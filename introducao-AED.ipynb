## Modelagem

A modelagem de dados envolve a criação de um modelo que pode descrever padrões complexos nos dados. Esses modelos podem então ser usados para fazer previsões sobre dados futuros ou para compreender melhor os fenômenos subjacentes aos dados. Nesse contexto, modelar significa também decidir quais dados serão coletados, ou mensarados, de forma que melhor representem o problema estudado. Ou seja, é durante a modelagem que definimos **o quê sera observado** (medido, coletado, mensurado, etc.). Isso implica definir quais varáveis representarão o problema estudado. 

Por exemplo, se estamos estudando os fatores que afetam o desempenho dos alunos em um curso universitário, nossa modelagem pode envolver a decisão de coletar dados sobre o tempo de estudo semanal dos alunos, sua frequência às aulas, se têm ou não um emprego paralelo, e assim por diante. Cada um desses é uma variável que pode ser medida e coletada. 

Nesse caso, a "nota final do curso" poderia ser a variável de interesse que estamos tentando prever ou explicar (também conhecida como a variável dependente). As variáveis que estamos coletando — tempo de estudo, frequência às aulas, emprego paralelo — são as variáveis que acreditamos que podem influenciar a variável de interesse (conhecidas como variáveis independentes). 

Portanto, a modelagem de dados neste contexto envolveria:

1. **Definir a questão de pesquisa ou o problema**: Neste caso, queremos entender os fatores que afetam o desempenho dos alunos em um curso universitário.

2. **Decidir quais dados coletar**: Baseado na questão de pesquisa, decidimos coletar dados sobre tempo de estudo, frequência às aulas e se o aluno tem ou não um emprego paralelo.

3. **Definir as variáveis**: A "nota final do curso" é nossa variável dependente, enquanto "tempo de estudo", "frequência às aulas" e "emprego paralelo" são nossas variáveis independentes.

4. **Coletar os dados**: Executamos a coleta de dados de acordo com o que foi definido.

5. **Analisar os dados**: Usamos técnicas estatísticas para analisar os dados e construir um modelo que possa prever a variável de interesse a partir das variáveis independentes.

Assim, a modelagem de dados é uma parte integral da AED, ajudando a determinar quais dados devem ser coletados e como eles serão analisados para responder à questão de pesquisa.


## Definição de Análise Exploratória de Dados

A Análise Exploratória de Dados (AED) é uma etapa elemntar na análise de dados que envolve uma série de técnicas para visualizar, resumir e entender a informação que nossos dados estão nos fornecendo. Esta fase da análise de dados é fundamental para gerar ideias (insights), testar hipóteses e verificar suposições.

A AED é uma abordagem que enfatiza a exploração visual dos dados, em vez de testes estatísticos formais ou modelagem preditiva. Seu objetivo principal é revelar a estrutura dos dados, detectar outliers ou anomalias, testar suposições, desenvolver modelos intuitivos e determinar a seleção adequada de ferramentas e técnicas para análises futuras como a Análise Inferencial; que realiza previsões ou inferências sobre uma população com base nos dados da amostra. 

Agora, vamos primeiro entender o que são variáveis e seus respectivos tipos.

# Variáveis

No contexto da Análise Exploratória de Dados, uma variável é uma característica, número ou quantidade que pode ser medida ou contada. As variáveis podem assumir vários valores e cada valor da variável pode corresponder a uma observação ou medida diferente.

As variáveis podem ser categorizadas principalmente em dois tipos: qualitativas e quantitativas.

## Tipos de Variáveis

**Qualitativas**. Também conhecidas como variáveis categóricas, elas são aquelas que expressam uma característica ou qualidade que não pode ser medida numericamente. Elas podem ser divididas em:

- **Variáveis Nominais**: não existe uma ordem ou ranking entre as categorias. Exemplos incluem gênero, cor dos olhos, tipo de material, etc.
   
- **Variáveis Ordinais**: as categorias têm uma ordem ou ranking específico. Exemplos incluem classificações de filmes, níveis de educação, etc.

**Quantitativas**. Estas são variáveis que podem ser medidas numericamente. Elas podem ser divididas em:

- **Variáveis Discretas**: são contáveis e normalmente são números inteiros. Exemplos incluem o número de animais de estimação em uma casa, o número de alunos em uma sala de aula, etc.
   
- **Variáveis Contínuas**: podem assumir qualquer valor dentro de um intervalo específico. Exemplos incluem altura, peso, tempo, temperatura, etc.

## Variáveis Dependentes e Independentes

Na análise de dados, muitas vezes estamos interessados em entender **se existe uma correlação** entre duas variáveis. A **variável dependente** é o que estamos tentando prever ou explicar, enquanto a **variável independente** é a que acreditamos ter algum efeito sobre a variável dependente. 

Por exemplo, a Gabriela tem uma loja de roupas. Ela quer saber com mais promoções, ela venderá mais ou menos roupas. Então, temos duas vairiáveis: porcentagem de promoção e quantidade de rouaps vendidas. A variável que ela quer prever é a quantidade de roupa vendida, logo esta será sua variável dependente.

Outro exemplo, se quisermos prever o desempenho de um aluno com base na quantidade de horas que ele estuda por semana, o desempenho do aluno seria a variável dependente, ou seja, **a variável que queremos prever ou estimar é a variável dependente**. Já a quantidade de horas de estudo seria a variável independente. Acreditamos que o desempenho do aluno (variável dependente) depende de, ou é influenciado por, a quantidade de horas que ele estuda (variável independente). Da mesma forma, em um experimento científico para testar os efeitos de um medicamento, a variável independente seria a administração do medicamento, enquanto a variável dependente poderia ser a melhora ou piora dos sintomas.

> **Exemplo**. Vamos prever uma variável contínua (o desempenho do aluno) com base em outra variável contínua (as horas de estudo). Podemos usar uma regressão linear simples para este exemplo. Aqui está um exemplo de código usando a biblioteca `sklearn` do Python.
> 
> No exemplo acima, estamos simulando um conjunto de dados onde o desempenho do aluno é proporcional às horas de estudo, com algum ruído adicionado para tornar os dados mais realistas. Em seguida, treinamos um modelo de regressão linear no conjunto de treinamento e fazemos previsões no conjunto de teste. Por fim, comparamos as previsões do modelo com os valores reais e plotamos a linha de regressão.

```python
# Importando as bibliotecas necessárias
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt
import pandas as pd

# Definindo a semente do gerador de números aleatórios para reprodutibilidade
np.random.seed(0)

# Gerando dados fictícios
horas_estudo = np.random.normal(10, 2, 100) # Gerando horas de estudo aleatórias
desempenho_aluno = horas_estudo * 2 + np.random.normal(0, 2, 100) # Desempenho é proporcional às horas de estudo, adicionamos algum ruído

# Reformatando os dados para usar com sklearn
horas_estudo = horas_estudo.reshape(-1,1)
desempenho_aluno = desempenho_aluno.reshape(-1,1)

# Dividindo os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(horas_estudo, desempenho_aluno, test_size=0.2, random_state=0)

# Treinando o modelo
regressor = LinearRegression()  
regressor.fit(X_train, y_train) # Treinando o algoritmo

# Fazendo previsões no conjunto de teste
y_pred = regressor.predict(X_test)

# Comparando a previsão com os dados reais
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
print(df)

# Plotando a linha de regressão
plt.scatter(X_test, y_test,  color='gray')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.show()
```

## População e Amostra

População é o conjunto completo de observações, indivíduos, itens ou eventos que queremos estudar, enquanto a amostra é um subconjunto dessa população. Por exemplo, se quisermos estudar a altura dos estudantes universitários no Brasil, a população seria todos os estudantes universitários no Brasil. Já uma amostra poderia ser os discentes matriculados na turma de Estatística do curso de Relações Públicas da Universidade Federal de Alagoas por exemplo.

> **Exemplo**. Suponha que temos a altura de todos os estudantes universitários no Brasil e queremos tirar uma amostra desses dados para estudos adicionais. Usaremos a biblioteca `numpy` para simular os dados da população e a função `random.choice` para selecionar uma amostra aleatória. Para fins de ilustração, suponhamos que a altura dos estudantes universitários no Brasil siga uma distribuição normal com uma média de 170 cm e um desvio padrão de 10 cm.
> 
> Estamos gerando (simulando) uma "população" de alturas de estudantes universitários e, então, tirmamos uma "amostra" dessa população. Depois, calculamos algumas medidas-resumo como média, mediana e moda da altura para a população e para a amostra.
> 
> Note que, neste exemplo de amostragem aleatória, a média da amostra pode não ser exatamente igual à média da população, devido à variação inerente ao processo de amostragem. No entanto, se a amostra for suficientemente grande e for selecionada aleatoriamente, a média da amostra deve ser uma boa estimativa da média da população.
>
> No exemplo a seguir, apresentamos também outros tipos de amostragem além da aleatória que serão estudados na próxima seção.

```python
import pandas as pd

def descreve_datasets(dataset1, dataset2=None, nome1='Dataset 1', nome2='Dataset 2', idioma='pt'):
    descricao1 = dataset1.describe()

    if dataset2 is not None:
        descricao2 = dataset2.describe()
        descricao = pd.concat([descricao1, descricao2], axis=1, keys=[nome1, nome2])
    else:
        descricao = descricao1.to_frame()
        descricao.columns = [nome1]

    if idioma == 'pt':
        descricao = descricao.rename(index={
            'count': 'contagem',
            'mean': 'média',
            'std': 'desvio padrão',
            'min': 'mínimo',
            '25%': 'Q1 (25%)',
            '50%': 'mediana',
            '75%': 'Q3 (75%)',
            'max': 'máximo'
        })
    elif idioma == 'en':
        descricao = descricao.rename(index={
            'count': 'count',
            'mean': 'mean',
            'std': 'standard deviation',
            'min': 'minimum',
            '25%': 'Q1 (25%)',
            '50%': 'median',
            '75%': 'Q3 (75%)',
            'max': 'maximum'
        })
    else:
        raise ValueError('Idioma não suportado')

    print(descricao.to_markdown() + "\n")


import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans

# Definindo a semente do gerador de números aleatórios para reprodutibilidade
np.random.seed(19)

# Gerando dados da população
populacao_tam = 100000
populacao = np.random.normal(170, 10, populacao_tam) # Suponha que temos 100000 estudantes universitários no Brasil
populacao_df = pd.DataFrame(populacao, columns=['altura'])

# Selecionando uma amostra para cada tipo de amostragem
amostra_tam = 500
# Amostragem Aleatória
amostra_aleatoria = np.random.choice(populacao, amostra_tam)

# Amostragem Sistematica
n = len(populacao_df) // amostra_tam # pega cada nth membro da população
# FIXME o plot estratificado n dá com intervalos iguais
print(" N é " + str(n))
amostra_sistematica = populacao_df.iloc[::n, :].values.flatten()

# Amostragem Estratificada
populacao_df['estrato'] = pd.cut(populacao_df['altura'], bins=[0, 170, np.inf], labels=['abaixo da média', 'acima da média'])
strat_train_set, strat_test_set = train_test_split(populacao_df, test_size=0.25, stratify=populacao_df['estrato'])
amostra_estratificada = strat_test_set['altura'].values

# Amostragem por Conglomerados
# kmeans = KMeans(n_clusters=10, random_state=0).fit(populacao_df[['altura']])
kmeans = KMeans(n_clusters=3, n_init=10, random_state=0).fit(populacao_df[['altura']])

populacao_df['cluster'] = kmeans.labels_
amostra_por_conglomerados = populacao_df[populacao_df['cluster'] == 0]['altura'].values

# Criando uma lista com todos os tipos de amostragem
tipos_de_amostragem = [("Amostragem Aleatória", amostra_aleatoria), 
                       ("Amostragem Sistemática", amostra_sistematica), 
                       ("Amostragem Estratificada", amostra_estratificada), 
                       ("Amostragem por Conglomerados", amostra_por_conglomerados)]

# Gerando um gráfico de histograma para cada tipo de amostragem
for tipo, amostra in tipos_de_amostragem:
    plt.figure(figsize=(12,6))
    plt.hist(populacao, bins=100, alpha=0.5, label='População')#, density=True)
    plt.hist(amostra, bins=100, alpha=0.5, label=tipo)#, density=True)

    # Adicionando a legenda
    plt.legend(loc='upper right')

    # Adicionando títulos e labels
    plt.title(f'Distribuição da Altura da População - {tipo}')
    plt.xlabel('Altura (cm)')
    plt.ylabel('Frequência')

    # Imprimindo statísticas de resumo da população e da amostra
    print("\n\n===========   " + tipo.upper() + "   =============================================================================================")
    print("\n------------ Medidas-Resumo ------------\n")
    descreve_datasets(pd.Series(populacao), pd.Series(amostra), nome1="População", nome2="Amostra", idioma='pt')

    # Mostrando o gráfico
    plt.show()
    print()
```

## Técnicas de Amostragem

A amostragem é o processo de selecionar uma subparte de uma população para estudo. Isso é feito para economizar tempo, diminuir custos e, muitas vezes, para viabilizar o estudo. Existem muitas técnicas de amostragem diferentes, cada uma com suas próprias vantagens e desvantagens. Selecionar a amostra apropriada é uma das etapas mais importantes no processo de análise de dados. Vários métodos de amostragem podem ser empregados, dependendo do tipo de dados, do objetivo da pesquisa e dos recursos disponíveis. Aqui estão algumas técnicas de amostragem comumente usadas.


1. **Amostragem Aleatória Simples**: Esta é a forma mais básica de amostragem. Em uma amostragem aleatória simples, cada membro da população tem igual probabilidade de ser selecionado para a amostra. Por exemplo, se tivermos uma população de 100 pessoas e quisermos selecionar uma amostra de 10, poderíamos simplesmente colocar todos os nomes em um chapéu e desenhar 10 deles aleatoriamente. A amostragem aleatória simples é frequentemente usada quando a população é homogênea, ou seja, não existem subgrupos específicos dentro da população que precisamos representar.

1. **Amostragem Sistemática**: A amostragem sistemática envolve a seleção de elementos em intervalos regulares através da lista da população. Por exemplo, se quisermos selecionar uma amostra de 100 pessoas de uma população de 1000, podemos selecionar a cada 10ª pessoa na lista.

1. **Amostragem Estratificada**: A amostragem estratificada é usada quando queremos garantir que subgrupos específicos da população sejam adequadamente representados na amostra. Nesta técnica, a população é dividida em estratos ou grupos homogêneos (por exemplo, gênero, faixa etária, etnia, etc.), e depois amostras aleatórias são retiradas de cada estrato. Por exemplo, se uma universidade quiser pesquisar opiniões dos alunos, ela pode decidir que quer a mesma quantidade de opiniões de estudantes de cada ano. Então, eles poderiam usar amostragem estratificada para garantir que selecionassem igual número de estudantes de primeiro, segundo, terceiro e quarto ano.

1. **Amostragem por Conglomerados**: Quando a população está dispersa geograficamente, a amostragem por conglomerados pode ser a mais eficiente. Neste método, a população é dividida em grupos ou "conglomerados" com base na localização geográfica, e então conglomerados inteiros são selecionados aleatoriamente para a amostra. Por exemplo, se uma empresa de pesquisa de mercado quisesse obter opiniões sobre um novo produto em todo o país, eles poderiam dividir o país em regiões e selecionar aleatoriamente algumas regiões para pesquisar.

## Erros de Amostragem

Ao trabalhar com amostras, é importante estar ciente de possíveis erros que podem surgir no processo de amostragem e que podem afetar a representatividade e precisão da amostra em relação à população. A seguir, apresentamos alguns tipos de erros de amostragem.

1. **Viés de seleção**: O viés de seleção ocorre quando a maneira como você seleciona sua amostra faz com que ela seja não representativa da população. Por exemplo, se você estiver realizando uma pesquisa online, sua amostra provavelmente será enviesada em direção a pessoas que têm acesso à Internet e que se sentem confortáveis em usá-la, o que pode não representar toda a população.

1. **Erro de amostragem**: Mesmo com um processo de seleção aleatória perfeito, ainda haverá um erro de amostragem, que é a diferença entre a verdadeira média da população e a média da amostra. De uma forma geral, quanto maior a amostra, menor será o erro de amostragem. Há um ponto de diminuição dos retornos, onde amostras maiores não reduzirão muito o erro de amostragem.

1. **Dados Atípicos e Outliers**: Dados típicos são valores que são significativamente diferentes da maioria dos outros valores na amostra. Dependendo do que você está estudando, os dados atípicos podem ser medidas naturalmente muito discrepantes, por exemplo a população das capitais brasileira: São Paulo e Rio de Janeiro são casos atípicos com quantidade de habitantes muito maior que as demais capitais. No exemplo anterior, a população São Paulo e Rio de Janeiro não se tratam de erro de mensiração, de digitação, etc., são valores simplesmentes raros. Quando os daos atípicos são provinientes de erros ou prováveis erros, chamamo-os de outliers, por exemplo, um erro de digitação ao inserir os dados. Em qualquer caso, dados atípicos podem distorcer os resultados e muitas vezes são excluídos da análise.

1. **Viés de não resposta**: Esse tipo de erro ocorre quando as respostas não são coletadas de toda a amostra. Por exemplo, se você enviar uma pesquisa por e-mail para 200 pessoas e apenas 50 responderem, você pode ter um viés de não resposta se as pessoas que não responderam forem sistematicamente diferentes daquelas que responderam.

# Medidas Resumo de Posição (Medidas de Tendência Central)

As medidas de posição são estatísticas que nos permitem entender os diferentes aspectos dos dados de uma perspectiva central. Em outras palavras, essas medidas tentam definir um "centro" ou "posição média" para os dados. As medidas de posição mais comuns são a média, a mediana e a moda.

## Média

Esta é provavelmente a medida de posição mais conhecida. A média de um conjunto de números é calculada somando-se todos os números e depois dividindo-se pelo número total de números. Por exemplo, supondo que tenhamos três pessoas cujas alturas são 150cm, 160am e 180cm, então a média seria calculada da seguinte forma: $média = ( 150 + 160 + 185) / 3 = 165$.

> **Exemplo**. Por exemplo, se tivermos as alturas de sete pessoas, demonstrada pela variável `heights` no exemplo a seguir, podemos calcular a média somando todas as alturas e divindo a soma por sete, como mostrado no código a seguir.

```python
import numpy as np

heights = np.array([150, 160, 165, 170, 180, 190, 190])
mean_height = np.mean(heights)
print(f"Média das alturas: {mean_height} cm")
```

## Mediana

A mediana é a medida que indica o valor central de um conjunto de números. Para encontrar a mediana, você organiza os números em ordem crescente e identifica o número no meio.

Por exemplo, no conjunto de alturas anterior, podemos calcular a mediana da seguinte maneira:

```python
median_height = np.median(heights)
print(f"Mediana das alturas: {median_height} cm")
```

A mediana é útil porque não é afetada por outliers, portanto, é uma medida de posição central mais robusta que a média em muitos casos.

## Moda

A moda é a medida que indica o valor que ocorre com mais frequência em um conjunto de dados. Por exemplo, baseando-se no conjunto de dados de alturas do exemplo a seguir, demonstrada pela variável `heights`, podemos calcular a moda da seguinte maneira:

```python
from scipy import stats

mode_height = stats.mode(heights)
print(f"Moda das alturas: {mode_height.mode[0]} cm")
```

A moda pode ser útil para entender a tendência ou a preferência mais comum entre um grupo. Em alguns casos, um conjunto de dados pode ter mais de uma moda (bimodal, multimodal) ou nenhuma moda (nenhum valor se repete).

Essas medidas de posição são fundamentais na Análise Exploratória de Dados, pois fornecem uma visão geral rápida do "centro" dos dados. No entanto, é importante lembrar que elas não fornecem um quadro completo e devem ser usadas em conjunto com outras medidas, como as medidas de dispersão, para uma compreensão mais completa dos dados. Medidas de Dispersão é o próximo assunto que estudaremos.



# Medidas Resumo de Dispersão

As medidas de dispersão, também conhecidas como medidas de variabilidade, mostram o quão espalhados os dados estão, ou como eles variam. Estas medidas incluem a amplitude, a variância, o desvio padrão e a faixa interquartil.

## Amplitude Total (Faixa Total)

A amplitude é a diferença entre o valor máximo e o valor mínimo em um conjunto de dados. É a medida de dispersão mais simples, mas é extremamente sensível a outliers.

> **Exemplo**. Seja um conjunto de dados de alturas geradas aleatoriamente com média = 170cm, desvio padrão = 10cm e quantidade de pessoas = 20 média. Essas alturas são representadas por barras em um gráfico, da menor para a maior, ilustrando visualmente a amplitude de alturas. Subtraindo a maior altura pela menor altura, temos a amplitude.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Criando e ordenando uma lista de alturas: média = 170cm, desvio padrão = 10cm, qutde. de elementos = 20
heights = np.random.normal(170, 5, 10)
sorted_heights = np.sort(heights)
range_height = np.ptp(heights)
range_height = round(range_height, 2)

# Plotando um gráfico de barras para ilustrar a amplitude: ORDENADA
plt.figure(figsize=(6,4))
sns.barplot(x=list(range(len(sorted_heights))), y=sorted_heights)
plt.ylim(140, 200)  # limites do eixo y para visualização clara
plt.title(f"Amplitude das Alturas = {range_height}")
plt.xlabel('Índice: Número de identificação da pessoa')
plt.ylabel('Altura (cm)')

# Adicionando linhas para mostrar a amplitude
min_height = min(np.round(sorted_heights, 2))
max_height = max(np.round(sorted_heights, 2))
plt.axhline(y=min_height, color='r', linestyle='--', label=f'Altura mínima = {min_height}')
plt.axhline(y=max_height, color='b', linestyle='--', label=f'Altura máxima = {max_height}')

# Adicionando legenda
plt.legend()

# Adicionando anotação para a amplitude
mid_index = len(sorted_heights) // 2
plt.annotate('', xy=(mid_index, min_height), xytext=(mid_index, max_height),
             arrowprops=dict(arrowstyle='<->', color='black'))
plt.text(mid_index + 1, (min_height + max_height)/2, f'AMPLITUDE = {range_height} cm', 
         rotation=90, verticalalignment='center', color='black')

plt.show()
```

## Variância e Desvio Padrão

A variância e o desvio padrão são medidas estatísticas fundamentais na compreensão da dispersão de um conjunto de dados, auxiliando a capturar a consistência, a variabilidade e a confiabilidade dos dados. A variância, em particular, é útil para **identificar o quanto os valores em um conjunto de dados divergem da média**. Um valor de variância ou de desvio padrão alto indica uma grande dispersão dos dados, o que pode sinalizar uma grande variabilidade no fenômeno que estamos medindo. 

Ou seja, essas medidas fornecem uma visão crucial sobre a estabilidade e a confiabilidade dos dados, permitindo decisões informadas em diversas áreas, desde investimentos financeiros até pesquisas científicas e controle de qualidade na produção. Mostraremos mais detalhes sobre a variância e o desvio padrão a seguir.

- A **variância** é uma medida da variabilidade que utiliza todas as observações de dados. Ela é calculada como a média das diferenças quadradas da média. A variância fornece um indicador de quão longe os valores estão espalhados a partir da média.
    - *Por exemplo, considere os conjuntos de números [4, 5, 6, 6, 6] e [1, 2, 6, 10, 15]. Embora ambos tenham a mesma média de 5,5, o primeiro conjunto tem uma variância menor, pois seus valores estão mais próximos da média, enquanto o segundo conjunto tem uma variância maior devido ao maior espalhamento dos valores em relação à média.*

- O **desvio padrão** é simplesmente a raiz quadrada da variância. Ele é útil porque, ao contrário da variância, está na mesma unidade que os dados. Isso facilita a interpretação e a comparação com a média.
    - *Continuando com o exemplo acima, as variâncias dos dois conjuntos são aproximadamente 0,8 e 32,3 respectivamente. No entanto, esses números estão em unidades quadradas (neste caso, "quadrados de números"), o que pode ser difícil de interpretar. Ao tirar a raiz quadrada desses valores para obter o desvio padrão, obtemos aproximadamente 0,9 e 5,7 respectivamente, o que nos dá uma ideia mais clara da dispersão dos dados em relação à média, na mesma unidade dos dados originais.*


> **Exemplo**. O gráfico a seguir, chamado de gráfico de dispersão, oferece uma representação visual dos conceitos de média e desvio padrão, além de mostrar a distribuição dos dados, isto é, das alturas.

> 3. **Área sombreada em vermelho (Desvio Padrão):** O desvio padrão é uma medida de dispersão que mostra o quanto as alturas variam em relação à média. A área sombreada em vermelho representa essa variação - cobrindo as alturas que estão a um desvio padrão acima e abaixo da média. Em uma distribuição normal, cerca de 68% dos dados se encontram dentro de um desvio padrão da média.

# Medidas Resumo de Quantis

Quantis são pontos estabelecidos em intervalos regulares da função de distribuição cumulativa de uma variável aleatória. De maneira mais simples, os quantis são valores derivados do conjunto de dados que dividem este conjunto de dados em partes iguais. As medidas de quantis mais comuns são os quartis, os decils e os percentis.

## Percentis

Percentis são medidas que dividem um conjunto de dados em 100 partes iguais. São muito úteis na estatística para entender a distribuição dos dados. Por exemplo, o percentil 50, também conhecido como mediana, é o valor no qual 50% dos dados são menores e 50% dos dados são maiores.

Os percentis são muito úteis para entender onde um ponto de dados específico se encaixa no contexto de todo o conjunto de dados. Por exemplo, se a altura de uma pessoa estiver no percentil 85, isso significa que essa pessoa é maior que 85% das demais.

## Quartis

Os quartis são um tipo específico de quantil que dividem um conjunto de dados ordenados em **quatro partes** iguais. Os quartis são úteis para fornecer uma visão rápida sobre a distribuição dos dados, identificar outliers e entender a dispersão dos dados.

Existem três quartis:

1. **Primeiro quartil (Q1)**: também conhecido como quartil inferior, representa o valor abaixo do qual se encontram **25%** dos dados quando ordenados do menor para o maior. Em outras palavras, Q1 é o valor que divide os primeiros 25% dos dados do restante.

2. **Segundo quartil (Q2)**: é a **mediana**, que divide o conjunto de dados ao meio, ou seja, **50%** dos dados são menores que Q2 e 50% dos dados são maiores.

3. **Terceiro quartil (Q3)**: também conhecido como quartil superior, é o valor abaixo do qual **75%** dos dados se encontram. Q3, portanto, separa os 25% superiores dos dados do restante.

## Medida de Dispersão de Quantis: Amplitude Interquartil (Faixa Interquartil)

A Amplitude Interquartil (também conhecida como Faixa Interquartil ou _Interquartile Range - IQR_) é uma medida de dispersão estatística que representa a diferença entre o terceiro quartil (Q3) e o primeiro quartil (Q1). Ou seja, ela abrange o intervalo entre os 25% e 75% dos dados quando ordenados do menor para o maior.

A Amplitude Interquartil mede a dispersão em torno da mediana, o que a torna uma medida de dispersão robusta, ou seja, ela é resistente a outliers ou valores extremos. A Amplitude Interquartil é útil para entender a variabilidade dos dados quando a distribuição não é simétrica ou quando há outliers significativos. 

A Amplitude Interquartil $d_q$ é calculada pela seguinte diferença:

$d_q = Q3 - Q1$.

Isso resulta em um valor que indica a "largura" da parte central dos dados, proporcionando uma medida da dispersão dos dados ao redor da mediana:
- Se a Amplitude Interquartil é grande, isso indica que os dados estão mais dispersos.
- Se for pequena, isso indica que os dados estão mais agrupados.

## Boxplot ou Diagrama de Caixas

Um gráfico de caixa, ou boxplot, é uma excelente froma de visualizar os quantis de um conjunto de dados. A "caixa" do gráfico contém a maioria dos dados (do Q1 ao Q3) e a "linha" dentro da caixa é a mediana. Os "bigodes" do gráfico representam o alcance dos dados, geralmente excluindo os outliers.

O boxplot representa visualmente muto informação estatística sobre um conjunto de dados como ilustra o exemplo a seguir.

> **Exemplo**. Continuando com nosso exemplo do conjunto de dados das alturas, vamos agora criar vários conjuntos de dados, cada um seguindo co uma distribuição de frequência diferente. Em seguida, vamos criar o gráfico da distribuição de frequências e outro para do boxplot. Perceba que ambos gráficos para cada conjunto de dados estão alinhados em relação à sua escala no eixo X., para facilitar a análise do boxpot.

```python
# import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def plot_hist_boxplot(heights_list, titles, colors):
    for i in range(len(heights_list)):
        print(f'\n\n\n=================================== {titles[i].upper()} ===================================\n')
        # Criando a figura e os subplots
        fig, axs = plt.subplots(2, 1, figsize=(8, 10), sharex=True, gridspec_kw={'height_ratios': [1, 4]})
        heights = heights_list[i]
    
        # Boxplot horizontal
        sns.boxplot(x=heights, ax=axs[1], orient='h', width=0.1, color=colors[i])  # Diminuir a largura da caixa
        axs[1].set_ylim(-0.1, 0.1)  # Ajustando o limite do eixo Y
        axs[1].set_xlabel('')  # Remover o rótulo do eixo x para o boxplot
        axs[1].set_title(f'Boxplot - {titles[i]}')
        axs[1].annotate('Q1 (1º Quartil)', xy=(np.percentile(heights, 25), -0.05), xytext=(-100, 30), 
                        textcoords='offset points', 
                        arrowprops=dict(facecolor='black', edgecolor='black', arrowstyle='->'))
        axs[1].annotate('Mediana (Q2)', xy=(np.median(heights), 0.05), xytext=(-100, -50), 
                        textcoords='offset points', 
                        arrowprops=dict(facecolor='black', edgecolor='black', arrowstyle='->'))
        axs[1].annotate('Q3 (3º Quartil)', xy=(np.percentile(heights, 75), -0.05), xytext=(-100, 50), 
                        textcoords='offset points', 
                        arrowprops=dict(facecolor='black', edgecolor='black', arrowstyle='->'))

        # Histograma
        sns.histplot(heights, ax=axs[0], bins=20, kde=False, color=colors[i]).set(xlabel='Altura (cm)', ylabel='Frequência')
        axs[0].set_title(f'Histograma - {titles[i]}')
    
        plt.xlabel('Altura (cm)')
        plt.tight_layout()  # Ajustar layout
        plt.show()

# Dados
heights_1 = np.random.normal(170, 5, 1000)  # Dados normalmente distribuídos
heights_2 = np.random.uniform(160, 180, 1000)  # Dados uniformemente distribuídos
heights_3 = np.random.exponential(10, 1000) + 160  # Dados exponencialmente distribuídos
heights_4 = np.random.lognormal(5, 0.5, 1000)  # Dados log-normalmente distribuídos
heights_5 = np.random.binomial(200, 0.5, 1000)  # Dados binomialmente distribuídos
heights_6 = np.random.poisson(5, 1000)  # Dados distribuídos segundo Poisson

# Distribuição bimodal
heights_7 = np.concatenate([np.random.normal(160, 5, 500), np.random.normal(180, 5, 500)])

# Lista de dados
heights_list = [heights_1, heights_2, heights_3, heights_4, heights_5, heights_6, heights_7]
titles = ['Distribuição Normal', 'Distribuição Uniforme', 'Distribuição Exponencial',
          'Distribuição Log-Normal', 'Distribuição Binomial', 'Distribuição de Poisson', 'Distribuição Bimodal']
# https://matplotlib.org/stable/gallery/color/named_colors.html
colors = ["lightblue", "lightpink", "aquamarine", "wheat", "lightsalmon", "plum", "lightgrey"]

plot_hist_boxplot(heights_list, titles, colors)
```

## Definição

A Distribuição de Frequências é uma representação tabular ou gráfica de dados que descreve a ocorrência de variáveis. Cada entrada na tabela ou gráfico contém a quantidade ou a percentagem de ocorrências de um determinado intervalo de valores dentro do conjunto de dados total. Este conceito é uma parte fundamental da estatística e ajuda a fornecer um entendimento básico sobre a natureza e a distribuição dos dados coletados.

Com relação ao tipo de variável, a Distribuição de Frequências apresentam as seguintes características:

- Variáveis Qualitativas: Para dados categóricos ou qualitativos, a distribuição de frequência contará o número de vezes que cada categoria aparece no conjunto de dados.
- Variáveis Quantitativas: Para dados quantitativos, a distribuição de frequência contará o número de dados dentro de certos intervalos.

| Intervalo de Notas | Frequência |
|--------------------|------------|
| 40-59              | 5          |

A interpretação dessa distribuição de frequências é a seguinte:

- 5 estudantes tiraram notas entre 40 e 59.
- 6 estudantes tiraram notas entre 60 e 79.
- 7 estudantes tiraram notas entre 80 e 100.


Portanto, uma distribuição de frequência, seja tabular ou gráfica, serve como uma ferramenta muito útil para resumir e interpretar grandes conjuntos de dados, tornando a informação compreensível e facilmente acessível. Geralmente, a representação da distribuição de frequências de grandes conjuntos de dados é feita através de gráficos de barras para melhor visualização. 


## Histogramas

Um histograma é uma representação gráfica de uma distribuição de frequência, amplamente utilizada para ilustrar a forma da distribuição dos dados. Ao utilizar retângulos de diferentes alturas, alinhados lado a lado, é possível visualizar com clareza as frequências (ou contagens) de dados dentro de intervalos específicos, também conhecidos como classes. A base de cada retângulo representa uma classe, enquanto a altura do retângulo indica a frequência dos dados nesse intervalo específico.

Para construir um histograma, os dados são primeiramente organizados em classes de igual largura. Estas classes são faixas de valores, e os dados são contados de acordo com a classe à qual pertencem. A escolha dos limites das classes pode ser crucial para a interpretação correta dos dados, e frequentemente, números redondos são usados para facilitar a leitura e interpretação do histograma.

O propósito de um histograma vai além da simples contagem de observações. Ele serve como uma ferramenta valiosa para visualizar e entender a distribuição subjacente dos dados. Ao observar a forma, a amplitude e as variações na altura dos retângulos do histograma, é possível identificar padrões, tais como simetria, assimetria, picos múltiplos (modas) e possíveis outliers.

Por exemplo, em um histograma que representa as notas de alunos em uma prova, é possível identificar rapidamente a concentração de notas, os valores mais frequentes e a variação na distribuição das notas. Isso pode oferecer infromações valiosos sobre o desempenho geral dos alunos, a dificuldade da prova, entre outros aspectos.

> **Exemplo** Vamos agora visualizar a mesma distribuição de frequências do exemplo anteior com dois gráficos: um gráficos de barras e um histograma. Perceba que, no histograma, a faixa de valor $50-89$ é a mais larga de todas, pois o histograma respeita a escala de valores dos intervalos de dados (faixas), diferente do gráfico de barras. Em outras palavras, o histograma apresenta mais informações sobre o conjunto de dados do que o gráfico de barras.

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados fornecidos
intervalos = ['30-49', '50-89', '90-100']
frequencia = [5, 6, 7]

# Cores para os intervalos
cores = ['red', 'green', 'blue']

# Convertendo os intervalos em uma lista de limites para os bins
limites = [int(intervalo.split('-')[0]) for intervalo in intervalos] + [int(intervalos[-1].split('-')[1])]

# Criando subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# Gráfico de barras
axs[0].bar(intervalos, frequencia, color=cores)
axs[0].set_title('Gráfico de Barras da Distribuição de Frequências das Notas dos Alunos')
axs[0].set_xlabel('Intervalo de Notas')
axs[0].set_ylabel('Frequência')

# Histograma - Criando um conjunto de dados baseado nos intervalos e frequências
for i, (inicio, fim) in enumerate(zip(limites[:-1], limites[1:])):
    dados = np.random.uniform(inicio, fim, frequencia[i])
    axs[1].hist(dados, bins=[inicio, fim], color=cores[i], edgecolor='black', alpha=0.7, rwidth=0.85)

axs[1].set_title('Histograma da Distribuição de Frequências das Notas dos Alunos')
axs[1].set_xlabel('Notas')
axs[1].set_ylabel('Frequência')

# Exibindo os gráficos
plt.tight_layout()
plt.show()
```

## Tipos de Distribuição de Frequências

Nesta seção, vamos aprofundar nos tipos de distribuições de frequência, discutindo sobre o comportamento das medidas resumo (como média, mediana e moda) em cada tipo de distribuição. Além disso, vamos relacionar essas distribuições com representações gráficas como o boxplot e as curvas de densidade de probabilidade. Curvas de densidade de probabilidade, ou simplesmente curvas de densidade, são uma representação suavizada do padrão subjacente dos dados em um histograma. Em outras palavras, essas curvas ajudam a visualizar a distribuição dos dados de uma maneira mais contínua e menos fragmentada do que as distribuições de frequência.


### Distribuição Uniforme

Em uma distribuição uniforme, todas as variáveis ou classes têm a mesma frequência ou probabilidade de ocorrência. Ou seja, em um conjunto de dados com distribuição uniforme, cada valor ocorre com a mesma frequência ou frequência similiar. Isso significa que a média, mediana e moda podem cair em qualquer ponto dentro de um memso intervalo, pois todos as observações têm frequências parecidas. 

O boxplot de uma distribuição uniforme mostrará uma caixa de tamanho uniforme se os dados estiverem perfeitamente distribuídos.

Na distribuição normal, a curva de densidade tem a forma clássica de sino, indicando que a maioria das observações está concentrada em torno da média, e se torna menos frequente à medida que se afasta da média em ambas as direções.

### Distribuição Normal

Uma distribuição normal é caracterizada pela sua forma de sino simétrica. Neste caso, a média, mediana e a moda são iguais e se localizam no centro do pico. 

O boxplot associado a uma distribuição normal será simétrico, indicando que os dados são distribuídos de maneira igual em torno da média, porém ele apresenta uma concentração maior entre Q1 e Q3 quando comparado com o boxplot da Distribuição Uniforme por exemplo.

Em uma distribuição uniforme, a curva de densidade seria uma linha reta, indicando que cada valor tem a mesma probabilidade de ocorrência.

### Distribuição Binomial

Em uma Distribuição Binomial, há apenas dois possíveis resultados ou classes, como "sucesso" e "fracasso". A média e mediana podem variar, mas geralmente ficam no centro se ambos os resultados forem igualmente prováveis. 

O boxplot pode variar dependendo das probabilidades de sucesso e fracasso, e a simetria será evidente se as probabilidades forem iguais.

Na distribuição binomial, a curva de densidade pode assumir diferentes formas dependendo da probabilidade de sucesso e do número de tentativas. Não é uma curva suave, mas mostra a probabilidade de cada número específico de sucessos.

### Distribuição Assimétrica (Skewness)

A Distribuição Assimétrica é inclinada, podendo ser positiva ou negativa. Em uma Distribuição Assimétrica positiva, a cauda à direita é mais longa, implicando que a média é maior que a mediana, que é, por sua vez, maior que a moda. Em uma Distribuição Assimétrica negativa, a cauda à esquerda é mais longa, indicando que a média é menor que a mediana, que é menor que a moda. 

O boxplot mostra claramente a assimaetria da Distribuição Assimétrica.

Em distribuições com skewness, a curva de densidade inclina-se para o lado onde os dados são menos frequentes, refletindo a assimetria dos dados.

### Distribuição Bimodal

Uma distribuição bimodal tem dois picos distintos, indicando que há duas modas. A média e a mediana situam-se entre esses dois picos. Na Distribuição Bimodal, a moda não serve como uma única medida representativa da centralidade dos dados, pois existem dois picos distintos ou modas. Nesses casos, é importante considerar ambas as modas para entender completamente a distribuição dos dados. 

O boxplot pode mostrar duas caixas distintas ou uma distância interquartil mais ampla, dependendo da separação das modas e da dispersão dos dados.

Em uma distribuição bimodal, a curva de densidade terá dois picos para representar os dois conjuntos distintos de dados dentro da amostra.


> **Exemplo**. Simulamos vários conjuntos de dados gerados aleatoriamente, cada um tem uma distribuição específica. Cada conjunto de dados é visualmente apresentada com seu histograma correspondente e boxplot . A partir desses gráficos, você pode observar claramente as características únicas de cada tipo de distribuição e como a média, a mediana e a moda, variam de acordo. Perceba que ambos gráficos para cada conjunto de dados estão alinhados em relação à sua escala no eixo X.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configuração para o estilo dos gráficos
sns.set(style="whitegrid")

# Dados de exemplo
data = {
    'Uniforme': np.random.rand(1000),
    'Normal': np.random.randn(1000),
    'Binomial': np.random.binomial(n=10, p=0.5, size=1000),
    'Positivamente Inclinada': np.random.exponential(scale=1, size=1000),
    'Bimodal': np.concatenate((np.random.normal(loc=5, scale=2, size=500),
                               np.random.normal(loc=10, scale=2, size=500)))
}

# Criando gráficos
fig, axes = plt.subplots(2, len(data.keys()), figsize=(15,10))

# Iterando através dos dados para criar histogramas e boxplots
for i, dist in enumerate(data.keys()):
    sns.histplot(data[dist], bins=20, kde=True, ax=axes[0, i]).set(title=f'Distribuição {dist}', xlabel='Observações', ylabel='Frequência')
    sns.boxplot(x=data[dist], ax=axes[1, i]).set(xlabel='Observações')


# Ajustando os layout
plt.tight_layout()
plt.show()

```

Quando construídos adequadamente, **os gráficos traduzem as complexidades numéricas em uma linguagem visual simples**, proporcionando percepções instantâneas, identificando padrões e indicando tendências sobre os conjunto de dados.

Ao longo das seções anteriores, utilizamos várias visualizações gráficas, adaptados para apresentar diferentes tipos de dados e informações. Da simplicidade dos gráficos de barras à riqueza informativa dos histogramas, por exemplo, cada tipo de gráfico serve a um propósito distinto, auxiliando na narrativa orientada a dados. Dessa maneira, a visualização de dados permite que os analistas, decisores e até o público em geral entendam os dados de forma mais eficiente.

## Tipos Essenciais de Gráficos e Suas Aplicações

A arte de visualizar dados envolve a escolha criteriosa dos formatos gráficos que melhor representam e comunicam as informações desejadas. Cada tipo de gráfico é intrínseco a determinadas formas de dados e objetivos de análise. Vejamos em detalhes alguns dos gráficos mais frequentemente empregados.

### Gráfico de Barras
Este gráfico é excepcional para representar e comparar quantidades ou frequências associadas a diferentes categorias. É visualmente intuitivo, tornando fácil a identificação de diferenças significativas entre categorias.

**Aplicação Prática:** Se desejarmos comparar o número de vendas de diferentes produtos em um mês, o gráfico de barras oferece uma visualização clara e concisa das vendas relativas.

```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C']
vendas = [150, 200, 175]

plt.bar(produtos, vendas, color=['blue', 'green', 'red'])
plt.title('Vendas de Produtos em Janeiro')
plt.xlabel('Produtos')
plt.ylabel('Número de Vendas')
plt.show()
```

### Gráfico de Setores (Pizza)
Apesar de comum, o uso deve ser moderado para evitar complexidade visual. É eficiente para mostrar proporções relativas de categorias em um conjunto total.

**Aplicação Prática:** Ao visualizar a distribuição percentual de respostas de uma pesquisa de opinião, o gráfico de pizza destaca a proporção de cada resposta.

```python
import matplotlib.pyplot as plt
import squarify

# Dados existentes
respostas = ['Concordo', 'Discordo', 'Neutro']
percentuais = [50, 30, 20]
cores = ['gold', 'lightskyblue', 'lightcoral']

# Gráfico de pizza
plt.subplot(121)
plt.pie(percentuais, labels=respostas, colors=cores, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Gráfico de Pizza')

# Treemap
plt.subplot(122)
squarify.plot(sizes=percentuais, label=respostas, color=cores, alpha=0.7, text_kwargs={'fontsize':10})
plt.title('Treemap')
plt.axis('off')  # Desativa os eixos para o treemap

# Ajusta o layout para evitar sobreposição
plt.tight_layout()

# Mostra os gráficos
plt.show()
```

### Gráfico de Linhas
Este é indispensável para representar tendências temporais, mostrando a evolução de variáveis numéricas ao longo do tempo.

**Aplicação Prática:** Utilizado para traçar a trajetória de crescimento das vendas anuais, oferecendo insights sobre padrões sazonais ou anuais.

```python
anos = range(2018, 2023)
vendas_anuais = [200, 250, 300, 350, 400]

plt.plot(anos, vendas_anuais, marker='o', linestyle='-', color='purple')
plt.title('Crescimento das Vendas Anuais')
plt.xlabel('Ano')
plt.ylabel('Vendas (em milhares)')
plt.grid(True)
plt.show()
```

### Gráfico de Dispersão
Ideal para explorar e visualizar relações entre duas variáveis contínuas, facilitando a identificação de correlações.

**Aplicação Prática:** Ao analisar a relação entre despesas com publicidade e receita gerada, o gráfico de dispersão proporciona uma visão clara da correlação.

```python
despesas_publicidade = [50, 100, 150, 200, 250]
receita = [200, 400, 450, 600, 700]

plt.scatter(despesas_publicidade, receita, color='green')
plt.title('Relação entre Despesas com Publicidade e Receita')
plt.xlabel('Despesas com Publicidade (em milhares)')
plt.ylabel('Receita (em milhares)')
plt.grid(True)
plt.show()
```

### Histograma
Um recurso poderoso para apresentar a distribuição de dados numéricos, oferecendo insights sobre a forma, amplitude e centralidade dos dados.

**Aplicação Prática:** Utilizado para entender a distribuição de pontuações em um teste, facilitando a identificação de padrões de desempenho dos estudantes.

```python
import numpy as np

pontuacoes = np.array([55, 60, 65, 70, 70, 75, 80, 85, 85, 90, 95, 100])

plt.hist(pontuacoes, bins=9, color='orange', alpha=0.7)
plt.title('Distribuição das Pontuações do Teste')
plt.xlabel('Pontuação')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

### Boxplot
Este gráfico oferece uma representação visual compacta das medidas estatísticas centrais e da dispersão, sendo altamente eficaz para identificar outliers.

**Aplicação Prática:** Em estudos clínicos, pode ser utilizado para visualizar a variação dos níveis de colesterol entre diferentes grupos etários.

```python
import seaborn as sns
import pandas as pd

colesterol = [180, 200, 220, 240, 260, 220, 205, 210, 180, 230]
grupo_etario = ['Jovens'] * 5 + ['Idosos'] * 5

data = pd.DataFrame({'Colesterol': colesterol, 'Grupo Etário': grupo_etario})
sns.boxplot(x='Grupo Etário', y='Colesterol', data=data)
plt.title('Variação dos Níveis de Colesterol entre Grupos Etários')
plt.show()
```

## Estratégias para Seleção Adequada de Gráficos

A acuidade na seleção do gráfico certo é um passo fundamental na comunicação eficaz dos dados. A seguir, estão estratégias para a escolha de um gráfico:

- **Quantidades e Proporções:** Gráficos de barras e setores são ideais para representar e comparar quantidades ou proporções distintas entre diferentes categorias.
- **Tendências Temporais:** Prefira gráficos de linhas, pois eles são excepcionais para representar a evolução dos dados ao longo do tempo. 
- **Relações entre Variáveis:** Gráficos de dispersão, pois conseguimos dentificar padrões e correlações entre variáveis.
- **Distribuição de Dados:**  Histogramas são fundamentais para desvendar a estrutura dos dados e identificar padrões.
- **Dispersão e Centralidade:**  Boxplots sintetizam estes elementos eficientemente, oferecendo uma visão concisa das medidas centrais e da dispersão dos dados. 

Embora essas diretrizes sejam úteis, é crucial considerar o público-alvo e o contexto. A complexidade, a interpretabilidade e a estética visual desempenham papéis cruciais na eficácia da comunicação dos dados. Sempre teste e revise suas visualizações para garantir que sejam intuitivas, informativas e engajadoras para o seu público específico.

Por fim, a escolha de um gráfico possuem também um caráter subjetivo e algumas perguntas podem nos ajudar a nortear a escolha dos gráficos:

- Este gráfico ajuda a responder às minhas perguntas?
- Ele representa claramente os dados?
- As informações são facilmente interpretáveis para o público-alvo?
