# 2. Modelagem

Para realizarmos uma An√°lise Explorat√≥ria de Dados, o primeiro passo √© **compreender o problema** e **modela-lo**. A modelagem de dados envolve a cria√ß√£o de um modelo que pode descrever padr√µes complexos nos dados. Esses modelos podem ent√£o ser usados para fazer previs√µes sobre dados futuros ou para compreender melhor os fen√¥menos subjacentes aos dados. 

Nesse contexto, modelar significa tamb√©m decidir quais dados ser√£o coletados, ou mensarados, de forma que melhor representem o problema estudado. Ou seja, √© durante a modelagem que definimos **o qu√™ sera observado** (medido, coletado, mensurado, etc.). Isso implica definir quais var√°veis representar√£o o problema estudado. 

Por exemplo, se estamos estudando os fatores que afetam o desempenho dos alunos em um curso universit√°rio, nossa modelagem pode envolver a decis√£o de coletar dados sobre o tempo de estudo semanal dos alunos, sua frequ√™ncia √†s aulas, se t√™m ou n√£o um emprego paralelo, e assim por diante. Cada um desses √© uma vari√°vel que pode ser medida e coletada. 

Nesse caso, a "nota final do curso" poderia ser a vari√°vel de interesse que estamos tentando prever ou explicar (tamb√©m conhecida como a vari√°vel dependente). As vari√°veis que estamos coletando ‚Äî tempo de estudo, frequ√™ncia √†s aulas, emprego paralelo ‚Äî s√£o as vari√°veis que acreditamos que podem influenciar a vari√°vel de interesse (conhecidas como vari√°veis independentes). 

Portanto, a modelagem de dados neste contexto envolveria:

1. **Definir a quest√£o de pesquisa ou o problema**: Neste caso, queremos entender os fatores que afetam o desempenho dos alunos em um curso universit√°rio.

2. **Decidir quais dados coletar**: Baseado na quest√£o de pesquisa, decidimos coletar dados sobre tempo de estudo, frequ√™ncia √†s aulas e se o aluno tem ou n√£o um emprego paralelo.

3. **Definir as vari√°veis**: A "nota final do curso" √© nossa vari√°vel dependente, enquanto "tempo de estudo", "frequ√™ncia √†s aulas" e "emprego paralelo" s√£o nossas vari√°veis independentes.

4. **Coletar os dados**: Executamos a coleta de dados de acordo com o que foi definido.

5. **Analisar os dados**: Usamos t√©cnicas estat√≠sticas para analisar os dados e construir um modelo que possa prever a vari√°vel de interesse a partir das vari√°veis independentes.

Assim, a modelagem de dados √© uma parte integral da AED, ajudando a determinar quais dados devem ser coletados e como eles ser√£o analisados para responder √† quest√£o de pesquisa.

## 3.2 Depend√™ncia entre Vari√°veis

Na an√°lise de dados, muitas vezes estamos interessados em entender **se existe uma correla√ß√£o** entre duas ou mais vari√°veis. Para isso, precisamos criar hip√≥teses sobre qual(is) vari√°vel(is) influencia(m) no comportamento de outra(s). Isso requer definir qual(is) ser√°(√£o) a(s) vari√°vel(is) dependente(s) e independente(s). Finalmente, podemos compreender se e como essas vari√°veis se correlacionam. 

Para podermos fazer essa an√°lise, vamos primeiro compreender o que s√£o as Vari√°veis Dependentes e Independentes.

### Vari√°veis Dependentes e Independentes

Analisar a depend√™ncia entre vari√°veis quer dizer tentar compreender como uma vari√°vel (ou mais) influencia em outra(s) vari√°vel(is). Para isso, modelamos como a **vari√°vel dependente** a vari√°vel que achamos que √© influenciada por outra.  Por outro lado, a **vari√°vel independente** √© a que acreditamos ter algum efeito sobre a vari√°vel dependente, isto √©, a que influencia a vari√°vel dependente fazendo com que os valores da independente alterem os valores da dependente. 

Por exemplo,  oder√≠amos considerar a "quantidade de conte√∫do publicado em redes sociais" como uma vari√°vel independente e a "quantidade de novos seguidores", representadno o n√≠vel de engajamento do p√∫blico, como uma vari√°vel dependente. Nesse contexto, a hip√≥tese seria que a frequ√™ncia de publica√ß√µes (vari√°vel independente) influencia o engajamento do p√∫blico (vari√°vel dependente), seja aumentando-o ou diminuindo-o.


````python
import numpy as np
import matplotlib.pyplot as plt

# Gerar dados simulados
np.random.seed(0)
quantidade_postagens = np.linspace(1, 30, 100)  # De 1 a 30 postagens
engajamento = np.square(quantidade_postagens) / (0.1 * quantidade_postagens + 1) 
engajamento += np.random.normal(0, 30, 100)  # Adicionando um pouco de ru√≠do

# Criar o gr√°fico
plt.figure(figsize=(10, 6))
plt.scatter(quantidade_postagens, engajamento, color='blue')
plt.title('Rela√ß√£o entre Quantidade de Conte√∫do Publicado e Engajamento do P√∫blico')
plt.xlabel('Quantidade de Conte√∫do Publicado (Postagens por Dia)')
plt.ylabel('Qtde. de Novos Seguidores (Por Dia)')
plt.grid(True)
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Gerando dados fict√≠cios
np.random.seed(0)
nof_days = 20
datas = pd.date_range('2023-01-01', periods=nof_days, freq='D')
frequencia_postagens = np.random.randint(1, 4, size=nof_days)  # 1 a 3 postagens por dia
engajamento = frequencia_postagens * np.random.uniform(0.8, 1.2, size=nof_days)  # Engajamento relacionado √† frequ√™ncia

# Criando um DataFrame
dados = pd.DataFrame({'Data': datas, 'Frequ√™ncia de Postagens': frequencia_postagens, 'Engajamento': engajamento})

# Gr√°fico
# Ajustando o gr√°fico para ter escalas diferentes no eixo Y

plt.figure(figsize=(12, 3))

# Criando o primeiro eixo (para frequ√™ncia de postagens)
ax1 = plt.gca()  # Obt√©m o eixo atual
ax1.plot(dados['Data'], dados['Frequ√™ncia de Postagens'], label='Frequ√™ncia de Postagens', color='blue')
ax1.set_ylabel('Frequ√™ncia de Postagens', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Criando um segundo eixo (para engajamento) compartilhando o mesmo eixo X
ax2 = ax1.twinx()  # Cria um segundo eixo Y que compartilha o mesmo eixo X
ax2.plot(dados['Data'], dados['Engajamento'], label='Engajamento', color='green')
ax2.set_ylabel('Engajamento', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Configura√ß√µes adicionais do gr√°fico
plt.title('Correla√ß√£o entre Frequ√™ncia de Postagens e Engajamento em M√≠dias Sociais')
ax1.set_xlabel('Data')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
````


Agora que entendemos que duas ou mais vari√°veis podem estar correlacionadas, precisamos de mais precis√£o na nossa an√°lise. H√° v√°rios m√©todos para quantificar a correla√ß√£o entre vari√°veis. Apresentamos aqui duas formas de calcular o grau de correla√ß√µes entre vari√°aveis de acordo com a natureza da rela√ß√£o:


- Coeficiente de Pearson ($r$): para correla√ß√µes lineares
- Coeficiente de correla√ß√£o de Kendall ($\tau$): para correla√ß√µes n√£o lineares

### Coeficiente de Pearson: Medindo uma Correla√ß√£o Linear

O [Coeficiente de Correla√ß√£o de Pearson](https://en.wikipedia.org/wiki/Pearson_correlation_coefficient), denotado por $r$, √© uma medida estat√≠stica que quantifica a rela√ß√£o linear entre duas vari√°veis quantitativas. O Ele √© utilizado em diversas √°reas para medir a intensidade e a dire√ß√£o de uma rela√ß√£o linear entre duas vari√°veis. O Coeficiente de Correla√ß√£o de Pearson varia entre $-1$ e $+1$, onde:

- **$+1$** indica uma correla√ß√£o positiva perfeita (quando uma vari√°vel aumenta, a outra tamb√©m aumenta na mesma propor√ß√£o).
- **$0$** indica que n√£o h√° correla√ß√£o linear entre as vari√°veis.
- **$-1$** indica uma correla√ß√£o negativa perfeita (quando uma vari√°vel aumenta, a outra diminui na mesma propor√ß√£o).

Interpreta√ß√£o do Coeficiente:

- **0.9 a 1.0 (-0.9 a -1.0)**: Indica uma correla√ß√£o muito forte. Valores pr√≥ximos a 1 ou -1 mostram que as vari√°veis t√™m uma rela√ß√£o linear quase perfeita, sendo positiva ou negativa, respectivamente.
- **0.7 a 0.9 (-0.7 a -0.9)**: Representa uma correla√ß√£o forte, sugerindo uma rela√ß√£o linear significativa entre as vari√°veis.
- **0.5 a 0.7 (-0.5 a -0.7)**: Correla√ß√£o moderada, indicando uma rela√ß√£o linear razo√°vel, mas n√£o extremamente forte.
- **0.3 a 0.5 (-0.3 a -0.5)**: Uma correla√ß√£o fraca, que sugere uma rela√ß√£o linear mais fraca e menos consistente.
- **0 a 0.3 (0 a -0.3)**: Correla√ß√£o desprez√≠vel ou nenhuma correla√ß√£o, indicando pouca ou nenhuma rela√ß√£o linear entre as vari√°veis.

√â importante notar que esses intervalos s√£o orienta√ß√µes gerais e podem variar em diferentes contextos. Al√©m disso, o Coeficiente de Pearson n√£o detecta rela√ß√µes n√£o lineares e √© sens√≠vel a outliers, o que pode afetar a interpreta√ß√£o em certas situa√ß√µes de dados.

**Limita√ß√µes**:

- **Somente para Rela√ß√µes Lineares**: O coeficiente de Pearson s√≥ mede a for√ßa de uma rela√ß√£o linear. Rela√ß√µes n√£o lineares podem n√£o ser bem representadas.
- **Influ√™ncia de Valores Extremos**: Outliers podem distorcer significativamente o resultado do coeficiente.
- **N√£o Implica Causalidade**: Uma correla√ß√£o alta (positiva ou negativa) n√£o implica que uma vari√°vel cause mudan√ßas na outra.

Por fim, Coeficiente de Correla√ß√£o de Pearson √© muito √∫til para identificar e quantificar correla√ß√µes lineares entre vari√°veis. No entanto, √© importante sempre considerar outras an√°lises para compreender completamente a natureza das rela√ß√µes entre vari√°veis como, por exemplo, o coeficiente de Kendall que estudaremos a seguir para computar correla√ß√µes n√£o lineares.

### Coeficiente de Kendall: Medindo uma Correla√ß√£o N√£o Linear

O Coeficiente de Correla√ß√£o de Kendall, frequentemente representado por $\tau$, √© uma medida estat√≠stica que ideintifica e quantifica a for√ßa e a dire√ß√£o de uma rela√ß√£o n√£o linear entre duas vari√°veis. Diferentemente do Coeficiente de Pearson, que mede correla√ß√µes lineares, o Coeficiente de Kendall √© mais adequado para identificar rela√ß√µes, sejam elas lineares ou n√£o. O Coeficiente de Kendall mede a concord√¢ncia entre os pares de observa√ß√µes, variando entre -1 e +1, onde:

- **+1** indica uma correla√ß√£o positiva perfeita (todas as observa√ß√µes concordam).
- **0** indica que n√£o h√° correla√ß√£o n√£o linear entre as vari√°veis.
- **-1** indica uma correla√ß√£o negativa perfeita (todas as observa√ß√µes discordam).

**Interpreta√ß√£o do Coeficiente**:

A interpreta√ß√£o do Coeficiente de Kendal baseia-se na magnitude do valor, que indica a for√ßa da associa√ß√£o n√£o linear entre as vari√°veis. Os intervalos t√≠picos para interpreta√ß√£o podem ser:

- $-1$ a $-0.5$: Correla√ß√£o negativa forte. Indica uma forte tend√™ncia de que, quando uma vari√°vel aumenta, a outra diminui de forma n√£o linear.
- $-0.5$ a $-0.2$: Correla√ß√£o negativa moderada. Uma tend√™ncia menos pronunciada de que as vari√°veis se movem em dire√ß√µes opostas.
$-0.2$ a $0.2$: Correla√ß√£o fraca ou nenhuma correla√ß√£o. Sinaliza uma associa√ß√£o muito fraca ou inexistente entre as vari√°veis.
- $0.2$ a $0.5$: Correla√ß√£o positiva moderada. Mostra uma tend√™ncia moderada de que as vari√°veis se movem juntas, mas n√£o de forma linear.
- $0.5$ a $1$: Correla√ß√£o positiva forte. Indica uma forte associa√ß√£o na qual as vari√°veis tendem a aumentar ou diminuir juntas, embora a rela√ß√£o n√£o seja linear.

√â importante ressaltar que esses intervalos s√£o guias gerais e podem variar dependendo do contexto espec√≠fico e da natureza dos dados. Al√©m disso, √© essencial considerar outros fatores, como o tamanho da amostra e a distribui√ß√£o das vari√°veis, ao interpretar o Coeficiente de Kendal.

**Limita√ß√µes**:

- **Complexidade Computacional**: Pode ser computacionalmente mais intensivo, especialmente para grandes conjuntos de dados.
- **Limitado a Rela√ß√µes Monot√¥nicas**: N√£o detecta bem rela√ß√µes n√£o monot√¥nicas.
- **N√£o Indica Causalidade**: Como com Pearson, uma correla√ß√£o alta (positiva ou negativa) n√£o implica causalidade.

O Coeficiente de Kendall √© √∫til para analisar rela√ß√µes entre vari√°veis em muitas √°reas, como comunica√ß√£o social, biologia e economia. √â mais √∫til ainda **quando as rela√ß√µes entre as vari√°veis s√£o suspeitas de serem n√£o lineares ou quando os dados cont√™m outliers**. Por fim, ao usar o Coeficiente de Kendall, √© importante considerar o contexto dos dados e complementar a an√°lise com outras medidas e m√©todos estat√≠sticos para obter uma compreens√£o completa das rela√ß√µes em estudo. ## 4.1 Popula√ß√£o e Amostra

Segundo Bussab e Morettin (2017), _"Popula√ß√£o √© o conjunto de todos os elementos ou resultados sob investiga√ß√£o. Amostra √© qualquer subconjunto da popula√ß√£o."_ [1]. Quando uma pesquisa √© feita com toda a popula√ß√£o, ela √© chamada de **pesquisa censit√°ria**. Pesquisas censit√°rias nos trazem uma vis√£o completa sobre o estudo, pois todos os dados da popula√ß√£o foram coeltados. Todavia, √© muito custoso ou invi√°vel realizar tal pesquisa. Por exemplo, ao realizar um censo,entrevistando todos os consumidores de um produto, para realizar uma pesquisa de de opini√£o p√∫blica, poder custar muito caro ou ser fisicamente invi√°vel. √â o caso de uma institui√ß√£o que s√≥ tem um profissional para realizar a pesquisa de opini√£o p√∫blica: √© custoso contratar mais profissionais para entrevistar todos os e √© invi√°vel que essa pessoa sozinha consiga entrevistar todos os consumidores.

## 4.2 Tipos de Amostragem

Um fator essencial na an√°lise de dados √© a sele√ß√£o de **amostras representativas**, ou seja, amostras que contenham partes quantitativamente equivalentes √†s suas propor√ß√µes na popula√ß√£o. Dessa forma, aumentamos as chances de que os resultados obtidos da an√°lise das amostras sejam correspondam bem √† representa√ß√£o da popula√ß√£o. Em outras palavras, a amostra deve incluir uma variedade de elementos que reflitam a diversidade da popula√ß√£o total de forma que as propor√ß√µes de diferentes subgrupos dentro da amostra devem corresponder √†s da popula√ß√£o.

Podemos dividir as os tipos de amostragem entre probabil√≠sticos e n√£o probail√≠sticos (Anuncia√ß√£o, 2021), conforme descritas a seguir:

- **Amostragem Probabil√≠stica**: cada membro da popula√ß√£o √© escolhido aleatoriamente e sabemos qual √© a probabilidade de cada membro ser selecionado para a amostra. Isso garante uma representatividade mais precisa da popula√ß√£o total, permitindo generaliza√ß√µes mais confi√°veis. 

- **Amostragem N√£o Probabil√≠stica**: sele√ß√£o intencional de membros para a amostra, sem usar um processo aleat√≥rio. Isso pode incluir escolhas baseadas em crit√©rios espec√≠ficos ou conveni√™ncia.

A Tabela a seguir, adaptada de (Anuncia√ß√£o, 2021), compila as vantagens de cada tipo de amostragem:

| Crit√©rio             | Probabil√≠stica                           | N√£o-Probabil√≠stica                       |
|----------------------|------------------------------------------|-----------------------------------------|
| Sele√ß√£o              | Aleat√≥ria, cada membro tem uma probabilidade conhecida | Intencional, sem processo aleat√≥rio         |
| Representatividade   | Alta                                     | Baixa                                    |

| Generaliza√ß√£o        | Poss√≠vel e confi√°vel                     | Limitada                                 |
| Aplica√ß√µes           | Pesquisas amplas, estudos demogr√°ficos   | Pesquisas explorat√≥rias, estudos de caso |
| Exemplos em RP       | An√°lise demogr√°fica de um p√∫blico-alvo para uma campanha de conscientiza√ß√£o | Pesquisa de opini√£o de clientes em um evento promocional espec√≠fico |

- **Amostragem Probabil√≠stica em RP**: A complexidade da amostragem probabil√≠stica em Rela√ß√µes P√∫blicas pode ser ilustrada por um projeto como a realiza√ß√£o de uma an√°lise demogr√°fica detalhada do p√∫blico-alvo para uma campanha de conscientiza√ß√£o. Este processo exigiria primeiro a defini√ß√£o clara da popula√ß√£o-alvo, seguida pela cria√ß√£o de um mecanismo aleat√≥rio para a sele√ß√£o de indiv√≠duos, garantindo que cada membro da popula√ß√£o tenha uma chance conhecida e igual de ser escolhido. Isso pode envolver o uso de softwares de randomiza√ß√£o e a necessidade de acessar ou criar uma lista abrangente da popula√ß√£o. Al√©m disso, para garantir a representatividade, a amostra teria que ser suficientemente grande e bem distribu√≠da, considerando v√°rias vari√°veis demogr√°ficas, como idade, g√™nero, localiza√ß√£o geogr√°fica, entre outras. - **Amostragem N√£o Probabil√≠stica em RP**: Em contraste, a amostragem n√£o-probabil√≠stica, embora menos complexa, apresenta seus pr√≥prios desafios. A sele√ß√£o intencional e a natureza oportunista dessa amostragem simplificam a coleta de dados, mas requerem cuidado na interpreta√ß√£o e aplica√ß√£o dos resultados obtidos.

3. **Amostragem Estratificada**: Utilizada para assegurar que todos os segmentos importantes da popula√ß√£o sejam representados como, por exemplo, idade, g√™nero ou localiza√ß√£o geogr√°fica. Em Rela√ß√µes P√∫blicas, isso poderia ser usado para garantir representatividade em uma pesquisa sobre a percep√ß√£o de uma campanha publicit√°ria. 4. **Amostragem por Conglomerados**: Esta t√©cnica pode ser usada para facilitar a realizar pesquisas de mercado em diferentes localiza√ß√µes geogr√°ficas. Por exemplo, se uma ag√™ncia de RP quiser avaliar a percep√ß√£o de uma marca em diferentes cidades, ela pode dividir o mercado em conglomerados (regi√µes) e selecionar algumas dessas regi√µes para realizar a pesquisa. Isso permite uma compreens√£o mais abrangente da percep√ß√£o regional da marca.

5. **Amostragem por Quotas**: Esta t√©cnica envolve selecionar amostras com base em quotas pr√©-definidas para diferentes subgrupos dentro da popula√ß√£o. O objetivo √© garantir que caracter√≠sticas espec√≠ficas sejam representadas na amostra. Por exemplo, em uma campanha de RP voltada para a conscientiza√ß√£o sobre sa√∫de, pode ser importante incluir quotas espec√≠ficas de homens e mulheres de diferentes faixas et√°rias. Assim, a amostragem por quotas asseguraria a inclus√£o proporcional desses grupos para obter uma vis√£o abrangente das diferentes percep√ß√µes e atitudes em rela√ß√£o √† campanha.

6. **Amostragem de Conveni√™ncia**: Esta abordagem utiliza indiv√≠duos facilmente acess√≠veis ou dispon√≠veis para a pesquisa. √â uma forma r√°pida e econ√¥mica de coletar dados, mas com poucas chances de ser representativa da popula√ß√£o total. Por exemplo, durante o lan√ßamento de um novo produto, um profissional de RP pode realizar uma pesquisa r√°pida com os participantes do evento de lan√ßamento. Essa amostra de conveni√™ncia permite coletar feedback instant√¢neo sobre o produto, embora os participantes possam n√£o representar todos os potenciais consumidores do produto.

7. **Amostragem de Bola de Neve**: Utilizada para acessar e estudar popula√ß√µes que s√£o dif√≠ceis de identificar ou alcan√ßar. Come√ßa com um pequeno grupo de participantes conhecidos que, por sua vez, indicam outros membros de sua rede. Por exemplo, ao estudar a influ√™ncia de opini√µes de l√≠deres de pensamento em um nicho de mercado espec√≠fico, um profissional de RP pode come√ßar entrevistando alguns influenciadores conhecidos nesse nicho. Esses influenciadores, por sua vez, podem indicar outros colegas ou contatos relevantes, ampliando a pesquisa atrav√©s da rede de contatos. Por isso o √© conhecido como "bola de neve".

````python
import pandas as pd

def descreve_datasets(dataset1, dataset2=None, nome1='Dataset 1', nome2='Dataset 2', idioma='pt'):
    descricao1 = dataset1.describe()

    if dataset2 is not None:
        descricao2 = dataset2.describe()
        descricao = pd.concat([descricao1, descricao2], axis=1, keys=[nome1, nome2])
    else:
        descricao = descricao1.to_frame()
        descricao.columns = [nome1]

    if idioma == 'pt':
        descricao = descricao.rename(index={
            'count': 'contagem',
            'mean': 'm√©dia',
            'std': 'desvio padr√£o',
            'min': 'm√≠nimo',
            '25%': 'Q1 (25%)',
            '50%': 'mediana',
            '75%': 'Q3 (75%)',
            'max': 'm√°ximo'
        })
    elif idioma == 'en':
        descricao = descricao.rename(index={
            'count': 'count',
            'mean': 'mean',
            'std': 'standard deviation',
            'min': 'minimum',
            '25%': 'Q1 (25%)',
            '50%': 'median',
            '75%': 'Q3 (75%)',
            'max': 'maximum'
        })
    else:
        raise ValueError('Idioma n√£o suportado')

    print(descricao.to_markdown() + "\n")


import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans

# Definindo a semente do gerador de n√∫meros aleat√≥rios para reprodutibilidade
np.random.seed(19)

# Gerando dados da popula√ß√£o
populacao_tam = 100000
populacao = np.random.normal(170, 10, populacao_tam) # Suponha que temos 100000 estudantes universit√°rios no Brasil
populacao_df = pd.DataFrame(populacao, columns=['altura'])

# Selecionando uma amostra para cada tipo de amostragem
amostra_tam = 5000
# Amostragem Aleat√≥ria
amostra_aleatoria = np.random.choice(populacao, amostra_tam)

# Amostragem Sistematica
n = len(populacao_df) // amostra_tam # pega cada nth membro da popula√ß√£o
# FIXME o plot estratificado n t√° com intervalos iguais
# print(" N √© " + str(n))
amostra_sistematica = populacao_df.iloc[::n, :].values.flatten()

# Amostragem Estratificada
populacao_df['estrato'] = pd.cut(populacao_df['altura'], bins=[0, 170, np.inf], labels=['abaixo da m√©dia', 'acima da m√©dia'])
strat_train_set, strat_test_set = train_test_split(populacao_df, test_size=0.25, stratify=populacao_df['estrato'])
amostra_estratificada = strat_test_set['altura'].values

# Amostragem por Conglomerados
# kmeans = KMeans(n_clusters=10, random_state=0).fit(populacao_df[['altura']])
kmeans = KMeans(n_clusters=3, n_init=10, random_state=0).fit(populacao_df[['altura']])

populacao_df['cluster'] = kmeans.labels_
amostra_por_conglomerados = populacao_df[populacao_df['cluster'] == 0]['altura'].values

# Criando uma lista com todos os tipos de amostragem
tipos_de_amostragem = [("Amostragem Aleat√≥ria", amostra_aleatoria), 
                       ("Amostragem Sistem√°tica", amostra_sistematica), 
                       ("Amostragem Estratificada", amostra_estratificada), 
                       ("Amostragem por Conglomerados", amostra_por_conglomerados)]

# Gerando um gr√°fico de histograma para cada tipo de amostragem
for tipo, amostra in tipos_de_amostragem:
    plt.figure(figsize=(12,6))
    plt.hist(populacao, bins=100, alpha=0.5, label='Popula√ß√£o')#, density=True)
    plt.hist(amostra, bins=100, alpha=0.5, label=tipo)#, density=True)

    # Adicionando a legenda
    plt.legend(loc='upper right')

    # Adicionando t√≠tulos e labels
    plt.title(f'Distribui√ß√£o da Altura da Popula√ß√£o - {tipo}')
    plt.xlabel('Altura (cm)')
    plt.ylabel('Frequ√™ncia')

    # Imprimindo stat√≠sticas de resumo da popula√ß√£o e da amostra
    print("\n\n===========   " + tipo.upper() + "   ===========")
    print("\n------------ Medidas-Resumo ------------\n")
    descreve_datasets(pd.Series(populacao), pd.Series(amostra), nome1="Popula√ß√£o", nome2="Amostra", idioma='pt')

    # Mostrando o gr√°fico
    plt.show()
    print()
````

## 4.4 Erros de Amostragem

Ao trabalhar com amostras, erros podem surgir no processo de amostragem e podem afetar a representatividade e precis√£o da amostra em rela√ß√£o √† popula√ß√£o. Esses erros ocorrem quando uma amostra selecionada de uma popula√ß√£o n√£o reflete exatamente as caracter√≠sticas da popula√ß√£o total. Em outras palavras, √© a diferen√ßa entre os resultados obtidos na amostra e os verdadeiros valores que seriam obtidos se toda a popula√ß√£o fosse examinada.

Os erros amostrais s√£o inerentes ao processo de amostragem, pois √© praticamente imposs√≠vel que uma pequena parte represente perfeitamente o todo. Eles podem ser influenciados por v√°rios fatores, como o tamanho da amostra, o m√©todo de sele√ß√£o e a variabilidade inerente √† popula√ß√£o. No campo din√¢mico das Rela√ß√µes P√∫blicas, onde a compreens√£o precisa do p√∫blico √© essencial, compreender e minimizar os erros de amostragem torna-se, pois, fundamental para garantir a confiabilidade e a validade dos resultados de uma pesquisa.

Nas Rela√ß√µes P√∫blicas, alguns erros de amostragem s√£o mais relevantes devido ao impacto direto que eles t√™m na interpreta√ß√£o de dados e na tomada de decis√µes estrat√©gicas. Destacam-se:

1. **Erro de Vi√©s de Sele√ß√£o**. O vi√©s de sele√ß√£o ocorre quando h√° um vi√©s na sele√ß√£o da amostra, ou seja, a maneira como voc√™ seleciona sua amostra faz com que ela seja n√£o representativa da popula√ß√£o. Uma sele√ß√£o enviesada de amostra pode levar a estalos (_insights_) e estrat√©gias baseadas em uma compreens√£o distorcida do p√∫blico. Por exemplo, se apenas usu√°rios ativos de m√≠dias sociais s√£o inclu√≠dos em uma pesquisa sobre percep√ß√µes de marca, isso pode excluir opini√µes de segmentos importantes do p√∫blico que n√£o s√£o t√£o ativos ou que n√£o usam as m√≠dias sociais utilizadas na pesquisa.

1. **Erro de Amostragem Aleat√≥rio**: Mesmo com um processo de sele√ß√£o aleat√≥ria perfeito, ainda haver√° um erro de amostragem, que √© a diferen√ßa entre a verdadeira m√©dia da popula√ß√£o e a m√©dia da amostra. De uma forma geral, quanto maior a amostra, menor ser√° o erro de amostragem. H√° um ponto de diminui√ß√£o, onde amostras maiores n√£o reduzir√£o muito o erro de amostragem.

1. **Erro de N√£o Resposta**: Comum em pesquisas e sondagens, esse erro ocorre quando indiv√≠duos selecionados para a amostra n√£o respondem. Em RP, isso pode resultar em uma compreens√£o incompleta ou distorcida do p√∫blico-alvo, especialmente se os n√£o respondentes compartilharem caracter√≠sticas ou opini√µes comuns.

1. **Erro de Cobertura**: Este erro acontece quando parte da popula√ß√£o foi omitida no momento de escolher as amostras. n√£o √© utilizada toda a popula√ß√£o para fazer a lista ou o quadro da amostra (a partir do qual a amostra √© selecionada) n√£o inclui todos os elementos da popula√ß√£o-alvo. Assim, mesmo com uma sele√ß√£o aleat√≥ria, a amostra pode n√£o representar toda a popula√ß√£o. Esse erro ocorre quando uma amostra que n√£o cobre toda a popula√ß√£o de interesse. Em RP, ao avaliar a efic√°cia de campanhas ou o alcance da marca, uma amostra pode n√£o cobrir toda a popula√ß√£o de interesse pode levar a conclus√µes imprecisas sobre a percep√ß√£o geral da marca.

1. **Erro de Vi√©s de Medi√ß√£o**: O erro de vi√©s de medi√ß√£o, tamb√©m conhecido como erro de medi√ß√£o, √© um tipo de erro que ocorre quando h√° uma distor√ß√£o sistem√°tica nos dados coletados durante o processo de pesquisa. Este erro n√£o est√° relacionado √† escolha da amostra, mas sim √† forma como as informa√ß√µes s√£o coletadas, processadas ou interpretadas. Pode ocorrer devido a perguntas mal formuladas em pesquisas ou interpreta√ß√µes equivocadas dos dados. Em Rela√ß√µes P√∫blicas, o erro de vi√©s de medi√ß√£o pode ter um impacto significativo, pois pode levar a uma compreens√£o distorcida do p√∫blico, efic√°cia das campanhas ou percep√ß√£o da marca. Por exemplo, se uma pesquisa para avaliar a percep√ß√£o da marca usa perguntas sugestivas ou amb√≠guas, os resultados podem n√£o refletir verdadeiramente as atitudes dos consumidores.

1. **Erro de Vi√©s de Informante**: Ocorre quando os dados coletados de participantes ou respondentes n√£o s√£o precisos ou s√£o distorcidos devido a v√°rias raz√µes. Esse tipo de erro afeta a qualidade e a confiabilidade das informa√ß√µes obtidas em uma pesquisa. As principais causas do erro de vi√©s de informante incluem a vontade de agradar o entrevistador ou √† percep√ß√£o social, esquecimento ou mem√≥ria seletiva, entendimento incorreto (interpretar mal as perguntas ou o prop√≥sito da pesquisa), sensibilidade ao assunto como tabus, entre outros.

1. **Erro de Vi√©s de Contexto**: Ocorre quando as circunst√¢ncias ou o ambiente em que a pesquisa √© conduzida influenciam as respostas dos participantes de maneira a distorcer os resultados. Esse tipo de vi√©s pode ser introduzido por diversos fatores relacionados ao contexto da coleta de dados, afetando a maneira como os respondentes interpretam as perguntas e como escolhem suas respostas. Por exemplo, fatores comom ambiente f√≠sico (formal ou informal, temperatura desagrad√°vel), contexto social ou cultural, eventos recentes (falecimento de ente pr√≥ximo), contexto temporal da pesquisa (ap√≥s um time de futebol ser campe√£o), dentre outros.

1. **Erro de Vi√©s de Volunt√°rio**: Em RP, especialmente em pesquisas abertas ou convites para feedback, os indiv√≠duos que escolhem participar podem n√£o ser representativos do p√∫blico maior, levando a compreens√µes potencialmente enviesadas.


1. **Dados At√≠picos e Outliers**: Dados t√≠picos s√£o valores que s√£o significativamente diferentes da maioria dos outros valores na amostra. O erro associado a dados at√≠picos e outliers refere-se ao impacto que observa√ß√µes extremas ou an√¥malas podem ter sobre os resultados de uma pesquisa ou an√°lise de dados. As causas comuns de dados at√≠picos s√£o erros de coleta ou no processamento dos dados ou erros acidentais na coleta de dados -- ambos _outliers_ --, varia√ß√£o natural de uma opula√ß√£o heterog√™nea


Cada um desses erros pode afetar a efic√°cia e a precis√£o das estrat√©gias e campanhas em Rela√ß√µes P√∫blicas, tornando fundamental para os profissionais da √°rea entender e trabalhar para mitigar esses erros em suas pesquisas e an√°lises. Para minimizar os erro de amostragem, os profissionais de RP podem procurar criar um ambiente neutro para a pesquisa, considerar cuidadosamente o momento da pesquisa e utilizar diferentes m√©todos e contextos para coletar dados, aumentando assim a confiabilidade dos resultados obtidos.

````python
import math

def tamanho_amostra_cochran_finita(populacao, margem_erro, confianca, proporcao_estimada=0.5):
    """
    Calcula o tamanho da amostra para uma popula√ß√£o finita usando a f√≥rmula de Cochran.

    :param populacao: Tamanho total da popula√ß√£o.
    :param margem_erro: Margem de erro desejada (em decimal).
    :param confianca: N√≠vel de confian√ßa desejado (em decimal).
    :param proporcao_estimada: Propor√ß√£o estimada da caracter√≠stica na popula√ß√£o (padr√£o √© 0.5).
    :return: Tamanho da amostra calculado.
    """
    z_score = abs(math.erf(confianca / math.sqrt(2)))  # Z-score para o n√≠vel de confian√ßa
    tamanho_amostra_infinita = (z_score ** 2) * proporcao_estimada * (1 - proporcao_estimada) / (margem_erro ** 2)
    tamanho_amostra = tamanho_amostra_infinita / (1 + (tamanho_amostra_infinita - 1) / populacao)
    return round(tamanho_amostra)

def tamanho_amostra_cochran_infinita(margem_erro, confianca, proporcao_estimada=0.5):
    """
    Calcula o tamanho da amostra para uma popula√ß√£o infinita usando a f√≥rmula de Cochran.

    :param margem_erro: Margem de erro desejada (em decimal).
    :param confianca: N√≠vel de confian√ßa desejado (em decimal).
    :param proporcao_estimada: Propor√ß√£o estimada da caracter√≠stica na popula√ß√£o (padr√£o √© 0.5).
    :return: Tamanho da amostra calculado.
    """
    z_score = abs(math.erf(confianca / math.sqrt(2)))  # Z-score para o n√≠vel de confian√ßa
    tamanho_amostra_infinita = (z_score ** 2) * proporcao_estimada * (1 - proporcao_estimada) / (margem_erro ** 2)
    return round(tamanho_amostra_infinita)

def tamanho_amostra_yamane(populacao, margem_erro):
    """
    Calcula o tamanho da amostra usando a f√≥rmula de Yamane para popula√ß√µes finitas.

    :param populacao: Tamanho total da popula√ß√£o.
    :param margem_erro: Margem de erro desejada (em decimal).
    :return: Tamanho da amostra calculado.
    """
    tamanho_amostra = populacao / (1 + populacao * (margem_erro ** 2))
    return round(tamanho_amostra)


# Exemplo de uso das fun√ß√µes
populacao_exemplo = 10000
margem_erro_exemplo = 0.05
confianca_exemplo = 0.95

tamanho_amostra_cochran = tamanho_amostra_cochran_finita(populacao_exemplo, margem_erro_exemplo, confianca_exemplo)
tamanho_amostra_cochran_infinita = tamanho_amostra_cochran_infinita(margem_erro_exemplo, confianca_exemplo)
tamanho_amostra_yamane = tamanho_amostra_yamane(populacao_exemplo, margem_erro_exemplo)

print(f'tamanho_amostra_cochran = {tamanho_amostra_cochran}')
print(f'tamanho_amostra_cochran_infinita = {tamanho_amostra_cochran_infinita}')
print(f'tamanho_amostra_yamane = {tamanho_amostra_yamane}')



````

## 5.1 M√©dia

### 5.1.1 Defini√ß√£o e Exemplos

A m√©dia aritm√©tica √© provavelmente a medida de posi√ß√£o mais conhecida. A m√©dia de um conjunto de dados √© calculada **somando-se todos os n√∫meros e depois dividindo-se pelo n√∫mero total de n√∫meros**. A f√≥rmula da m√©dia √©:

$ \bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i $

Onde:
- $\bar{x}$ representa a m√©dia da amostra,
- $n$ √© o n√∫mero total de observa√ß√µes na amostra,
- $x_i$ representa cada valor individual na amostra,
- $\sum$ √© o s√≠mbolo de somat√≥rio, indicando a soma de todos os valores $x_i$.

Por exemplo, para ter uma ideia geral do per√≠odo do entrevistado e da sua pretens√£o salarial, calculamos as **m√©dias** do per√≠odo e da pretens√£o salarial dos entrevistados do conjunto de dados CD-2; conforme calculado pelo programa a seguir. Esses valores m√©dios mostrados abaixo nos d√£o uma vis√£o geral em qual per√≠odo os entreistados est√£o e qual √© a sua pretens√£o salarial. No caso do per√≠odo, como se trata de uma vari√°vel discreta, pode-se arredondar par cima ou para baixo o valor da m√©dia para representar de forma discreta o per√≠odo dos entresvistados.

## 6.1 Amplitude Total (Faixa Total)

A amplitude √© a diferen√ßa entre o valor m√°ximo e o valor m√≠nimo em um conjunto de dados: 

$Amplitude\ Total = Valor\ m√°ximo - Valor\ m√≠nimo$


Por exemplo, no Conjunto de Dados CD-2, suponhamos que selecionamos uma amostra da pretens√£o salarial de 10 estudantes e calculemos a Amplitude Total desses dados. Apresentamos essa amostra em tabela e em um gr√°fico a seguir, que ilustra o c√°lculo da Amplitutde Total.

````python
import numpy as np

# Criando uma lista de pretens√µes salariais
salaries = np.round(np.random.normal(5000, 1000, 9)).astype(int)
salaries = np.append(salaries, 10000) # EXERCICIO: comentar essa linha e analisar a diferen√ßa
sorted_salaries = np.sort(salaries)
range_salary = round(np.ptp(sorted_salaries), 2)  # Calculando a amplitude

# Lista de nomes fict√≠cios para o eixo X
names = ["Ana", "Bruno", "Carlos", "Diana", "Eduardo", "Fernanda", "Gabriel", "Helena", "Igor", "J√∫lia"]

# Garantindo que a lista de nomes esteja na mesma ordem que os sal√°rios ordenados
# Isso n√£o √© estritamente necess√°rio se voc√™ estiver apenas etiquetando o eixo x, mas √© uma boa pr√°tica.
sorted_names = [x for _, x in sorted(zip(salaries, names))]

# Emparelhando nomes e sal√°rios usando zip e convertendo para uma lista
data = list(zip(sorted_names, sorted_salaries))

# Imprimindo a tabela com cabe√ßalhos 'Estudante' e 'Pretens√£o Salarial'
from IPython.display import display, Markdown

# Iniciando a string de tabela Markdown com os cabe√ßalhos
markdown_table = "| Estudante | Pretens√£o Salarial (R$) |\n"
markdown_table += "|-----------|----------------------|\n"

# Adicionando cada linha de dados √† string de tabela
for name, salary in zip(sorted_names, sorted_salaries):
    markdown_table += f"| {name} | {salary} |\n"

# Exibindo a tabela no Jupyter Notebook
display(Markdown(markdown_table))
````

````python
import matplotlib.pyplot as plt
import seaborn as sns


# Plotando um gr√°fico de barras para ilustrar a amplitude
plt.figure(figsize=(10, 4))
sns.barplot(x=sorted_names, y=sorted_salaries)
plt.ylim(0, 12000)  # limites do eixo y para visualiza√ß√£o clara
plt.title(f"Amplitude das Pretens√µes Salariais = R$ {range_salary}")
plt.xlabel('Estudante')
plt.ylabel('Pretens√£o Salarial (R$)')

# Adicionando linhas para mostrar a amplitude
min_salary = min(sorted_salaries)
max_salary = max(sorted_salaries)
plt.axhline(y=min_salary, color='r', linestyle='--', label=f'Pretens√£o m√≠nima = R$ {min_salary}')
plt.axhline(y=max_salary, color='b', linestyle='--', label=f'Pretens√£o m√°xima = R$ {max_salary}')

# Adicionando legenda
plt.legend()

# Adicionando anota√ß√£o para a amplitude
mid_index = len(sorted_salaries) // 2
plt.annotate('', xy=(mid_index, min_salary), xytext=(mid_index, max_salary),
             arrowprops=dict(arrowstyle='<->', color='black'))
plt.text(mid_index + 0.5, (min_salary + max_salary) / 2, f'AMPLITUDE = R$ {range_salary}',
         rotation=90, verticalalignment='center', color='black')

plt.show()
````

A amplitude deste conjunto de dados anteior ser√° muito alta devido ao valor at√≠pico da pretens√£o salarial da J√∫lia. Isso pode fazer com que a amplitude pare√ßa maior do que a variabilidade real para a maioria dos estudantes. Em suma, a Amplitude Total √© muito sens√≠vel aos valores at√≠picos.

Em outras palavras, a Amplitutde Total √© uma medida de dispers√£o que nos d√° uma **compreens√£o limitada da varia√ß√£o de um conjunto de dados**, pois foca nos valores m√°xioms e m√≠nimos dos dados. Ademais, ela √© sens√≠vel a valores at√≠picos (outliers), o que pode mascarar a verdadeira variabilidade dos dados. 

A Amplitutde Total tamb√©m ignora a distribui√ß√£o dos dados, por isso ela √© √∫til para termos uma no√ß√£o de quanto est√° variando os dados em rela√ß√£o aos valores m√°ximo e m√≠nimo. Por isso, quando aplic√°vel, deve-se remover os valores at√≠picos e analisar a variabilidade dos dados juntamente com outras medidas de dispers√£o que ser√£o estudadas a seguir.


## 6.2 Vari√¢ncia e Desvio Padr√£o

A vari√¢ncia e o desvio padr√£o s√£o medidas estat√≠sticas fundamentais na compreens√£o da dispers√£o de um conjunto de dados, permitindo uma compreens√£o mais completa sobre a variabilidade de um conjunto de dados. A vari√¢ncia e o desvio padr√£o s√£o √∫teis para **identificar o quanto os valores em um conjunto de dados divergem da m√©dia**. 

√â poss√≠vel notar essa compara√ß√£o com a m√©dia no c√°lculo da vari√¢ncia, pois a vari√¢ncia de um conjunto de dados √© definida como a m√©dia das diferen√ßas quadradas entre cada valor e a m√©dia do conjunto de dados. Para um conjunto de dados $X$ com $N$ observa√ß√µes (valores) $x_i$, onde $\bar{x}$ √© a m√©dia dos valores, a vari√¢ncia, denotada por $\sigma^2$, √© dada por:

$\sigma^2 = \frac{1}{N}\sum_{i=1}^{N}(x_i - \bar{x})^2$

Por conven√ß√£o, chamamos o **desvio padr√£o de $\sigma$** e como ele √© definido como a **raiz quadrada da vari√¢ncia**representamos a vari√¢ncia por $\sigma^2$.

$\sigma = Desvio\ Padr√£o = \sqrt{\sigma^2}$

$\sigma^2 = Vari√¢ncia$

Extrimos a raiz quadrada da vari√¢ncia e a chamamos de **desvio padr√£o** para termos uma  medida de dispers√£o dos dados em rela√ß√£o √† m√©dia em **na mesma medida dos dados originais**; uma vez que utilizamos o quadrado no c√°lculo da vari√¢ncia. Dessa maneira, o desvio padr√£o pode ser comparado com os valores do conjunto de dados diretamente. 

O desvio padr√£o √© uma das medidas de dispers√£o mais utilizadas, pois indica quanto, em m√©dia, os valores no conjunto de dados desviam da m√©dia $\bar{x}$. Um desvio padr√£o baixo indica que os dados est√£o agrupados mais perto da m√©dia, enquanto um desvio padr√£o alto indica que os dados est√£o mais espalhados.

Por exemplo, utilizando o conjunto de dados anterior, criamos o gr√°fico a seguir, chamado de gr√°fico de dispers√£o. Ele oferece uma representa√ß√£o visual dos conceitos de m√©dia e desvio padr√£o, al√©m de mostrar a distribui√ß√£o dos dados das pretens√µes salarias dos estudantes. 

- **Pontos no Gr√°fico:** Cada ponto no gr√°fico representa a pretens√£o salarial de um estudante. O eixo Y mostra a pretens√£o salarial em R\$ e o eixo X representa cada estudante.
- **Linha vermelha (M√©dia):** A linha vermelha horizontal mostra o valor da m√©dia das pretens√µes salarias dos estudantes. 
- **√Årea sombreada em vermelho (Desvio Padr√£o):** O desvio padr√£o √© uma medida de dispers√£o que mostra o quanto as pretens√µes salarias dos estudantes variam em rela√ß√£o √† m√©dia. A √°rea sombreada em vermelho representa essa varia√ß√£o, cobrindo as pretens√µes salarias que est√£o a um desvio padr√£o acima e abaixo da m√©dia. Perceba que **a pretens√£o salarial da J√∫lia est√° acima do desvio padr√£o**. Em uma distribui√ß√£o normal, cerca de 68% dos dados se encontram dentro de um desvio padr√£o da m√©dia.

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Utilizando o conjunto de dados definido anteriormente

# C√°lculo da m√©dia, vari√¢ncia e desvio padr√£o das pretens√µes salariais
mean_salary = np.mean(salaries)
variance_salary = np.var(salaries)
std_dev_salary = np.std(salaries)

# Criando um gr√°fico de dispers√£o para as pretens√µes salariais
plt.figure(figsize=(8, 4))
plt.ylim(0, 12000)
plt.scatter(sorted_names, sorted_salaries, label='Pretens√µes Salariais')
plt.hlines(mean_salary, 0, len(salaries)-1, colors='r', label='M√©dia')

# √Årea representando o desvio padr√£o
plt.fill_between(sorted_names, mean_salary-std_dev_salary, mean_salary+std_dev_salary, color='r', alpha=0.1, label='Desvio padr√£o')

# √Årea representando a vari√¢ncia
# Para a vari√¢ncia, calculamos o intervalo usando a raiz quadrada da vari√¢ncia para manter a mesma unidade dos dados
lower_bound = mean_salary - np.sqrt(variance_salary)
upper_bound = mean_salary + np.sqrt(variance_salary)
# plt.fill_between(sorted_names, lower_bound, upper_bound, color='b', alpha=0.1, label='Vari√¢ncia')

# Ajustando os t√≠tulos e r√≥tulos para os dados de pretens√£o salarial
plt.title('Gr√°fico de Dispers√£o das Pretens√µes Salariais dos Estudantes')
plt.xlabel('Estudante')
plt.ylabel('Pretens√£o Salarial (R$)')
# plt.xticks(rotation=45)  # Rota√ß√£o dos nomes para melhor visualiza√ß√£o
plt.legend(loc='upper left')

plt.tight_layout()  # Ajusta o layout para evitar sobreposi√ß√£o
plt.show()
````

## 6.3 Medidas Resumo de Quantis

Os quantis s√£o valores derivados do conjunto de dados quando o dividimos em partes iguais. As medidas de quantis mais comuns s√£o os quartis, os decils e os percentis.

Os percentis s√£o muito √∫teis para entender onde um ponto de dados espec√≠fico se encaixa no contexto de todo o conjunto de dados. Por exemplo, se a pretens√£o salarial de um estudante estiver no percentil 90, isso significa que esse estudante possui pretens√£o salarial maior que 90% dos demais.

### Quartis

Os quartis s√£o um tipo espec√≠fico de quantil que **dividem um conjunto de dados ordenados em quatro partes** iguais. Os quartis nos fornece s√£o √∫teis para fornecer uma vis√£o r√°pida sobre a distribui√ß√£o dos dados, identificar valores at√≠picos e entender a dispers√£o dos dados em torno da **mediana**.

Existem tr√™s quartis:

1. **Primeiro quartil (Q1)**: tamb√©m conhecido como quartil inferior, representa o valor abaixo do qual se encontram **25%** dos dados quando ordenados do menor para o maior. Em outras palavras, Q1 √© o valor que divide os primeiros 25% dos dados do restante.

2. **Segundo quartil (Q2)**: √© a **mediana**, que divide o conjunto de dados ao meio, ou seja, **50%** dos dados s√£o menores que Q2 e 50% dos dados s√£o maiores.

3. **Terceiro quartil (Q3)**: tamb√©m conhecido como quartil superior, √© o valor abaixo do qual **75%** dos dados se encontram. Q3, portanto, separa os 25% superiores dos dados do restante.



Por, exemplo, vamos calcular os quartis das pretens√µes salarias dos estudantes a seguir:

````python
import numpy as np

# Supondo que 'sorted_salaries' j√° tenha sido definido
# sorted_salaries = np.sort(np.append(np.round(np.random.normal(5000, 1000, 9)).astype(int), 10000))

# Calculando os quartis
quartil_1 = np.percentile(sorted_salaries, 25)  # Primeiro quartil (Q1)
quartil_2 = np.percentile(sorted_salaries, 50)  # Segundo quartil (Q2) ou Mediana
quartil_3 = np.percentile(sorted_salaries, 75)  # Terceiro quartil (Q3)

# Imprimindo os resultados
print("Quartis para as pretens√µes salariais:")
print(f"Primeiro Quartil (Q1): R$ {quartil_1:.0f}")
print(f"Segundo Quartil (Q2 ou Mediana): R$ {quartil_2:.0f}")
print(f"Terceiro Quartil (Q3): R$ {quartil_3:.0f}")
````

## 6.4 Amplitude Interquartil (Faixa Interquartil)

A Amplitude Interquartil (tamb√©m conhecida como Faixa Interquartil ou _Interquartile Range - IQR_) √© uma medida de dispers√£o estat√≠stica que representa a diferen√ßa entre o terceiro quartil (Q3) e o primeiro quartil (Q1):

$IQR = Q_3 - Q_1$

Ou seja, a Amplitude Interquartil abrange o intervalo entre os 25% e 75% dos dados quando ordenados do menor para o maior. Isso resulta em um valor que indica a "largura" da parte central dos dados, proporcionando **uma medida da dispers√£o dos dados ao redor da mediana**:

- Se a Amplitude Interquartil √© grande, isso indica que os dados est√£o mais dispersos.
- Se for pequena, isso indica que os dados est√£o mais agrupados, menos dispersos.


Como a Amplitude Interquartil mede a **dispers√£o em torno da mediana**, isso a torna **uma medida de dispers√£o robusta**, isto √©, ela √© resistente a valores at√≠picos. A Amplitude Interquartil √© √∫til para entender a variabilidade dos dados quando a distribui√ß√£o n√£o √© sim√©trica ou quando h√° valores at√≠picos significativos. 

````python
import numpy as np

# Calculando a Faixa Interquartil (IQR)
iqr = quartil_3 - quartil_1

# Imprimindo os resultados
print(f"Primeiro Quartil (Q1): R$ {quartil_1:.0f}")
print(f"Terceiro Quartil (Q3): R$ {quartil_3:.0f}")
print(f"Faixa Interquartil (IQR = Q3 - Q1): R$ {iqr:.0f}")
````

## 6.5 Boxplot (Diagrama de Caixas)

### Componentes de um Boxplot

Um gr√°fico de caixa, tamb√©m conhecido como **boxplot, permite visualizar a distribui√ß√£o, dispers√£o e simetria dos dados atrav√©s de seus quartis**. Cada elemento do boxplot est√° explicado a seguir:

- A **caixa central** do gr√°fico encapsula a metade intermedi√°ria dos dados, estendendo-se do primeiro quartil (Q1) ao terceiro quartil (Q3), o que fornece uma vis√£o da faixa interquartil ($IQR = Q_3 - Q_1$). 
- A linha que atravessa a caixa indica a **mediana** ($Q_2$), oferecendo uma compreens√£o clara do valor central do conjunto de dados. 
- Os **"bigodes"** se estendem a partir da caixa at√© os valores m√°ximos e m√≠nimos.
   - Comumente, exclu√≠mos os valores at√≠picos (_outliers_ ou n√£o) ao construirmos um boxplot. Ent√£o, os bigodes ficar√£o dentro de um limite especificado, 1,5 vezes a IQR acima do terceiro quartil ($Q_3$) e abaixo do primeiro quartil ($Q_1$). Assim, permite-se a identifica√ß√£o de valores extremos ou outliers, que s√£o representados por pontos individuais fora dos bigodes.

Dessa forma, o boxplot **resume visualmente v√°rias caracter√≠sticas** estat√≠sticas de um conjunto de dados.

## 7.2 Defini√ß√£o

A Distribui√ß√£o de Frequ√™ncias √© uma **representa√ß√£o tabular ou gr√°fica de dados que descreve a ocorr√™ncia de vari√°veis**. Cada entrada na tabela ou gr√°fico cont√©m a quantidade ou a percentagem de ocorr√™ncias de um determinado intervalo de valores dentro do conjunto de dados total. Este conceito √© uma parte fundamental da estat√≠stica e ajuda a fornecer um entendimento b√°sico sobre a natureza e a distribui√ß√£o dos dados coletados.

Com rela√ß√£o ao tipo de vari√°vel, a Distribui√ß√£o de Frequ√™ncias apresentam as seguintes caracter√≠sticas:

- **Vari√°veis Qualitativas**: Para dados categ√≥ricos ou qualitativos, a distribui√ß√£o de frequ√™ncia contar√° o n√∫mero de vezes que cada categoria aparece no conjunto de dados.
- **Vari√°veis Quantitativas**: Para dados quantitativos, a distribui√ß√£o de frequ√™ncia contar√° o n√∫mero de dados dentro de certos intervalos.

Por exemplo, suponha que um professor tenha registrado as notas das provas de uma turma. Nesse caso, temos uma vari√°vel quantitativa discreta, a nota, e ela varia de 0 a 100. O professor est√° interessado em entender a distribui√ß√£o dessas notas para analisar o desempenho geral da turma. Dessa maneira, a distribui√ß√£o de frequ√™ncias pode ser tabulada assim:

| Intervalo de Notas | Frequ√™ncia |
|--------------------|------------|
| 40-59              | 5          |
| 60-79              | 6          |
| 80-100             | 7          |

A interpreta√ß√£o dessa distribui√ß√£o de frequ√™ncias √© a seguinte:

- 5 estudantes tiraram notas entre 40 e 59.
- 6 estudantes tiraram notas entre 60 e 79.
- 7 estudantes tiraram notas entre 80 e 100.


Portanto, uma distribui√ß√£o de frequ√™ncia, seja tabular ou gr√°fica, serve como uma ferramenta muito √∫til para resumir e interpretar grandes conjuntos de dados, tornando a informa√ß√£o compreens√≠vel e facilmente acess√≠vel. Geralmente, a representa√ß√£o da distribui√ß√£o de frequ√™ncias de grandes conjuntos de dados √© feita atrav√©s de histogramas ou gr√°ficos de barras para melhor visualiza√ß√£o. 


## 7.3 Histogramas

### Defini√ß√£o e Utilidade

Um histograma √© uma representa√ß√£o gr√°fica de uma distribui√ß√£o de frequ√™ncia, amplamente utilizada para ilustrar a forma visual da distribui√ß√£o dos dados. Ao utilizar ret√¢ngulos de diferentes alturas, alinhados lado a lado, √© poss√≠vel visualizar com clareza as frequ√™ncias (ou contagens) de dados dentro de intervalos espec√≠ficos, tamb√©m conhecidos como classes. A **base de cada ret√¢ngulo representa uma classe**, enquanto **a altura do ret√¢ngulo indica a frequ√™ncia dos dados** nesse intervalo espec√≠fico.

O prop√≥sito de um histograma vai al√©m da simples contagem de observa√ß√µes. Ele serve como uma ferramenta para visualizar e entender a distribui√ß√£o subjacente dos dados. Ao observar a **forma**, a **amplitude** e as **varia√ß√µes na altura** dos ret√¢ngulos do histograma √© poss√≠vel identificar padr√µes, tais como simetria, assimetria, picos m√∫ltiplos (modas) e poss√≠veis valores at√≠picos.

Por exemplo, em um histograma que representa as notas de alunos em uma prova, √© poss√≠vel identificar rapidamente a concentra√ß√£o de notas, os valores mais frequentes e a varia√ß√£o na distribui√ß√£o das notas. Isso pode oferecer informa√ß√µes sobre o desempenho geral dos alunos, a dificuldade da prova, entre outros aspectos. Por exemplo vejamos o histograma a seguir do exemplo da Se√ß√£o 7.1 das notas dos alunos.

````python
import matplotlib.pyplot as plt
import numpy as np

# Dados fornecidos
intervalos = ['30-49', '50-89', '90-100']
frequencia = [5, 6, 7]

# Cores para os intervalos
cores = ['antiquewhite', 'slateblue', 'aquamarine']

# Criando o gr√°fico de barras
plt.figure(figsize=(3,3))  # Ajustando o tamanho do gr√°fico conforme necess√°rio
plt.bar(intervalos, frequencia, color=cores)
plt.title('Gr√°fico de Barras da Distribui√ß√£o de Frequ√™ncias das Notas dos Alunos')
plt.xlabel('Intervalo de Notas')
plt.ylabel('Frequ√™ncia')

# Exibindo o gr√°fico
plt.tight_layout()  # Ajusta automaticamente os subplots para que caibam no espa√ßo da figura
plt.show()
````

### Construindo um Histograma

Para construir um histograma, temos dois passos:

1. Os dados s√£o primeiramente **organizados em classes de igual largura** ou de larguras vari√°veis. Estas classes s√£o faixas de valores, e os dados s√£o contados de acordo com a classe √† qual pertencem, pois a escolha limites de classe apropriados podem ajudar a revelar a verdadeira natureza da distribui√ß√£o dos dados, enquanto a escolha inadequada pode ocultar padr√µes importantes ou levar a interpreta√ß√µes equivocadas.
2. Em seguida, constru√≠mos **um gr√°fico onde cada classe √© representada por uma barra** e a **altura de cada barra indica a frequ√™ncia dos dados** que caem dentro daquela classe. As barras tamb√©m podem representar a densidade de frequ√™ncia, no caso de larguras variadas

Portanto, o eixo horizontal do histograma (eixo X) representa os intervalos de classes, enquanto o eixo vertical (eixo y) representa as frequ√™ncias. 

Um aspecto importante sobre a constru√ß√£o de histograms s√£o o **tamanho das classes**. Classes muito largas podem suavizar demais a distribui√ß√£o, fazendo com que se percam detalhes sobre varia√ß√µes menores nos dados. Por outro lado, classes muito estreitas podem resultar em um histograma com muitas barras com pouca ou nenhuma observa√ß√£o, o que pode sugerir uma variabilidade excessiva que na realidade n√£o existe nos dados.

Por exemplo, vamos agora visualizar a mesma distribui√ß√£o de frequ√™ncias do exemplo anteior com dois gr√°ficos: um gr√°ficos de barras e um histograma. Perceba que, **no histograma, a faixa de valor $50-89$ √© a mais larga** de todas, pois o histograma respeita a escala de valores dos intervalos de dados (faixas), diferente do gr√°fico de barras. Em outras palavras, o histograma apresenta mais informa√ß√µes sobre o conjunto de dados do que o gr√°fico de barras.

````python
import matplotlib.pyplot as plt
import numpy as np

# Dados fornecidos
intervalos = ['30-49', '50-89', '90-100']
frequencia = [5, 6, 7]

# Cores para os intervalos
cores = ['antiquewhite', 'slateblue', 'aquamarine']

# Convertendo os intervalos em uma lista de limites para os bins
limites = [int(intervalo.split('-')[0]) for intervalo in intervalos] + [int(intervalos[-1].split('-')[1])]

# Criando subplots
fig, axs = plt.subplots(2, 1, figsize=(4,6))

# Gr√°fico de barras
axs[0].bar(intervalos, frequencia, color=cores)
axs[0].set_title('Gr√°fico de Barras da Distribui√ß√£o de Frequ√™ncias das Notas dos Alunos')
axs[0].set_xlabel('Intervalo de Notas')
axs[0].set_ylabel('Frequ√™ncia')

# Histograma - Criando um conjunto de dados baseado nos intervalos e frequ√™ncias
for i, (inicio, fim) in enumerate(zip(limites[:-1], limites[1:])):
    dados = np.random.uniform(inicio, fim, frequencia[i])
    axs[1].hist(dados, bins=[inicio, fim], color=cores[i], edgecolor='black', alpha=0.7, rwidth=0.85)

axs[1].set_title('Histograma da Distribui√ß√£o de Frequ√™ncias das Notas dos Alunos')
axs[1].set_xlabel('Notas')
axs[1].set_ylabel('Frequ√™ncia')

# Exibindo os gr√°ficos
plt.tight_layout()
plt.show()
````

## 7.4 Tipos de Distribui√ß√£o de Frequ√™ncias

Nesta se√ß√£o, vamos aprofundar nos tipos de distribui√ß√µes de frequ√™ncia, discutindo sobre o comportamento das medidas resumo (como m√©dia, mediana e moda) em cada tipo de distribui√ß√£o. Al√©m disso, vamos relacionar essas distribui√ß√µes com representa√ß√µes gr√°ficas como o boxplot e as curvas de densidade de probabilidade. Curvas de densidade de probabilidade, ou simplesmente curvas de densidade, s√£o uma representa√ß√£o suavizada do padr√£o subjacente dos dados em um histograma. Em outras palavras, essas curvas ajudam a visualizar a distribui√ß√£o dos dados de uma maneira mais cont√≠nua e menos fragmentada do que as distribui√ß√µes de frequ√™ncia.


### Distribui√ß√£o Uniforme

Em uma distribui√ß√£o uniforme, todas as vari√°veis ou classes t√™m a mesma frequ√™ncia ou probabilidade de ocorr√™ncia. Ou seja, em um conjunto de dados com distribui√ß√£o uniforme, cada valor ocorre com a mesma frequ√™ncia ou frequ√™ncia similiar. Isso significa que a m√©dia, mediana e moda podem cair em qualquer ponto dentro de um memso intervalo, pois todos as observa√ß√µes t√™m frequ√™ncias parecidas. 

O boxplot de uma distribui√ß√£o uniforme mostrar√° uma caixa de tamanho uniforme se os dados estiverem perfeitamente distribu√≠dos.

Na distribui√ß√£o normal, a curva de densidade tem a forma cl√°ssica de sino, indicando que a maioria das observa√ß√µes est√° concentrada em torno da m√©dia, e se torna menos frequente √† medida que se afasta da m√©dia em ambas as dire√ß√µes.

### Distribui√ß√£o Normal

Uma distribui√ß√£o normal √© caracterizada pela sua forma de sino sim√©trica. Neste caso, a m√©dia, mediana e a moda s√£o iguais e se localizam no centro do pico. 

O boxplot associado a uma distribui√ß√£o normal ser√° sim√©trico, indicando que os dados s√£o distribu√≠dos de maneira igual em torno da m√©dia, por√©m ele apresenta uma concentra√ß√£o maior entre Q1 e Q3 quando comparado com o boxplot da Distribui√ß√£o Uniforme por exemplo.

Em uma distribui√ß√£o uniforme, a curva de densidade seria uma linha reta, indicando que cada valor tem a mesma probabilidade de ocorr√™ncia.

### Distribui√ß√£o Binomial

Em uma Distribui√ß√£o Binomial, h√° apenas dois poss√≠veis resultados ou classes, como "sucesso" e "fracasso". A m√©dia e mediana podem variar, mas geralmente ficam no centro se ambos os resultados forem igualmente prov√°veis. 

O boxplot pode variar dependendo das probabilidades de sucesso e fracasso, e a simetria ser√° evidente se as probabilidades forem iguais.

Na distribui√ß√£o binomial, a curva de densidade pode assumir diferentes formas dependendo da probabilidade de sucesso e do n√∫mero de tentativas. N√£o √© uma curva suave, mas mostra a probabilidade de cada n√∫mero espec√≠fico de sucessos.

### Distribui√ß√£o Assim√©trica (Skewness)

A Distribui√ß√£o Assim√©trica √© inclinada, podendo ser positiva ou negativa. Em uma Distribui√ß√£o Assim√©trica positiva, a cauda √† direita √© mais longa, implicando que a m√©dia √© maior que a mediana, que √©, por sua vez, maior que a moda. Em uma Distribui√ß√£o Assim√©trica negativa, a cauda √† esquerda √© mais longa, indicando que a m√©dia √© menor que a mediana, que √© menor que a moda. 

O boxplot mostra claramente a assimaetria da Distribui√ß√£o Assim√©trica.

Em distribui√ß√µes com skewness, a curva de densidade inclina-se para o lado onde os dados s√£o menos frequentes, refletindo a assimetria dos dados.

### Distribui√ß√£o Bimodal

Uma distribui√ß√£o bimodal tem dois picos distintos, indicando que h√° duas modas. A m√©dia e a mediana situam-se entre esses dois picos. Na Distribui√ß√£o Bimodal, a moda n√£o serve como uma √∫nica medida representativa da centralidade dos dados, pois existem dois picos distintos ou modas. Nesses casos, √© importante considerar ambas as modas para entender completamente a distribui√ß√£o dos dados. 

O boxplot pode mostrar duas caixas distintas ou uma dist√¢ncia interquartil mais ampla, dependendo da separa√ß√£o das modas e da dispers√£o dos dados.

Em uma distribui√ß√£o bimodal, a curva de densidade ter√° dois picos para representar os dois conjuntos distintos de dados dentro da amostra.


> **Exemplo**. Simulamos v√°rios conjuntos de dados gerados aleatoriamente, cada um tem uma distribui√ß√£o espec√≠fica. Cada conjunto de dados √© visualmente apresentada com seu histograma correspondente e boxplot . A partir desses gr√°ficos, voc√™ pode observar claramente as caracter√≠sticas √∫nicas de cada tipo de distribui√ß√£o e como a m√©dia, a mediana e a moda, variam de acordo. Perceba que ambos gr√°ficos para cada conjunto de dados est√£o alinhados em rela√ß√£o √† sua escala no eixo X.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configura√ß√£o para o estilo dos gr√°ficos
sns.set(style="whitegrid")

# Dados de exemplo
data = {
    'Uniforme': np.random.rand(1000),
    'Normal': np.random.randn(1000),
    'Binomial': np.random.binomial(n=10, p=0.5, size=1000),
    'Positivamente Inclinada': np.random.exponential(scale=1, size=1000),
    'Bimodal': np.concatenate((np.random.normal(loc=5, scale=2, size=500),
                               np.random.normal(loc=10, scale=2, size=500)))
}

# Criando gr√°ficos
fig, axes = plt.subplots(2, len(data.keys()), figsize=(15,10))

# Iterando atrav√©s dos dados para criar histogramas e boxplots
for i, dist in enumerate(data.keys()):
    sns.histplot(data[dist], bins=20, kde=True, ax=axes[0, i]).set(title=f'Distribui√ß√£o {dist}', xlabel='Observa√ß√µes', ylabel='Frequ√™ncia')
    sns.boxplot(x=data[dist], ax=axes[1, i]).set(xlabel='Observa√ß√µes')


# Ajustando os layout
plt.tight_layout()
plt.show()

````
