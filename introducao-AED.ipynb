# 1. Definição de Análise Exploratória de Dados

A Análise Exploratória de Dados (AED) é uma etapa elementar na análise de dados que envolve uma série de técnicas para visualizar, resumir e entender a informação que nossos dados estão nos fornecendo. Esta fase da análise de dados é fundamental para gerar ideias (_insights_), testar hipóteses e verificar suposições.

A AED é uma abordagem que enfatiza a exploração visual dos dados, em vez de testes estatísticos formais ou modelagem preditiva. Seu objetivo principal é revelar a estrutura dos dados, detectar outliers ou anomalias, testar suposições, desenvolver modelos intuitivos e determinar a seleção adequada de ferramentas e técnicas para análises futuras como a Análise Inferencial, que realiza previsões ou inferências sobre uma população com base nos dados da amostra.

Para realizarmos uma Análise Exploratória de Dados, o primeiro passo é **compreender o problema** e **modela-lo**. A modelagem de dados envolve a criação de um modelo que pode descrever padrões complexos nos dados. Esses modelos podem então ser usados para fazer previsões sobre dados futuros ou para compreender melhor os fenômenos subjacentes aos dados. 

Nesse contexto, modelar significa também decidir quais dados serão coletados, ou mensarados, de forma que melhor representem o problema estudado. Ou seja, é durante a modelagem que definimos **o quê sera observado** (medido, coletado, mensurado, etc.). Isso implica definir quais varáveis representarão o problema estudado. 

Por exemplo, se estamos estudando os fatores que afetam o desempenho dos alunos em um curso universitário, nossa modelagem pode envolver a decisão de coletar dados sobre o tempo de estudo semanal dos alunos, sua frequência às aulas, se têm ou não um emprego paralelo, e assim por diante. Cada um desses é uma variável que pode ser medida e coletada. 

Nesse caso, a "nota final do curso" poderia ser a variável de interesse que estamos tentando prever ou explicar (também conhecida como a variável dependente). As variáveis que estamos coletando — tempo de estudo, frequência às aulas, emprego paralelo — são as variáveis que acreditamos que podem influenciar a variável de interesse (conhecidas como variáveis independentes). 

Portanto, a modelagem de dados neste contexto envolveria:

1. **Definir a questão de pesquisa ou o problema**: Neste caso, queremos entender os fatores que afetam o desempenho dos alunos em um curso universitário.

2. **Decidir quais dados coletar**: Baseado na questão de pesquisa, decidimos coletar dados sobre tempo de estudo, frequência às aulas e se o aluno tem ou não um emprego paralelo.

3. **Definir as variáveis**: A "nota final do curso" é nossa variável dependente, enquanto "tempo de estudo", "frequência às aulas" e "emprego paralelo" são nossas variáveis independentes.

4. **Coletar os dados**: Executamos a coleta de dados de acordo com o que foi definido.

5. **Analisar os dados**: Usamos técnicas estatísticas para analisar os dados e construir um modelo que possa prever a variável de interesse a partir das variáveis independentes.

Assim, a modelagem de dados é uma parte integral da AED, ajudando a determinar quais dados devem ser coletados e como eles serão analisados para responder à questão de pesquisa.

# 3. Variáveis

No contexto da Análise Exploratória de Dados, uma variável é uma característica, número ou quantidade que pode ser medida ou contada. As variáveis podem assumir vários valores e cada valor da variável pode corresponder a uma observação ou medida diferente.

As variáveis podem ser categorizadas principalmente em dois tipos: qualitativas e quantitativas. As variáveis também podem ser dependentes uma das outras e apresentarem correlações. A seguir, vamos estudar esses conceitos aplicando-os na área das Relações Públicas.

## 3.1 Tipos de Variáveis

- **Qualitativas**. Também conhecidas como variáveis categóricas, elas são aquelas que expressam uma característica ou qualidade que não pode ser medida numericamente. Elas podem ser divididas em:
    - **Variáveis Nominais**: não existe uma ordem ou classificação entre as categorias. Exemplos: gênero, tipo de mídia (TV, rádio, redes sociais), setor da empresa (tecnologia, saúde, educação), nacionalidade dos influenciadores, etc.
    - **Variáveis Ordinais**: antagonicamente às variáveis nominais, as categorias têm uma ordem ou classificação específico. Por exemplos: classificações de filmes, níveis de educação, níveis de satisfação do cliente (satisfeito, neutro, insatisfeito), prioridade de assuntos na agenda de comunicação (alta, média, baixa), etc.
- **Quantitativas**. Estas são variáveis mensuráveis, que podem ser medidas numericamente. Elas podem ser divididas em:
    - **Variáveis Discretas**: são contáveis e normalmente são números inteiros. Por exemplo, número de comunicados de imprensa emitidos, quantidade de menções da marca em mídias sociais, número de eventos organizados, quantidade de funcionários, etc.
   - **Variáveis Contínuas**: podem assumir qualquer valor dentro de um intervalo específico. Exemplos: tempo gasto em um site da empresa, orçamento de marketing, altura e peso das pessoas, horas da campnha de marketing digital, etc.
   
   
Por exemplo, vamos ver como diferentes tipos de dados são utilizados para analisar estratégicas de Relações Públicas. Ao analisar o tipo de mídia e o nível de satisfação do cliente, pode-se avaliar quais canais são mais eficazes. Da mesma forma, a relação entre o número de comunicados de imprensa e o alcance da campanha pode ajudar a entender a eficácia da comunicação da empresa. Consideremos, pois, as seguintes variáveis:

1. **Tipo de Mídia (Variável Nominal)**: Categorias sem ordem específica, como Redes Sociais, TV, Rádio, Impresso e Online. Isso indica os diferentes canais de comunicação utilizados.

2. **Nível de Satisfação do Cliente (Variável Ordinal)**: Uma escala com ordem que varia de Alto, Médio a Baixo. Esta variável mede a percepção e reação do cliente às iniciativas de RP.

3. **Número de Comunicados de Imprensa (Variável Discreta)**: Uma contagem de quantos comunicados de imprensa foram emitidos. Sendo um número inteiro, indica a frequência da comunicação formal da empresa.

4. **Alcance da Campanha (Variável Contínua)**: Mede o alcance das campanhas em termos de público, expresso em milhares de pessoas. Isso reflete a extensão do impacto de uma campanha.

### Exemplos de Variáveis nas Relações Públicas

A fim de entender como a Análise Exploratória de Dados pode auxiliar nas Relações Públicas, devemos utilizar métricas adequadas ao problema que queremos estudar. Um exemplo é estudarmos um problema contextualizado no engajamento em redes sociais. Nesse caso, as métricas podem ser representadas por diferentes variáveis relacionadas ao impacto das atividades de redes e mídias sociais.

Apresentamos a seguir variáveis que ajudam a entender como o público interage com o conteúdo em redes sociais e como estas podem ser úteis na extração de informações que podem auxiliar em estratégias de marketing e comunicação.

1. **Curtidas ou Reações**: Número de curtidas ou reações específicas (como amor, haha, triste, etc.) que uma postagem recebe.

2. **Comentários**: Número de comentários feitos em uma postagem. Comentários são um forte indicador de engajamento, pois requerem mais esforço por parte do usuário do que simplesmente curtir.

3. **Compartilhamentos ou Retweets**: Quantas vezes o conteúdo foi compartilhado ou retweetado. Os compartilhamentos ampliam o alcance do conteúdo e indicam um alto nível de engajamento.

4. **Cliques**: Total de cliques no conteúdo, link ou imagem da postagem. Isso mostra o interesse em saber mais sobre o conteúdo apresentado.

5. **Visualizações de Página/Perfil**: Número de vezes que a página ou perfil foi visualizado.

6. **Mencionações**: Quantas vezes a marca ou a conta foi mencionada em outras postagens.

7. **Novos Seguidores ou Assinantes**: O número de novos seguidores ou assinantes ganhos em um determinado período.

8. **Alcance**: Número total de pessoas que viram a postagem. Diferente de impressões, o alcance conta quantas pessoas únicas viram o conteúdo.

9. **Impressões**: Quantas vezes a postagem foi exibida em um feed, incluindo visualizações repetidas por uma única pessoa.

10. **Taxa de Engajamento**: Uma métrica composta que geralmente inclui curtidas, comentários e compartilhamentos divididos pelo número total de impressões ou alcance.

11. **Tempo de Visualização de Vídeo**: Quanto tempo os usuários gastam assistindo a um vídeo postado.

12. **Salvamentos ou Favoritos**: Quantas vezes as pessoas salvaram a postagem para ver mais tarde.

13. **Sentimento do Comentário**: Avaliação do tom dos comentários (positivo, neutro, negativo). Esta métrica qualitativa ajuda a entender a percepção do público sobre o conteúdo publicado.

14. **Tipo de Reação**: Além da quantidade de reações, a natureza específica das reações (como curtir, amar, rir, surpreender, entristecer, enfurecer) pode trazer informações sobre como o público está reagindo emocionalmente ao conteúdo.

15. **Categoria do Compartilhamento**: Classificação dos compartilhamentos baseada no contexto ou no conteúdo (por exemplo, compartilhamentos educativos, de entretenimento, informativos, etc.), ajudando a entender o motivo pelo qual o conteúdo está sendo compartilhado.

16. **Qualidade dos Comentários**: Avaliação da substância e relevância dos comentários, distinguindo entre feedback construtivo, perguntas, spam ou comentários irrelevantes.

17. **Fonte de Mencionações**: Classificação de onde as mencionações estão ocorrendo, como em blogs, fóruns, outras redes sociais, etc., para entender melhor o alcance e o impacto da marca.

18. **Demografia do Público**: Informações sobre a idade, gênero, localização, interesses e outros dados demográficos dos seguidores ou de quem interage com o conteúdo, fornecendo informações sobre o público-alvo.

19. **Tipo de Conteúdo**: Classificação do conteúdo postado (por exemplo, vídeo, imagem, texto, link, etc.), para analisar qual tipo de conteúdo gera mais engajamento.

20. **Feedback Direto dos Usuários**: Opiniões e sugestões coletadas diretamente dos usuários, por meio de pesquisas ou caixas de sugestões, oferecendo informações sobre a percepção do público.


As métricas anterirmente apresentadas fornecem uma visão sobre o desempenho do conteúdo e qual é o nível de interação do público com a marca, personalidade, etc. nas redes sociais. Elas podem ajudar a melhorar estratégias de comunicação, melhorar a qualidade do conteúdo e aumentar o engajamento do público.

## 3.2 Dependência entre Variáveis

Na análise de dados, muitas vezes estamos interessados em entender **se existe uma correlação** entre duas ou mais variáveis. Para isso, precisamos criar hipóteses sobre qual(is) variável(is) influencia(m) no comportamento de outra(s). Isso requer definir qual(is) será(ão) a(s) variável(is) dependente(s) e independente(s). Finalmente, podemos compreender se e como essas variáveis se correlacionam. 

Para podermos fazer essa análise, vamos primeiro compreender o que são as Variáveis Dependentes e Independentes.

### Variáveis Dependentes e Independentes

Analisar a dependência entre variáveis quer dizer tentar compreender como uma variável (ou mais) influencia em outra(s) variável(is). Para isso, modelamos como a **variável dependente** a variável que achamos que é influenciada por outra.  Por outro lado, a **variável independente** é a que acreditamos ter algum efeito sobre a variável dependente, isto é, a que influencia a variável dependente fazendo com que os valores da independente alterem os valores da dependente. 

Por exemplo,  oderíamos considerar a "quantidade de conteúdo publicado em redes sociais" como uma variável independente e a "quantidade de novos seguidores", representadno o nível de engajamento do público, como uma variável dependente. Nesse contexto, a hipótese seria que a frequência de publicações (variável independente) influencia o engajamento do público (variável dependente), seja aumentando-o ou diminuindo-o.


````python
import numpy as np
import matplotlib.pyplot as plt

# Gerar dados simulados
np.random.seed(0)
quantidade_postagens = np.linspace(1, 30, 100)  # De 1 a 30 postagens
engajamento = np.square(quantidade_postagens) / (0.1 * quantidade_postagens + 1) 
engajamento += np.random.normal(0, 30, 100)  # Adicionando um pouco de ruído

# Criar o gráfico
plt.figure(figsize=(10, 6))
plt.scatter(quantidade_postagens, engajamento, color='blue')
plt.title('Relação entre Quantidade de Conteúdo Publicado e Engajamento do Público')
plt.xlabel('Quantidade de Conteúdo Publicado (Postagens por Dia)')
plt.ylabel('Qtde. de Novos Seguidores (Por Dia)')
plt.grid(True)
plt.show()
````

## 3.3 Correlação entre Variáveis

### Introdução

Uma vez aprendido o que são as Variáveis Dependentes e Independentes, podemos avançar e fazer uma análise mais profunda: investigar se e como essas variáveis se correlacinonam. **Correlação entre variáveis** refere-se a uma relação estatística entre duas variáveis. Quando duas variáveis estão correlacionadas, significa que, quando uma muda, a outra também tende a mudar, de forma previsível. 

Contudo, é muito importante saber que essa relação não implica que uma variável seja a causa da mudança na outra. Ou seja, uma **correlação não implica necessariamente causalidade**. Em outras palavras, pode acontecer que a correlação é uma mera coincidência como são os casos das [correlações espúrias](https://pt.wikipedia.org/wiki/Rela%C3%A7%C3%A3o_esp%C3%BAria). Veja o exemplo de uma correlação espúria na figura a seguir:

<div style="text-align: center">
<img src="spurious-correlations.png"></img>
</div>

Por isso, é fundamental compreender bem o problema estudado para fazer uma modelagem adequada, isto é, **a escolha adequadas das variáveis e suas dependências** para que haja boas chances de que as possíveis correlações encontradas possam sugerir alguma causalidade.

Tomemos esse exemplo das Relações Públicas onde uma agência pode observar que o aumento da frequência de publicações em mídias sociais está correlacionado com um maior engajamento do público, por exemplo, medido em curtidas, compartilhamentos e comentários. A primeira impressão pode sugerir que postar mais frequentemente causa um aumento no engajamento. No entanto, essa correlação não garante causalidade. Pode haver outros influenciadores como a qualidade do conteúdo, eventos atuais relevantes ou campanhas de marketing simultâneas.

Para ilustar o exemplo anterior, o gráfico a seguir representa a correlação entre a frequência de postagens em mídias sociais e o engajamento do público. Como pode ser visto, ambas as variáveis parecem se mover juntas ao longo do tempo. No entanto, é importante lembrar que este gráfico mostra apenas uma correlação, não necessariamente uma causalidade direta. Podem haver muitos outros fatores que influenciam o engajamento e uma análise mais aprofundada seria necessária para estabelecer qualquer relação de causalidade.

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Gerando dados fictícios
np.random.seed(0)
nof_days = 20
datas = pd.date_range('2023-01-01', periods=nof_days, freq='D')
frequencia_postagens = np.random.randint(1, 4, size=nof_days)  # 1 a 3 postagens por dia
engajamento = frequencia_postagens * np.random.uniform(0.8, 1.2, size=nof_days)  # Engajamento relacionado à frequência

# Criando um DataFrame
dados = pd.DataFrame({'Data': datas, 'Frequência de Postagens': frequencia_postagens, 'Engajamento': engajamento})

# Gráfico
# Ajustando o gráfico para ter escalas diferentes no eixo Y

plt.figure(figsize=(12, 3))

# Criando o primeiro eixo (para frequência de postagens)
ax1 = plt.gca()  # Obtém o eixo atual
ax1.plot(dados['Data'], dados['Frequência de Postagens'], label='Frequência de Postagens', color='blue')
ax1.set_ylabel('Frequência de Postagens', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Criando um segundo eixo (para engajamento) compartilhando o mesmo eixo X
ax2 = ax1.twinx()  # Cria um segundo eixo Y que compartilha o mesmo eixo X
ax2.plot(dados['Data'], dados['Engajamento'], label='Engajamento', color='green')
ax2.set_ylabel('Engajamento', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Configurações adicionais do gráfico
plt.title('Correlação entre Frequência de Postagens e Engajamento em Mídias Sociais')
ax1.set_xlabel('Data')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
````


Agora que entendemos que duas ou mais variáveis podem estar correlacionadas, precisamos de mais precisão na nossa análise. Ou seja, precisamos de **medidas quantitativas que nos permita facilmente idenditificar se há alguma correlação e como ela é**, pois pode ser difícil identificar essas correlações em conjunto de dados muitos grandes.


Há vários métodos para quantificar a correlação entre variáveis. Apresentamos aqui duas formas de calcular o grau de correlações entre variáaveis de acordo com a natureza da relação:


- Coeficiente de Pearson ($r$): para correlações lineares
- Coeficiente de correlação de Kendall ($\tau$): para correlações não lineares

### Coeficiente de Pearson: Medindo uma Correlação Linear

O [Coeficiente de Correlação de Pearson](https://en.wikipedia.org/wiki/Pearson_correlation_coefficient), denotado por $r$, é uma medida estatística que quantifica a relação linear entre duas variáveis quantitativas. O Ele é utilizado em diversas áreas para medir a intensidade e a direção de uma relação linear entre duas variáveis. O Coeficiente de Correlação de Pearson varia entre $-1$ e $+1$, onde:

- **$+1$** indica uma correlação positiva perfeita (quando uma variável aumenta, a outra também aumenta na mesma proporção).
- **$0$** indica que não há correlação linear entre as variáveis.
- **$-1$** indica uma correlação negativa perfeita (quando uma variável aumenta, a outra diminui na mesma proporção).

Interpretação do Coeficiente:

- **0.9 a 1.0 (-0.9 a -1.0)**: Indica uma correlação muito forte. Valores próximos a 1 ou -1 mostram que as variáveis têm uma relação linear quase perfeita, sendo positiva ou negativa, respectivamente.
- **0.7 a 0.9 (-0.7 a -0.9)**: Representa uma correlação forte, sugerindo uma relação linear significativa entre as variáveis.
- **0.5 a 0.7 (-0.5 a -0.7)**: Correlação moderada, indicando uma relação linear razoável, mas não extremamente forte.
- **0.3 a 0.5 (-0.3 a -0.5)**: Uma correlação fraca, que sugere uma relação linear mais fraca e menos consistente.
- **0 a 0.3 (0 a -0.3)**: Correlação desprezível ou nenhuma correlação, indicando pouca ou nenhuma relação linear entre as variáveis.

É importante notar que esses intervalos são orientações gerais e podem variar em diferentes contextos. Além disso, o Coeficiente de Pearson não detecta relações não lineares e é sensível a outliers, o que pode afetar a interpretação em certas situações de dados.

**Limitações**:

- **Somente para Relações Lineares**: O coeficiente de Pearson só mede a força de uma relação linear. Relações não lineares podem não ser bem representadas.
- **Influência de Valores Extremos**: Outliers podem distorcer significativamente o resultado do coeficiente.
- **Não Implica Causalidade**: Uma correlação alta (positiva ou negativa) não implica que uma variável cause mudanças na outra.

Por fim, Coeficiente de Correlação de Pearson é muito útil para identificar e quantificar correlações lineares entre variáveis. No entanto, é importante sempre considerar outras análises para compreender completamente a natureza das relações entre variáveis como, por exemplo, o coeficiente de Kendall que estudaremos a seguir para computar correlações não lineares.

### Coeficiente de Kendall: Medindo uma Correlação Não Linear

O Coeficiente de Correlação de Kendall, frequentemente representado por $\tau$, é uma medida estatística que ideintifica e quantifica a força e a direção de uma relação não linear entre duas variáveis. Diferentemente do Coeficiente de Pearson, que mede correlações lineares, o Coeficiente de Kendall é mais adequado para identificar relações, sejam elas lineares ou não. O Coeficiente de Kendall mede a concordância entre os pares de observações, variando entre -1 e +1, onde:

- **+1** indica uma correlação positiva perfeita (todas as observações concordam).
- **0** indica que não há correlação não linear entre as variáveis.
- **-1** indica uma correlação negativa perfeita (todas as observações discordam).

**Interpretação do Coeficiente**:

A interpretação do Coeficiente de Kendal baseia-se na magnitude do valor, que indica a força da associação não linear entre as variáveis. Os intervalos típicos para interpretação podem ser:

- $-1$ a $-0.5$: Correlação negativa forte. Indica uma forte tendência de que, quando uma variável aumenta, a outra diminui de forma não linear.
- $-0.5$ a $-0.2$: Correlação negativa moderada. Uma tendência menos pronunciada de que as variáveis se movem em direções opostas.
$-0.2$ a $0.2$: Correlação fraca ou nenhuma correlação. Sinaliza uma associação muito fraca ou inexistente entre as variáveis.
- $0.2$ a $0.5$: Correlação positiva moderada. Mostra uma tendência moderada de que as variáveis se movem juntas, mas não de forma linear.
- $0.5$ a $1$: Correlação positiva forte. Indica uma forte associação na qual as variáveis tendem a aumentar ou diminuir juntas, embora a relação não seja linear.

É importante ressaltar que esses intervalos são guias gerais e podem variar dependendo do contexto específico e da natureza dos dados. Além disso, é essencial considerar outros fatores, como o tamanho da amostra e a distribuição das variáveis, ao interpretar o Coeficiente de Kendal.

**Limitações**:

- **Complexidade Computacional**: Pode ser computacionalmente mais intensivo, especialmente para grandes conjuntos de dados.
- **Limitado a Relações Monotônicas**: Não detecta bem relações não monotônicas.
- **Não Indica Causalidade**: Como com Pearson, uma correlação alta (positiva ou negativa) não implica causalidade.

O Coeficiente de Kendall é útil para analisar relações entre variáveis em muitas áreas, como comunicação social, biologia e economia. É mais útil ainda **quando as relações entre as variáveis são suspeitas de serem não lineares ou quando os dados contêm outliers**. Por fim, ao usar o Coeficiente de Kendall, é importante considerar o contexto dos dados e complementar a análise com outras medidas e métodos estatísticos para obter uma compreensão completa das relações em estudo. # 4. Amostragem

No mundo das Relações Públicas, assim como em outras áreas, frequentemente nos deparamos com a necessidade de entender e analisar grandes conjuntos de dados. Esses dados podem ser sobre opiniões do público, comportamento de consumo, entre outros. Porém, muitas vezes, é impraticável ou impossível analisar todos os elementos de um grupo (população). Então, utilizamos uma **amostra** dessa população.

## 4.1 População e Amostra

Segundo Bussab e Morettin (2017), _"População é o conjunto de todos os elementos ou resultados sob investigação. Amostra é qualquer subconjunto da população."_ [1]. Ou seja, uma amostra é um subconjunto selecionado de uma população maior. Na prática, a **amostra representa uma porção dos dados que buscamos entender**.

Quando uma pesquisa é feita com toda a população, ela é chamada de **pesquisa censitária**. Pesquisas censitárias nos trazem uma visão completa sobre o estudo, pois todos os dados da população foram coeltados. Todavia, é muito custoso ou inviável realizar tal pesquisa. Por exemplo, ao realizar um censo,entrevistando todos os consumidores de um produto, para realizar uma pesquisa de de opinião pública, poder custar muito caro ou ser fisicamente inviável. É o caso de uma instituição que só tem um profissional para realizar a pesquisa de opinião pública: é custoso contratar mais profissionais para entrevistar todos os e é inviável que essa pessoa sozinha consiga entrevistar todos os consumidores.

<div style="text-align: center">
<img src="figures/inviabilidade-opiniao-publica-com-populacao.png" width="350" ></img>
</div>

Por outro lado, quando uma pesquisa é feita com parte dessa população, ela é chamada de **pesquisa amostral**. Assim, essa pesquisa de Relações Públicas pode se concentrar em um grupo menor que representa esse universo maior de todos os consumidores. As vantagens de se utilizar uma amostra são:

- Praticidade: Muitas vezes é inviável coletar dados de toda a população.
- Custo e Tempo: Coletar e analisar dados de uma amostra é mais rápido e menos custoso.
- Viabilidade: Alguns testes e análises só são possíveis através de amostras.

Outro exemplo nas Relações Püblicas é a Análise de Engajamento. Nesse caso, estuda-se o engajamento dos seguidores em uma plataforma de mídia social escolhendo algumas postagens para análise ao invés de avaliar todas as postagens.

Mas quais critérios devemos utilizar para escolher a amostra? É o que vamos estudar na próxima seção.


## 4.2 Tipos de Amostragem

Um fator essencial na análise de dados é a seleção de **amostras representativas**, ou seja, amostras que contenham partes quantitativamente equivalentes às suas proporções na população. Dessa forma, aumentamos as chances de que os resultados obtidos da análise das amostras sejam correspondam bem à representação da população. Em outras palavras, a amostra deve incluir uma variedade de elementos que reflitam a diversidade da população total de forma que as proporções de diferentes subgrupos dentro da amostra devem corresponder às da população.

Podemos dividir as os tipos de amostragem entre probabilísticos e não probailísticos (Anunciação, 2021), conforme descritas a seguir:

- **Amostragem Probabilística**: cada membro da população é escolhido aleatoriamente e sabemos qual é a probabilidade de cada membro ser selecionado para a amostra. Isso garante uma representatividade mais precisa da população total, permitindo generalizações mais confiáveis. 

- **Amostragem Não Probabilística**: seleção intencional de membros para a amostra, sem usar um processo aleatório. Isso pode incluir escolhas baseadas em critérios específicos ou conveniência.

A Tabela a seguir, adaptada de (Anunciação, 2021), compila as vantagens de cada tipo de amostragem:

| Critério             | Probabilística                           | Não-Probabilística                       |
|----------------------|------------------------------------------|-----------------------------------------|
| Seleção              | Aleatória, cada membro tem uma probabilidade conhecida | Intencional, sem processo aleatório         |
| Representatividade   | Alta                                     | Baixa                                    |
| Custo e Recursos     | Maior                                  | Menor                                    |
| Complexidade         | Maior                                    | Menor                                    |
| Generalização        | Possível e confiável                     | Limitada                                 |
| Aplicações           | Pesquisas amplas, estudos demográficos   | Pesquisas exploratórias, estudos de caso |
| Exemplos em RP       | Análise demográfica de um público-alvo para uma campanha de conscientização | Pesquisa de opinião de clientes em um evento promocional específico |

A Tabela anterior inclui dois exemplos da área de Relações Públicas que são detalhados a seguir:

- **Amostragem Probabilística em RP**: A complexidade da amostragem probabilística em Relações Públicas pode ser ilustrada por um projeto como a realização de uma análise demográfica detalhada do público-alvo para uma campanha de conscientização. Este processo exigiria primeiro a definição clara da população-alvo, seguida pela criação de um mecanismo aleatório para a seleção de indivíduos, garantindo que cada membro da população tenha uma chance conhecida e igual de ser escolhido. Isso pode envolver o uso de softwares de randomização e a necessidade de acessar ou criar uma lista abrangente da população. Além disso, para garantir a representatividade, a amostra teria que ser suficientemente grande e bem distribuída, considerando várias variáveis demográficas, como idade, gênero, localização geográfica, entre outras. Isso exige um planejamento cuidadoso e uma análise estatística mais complexa para assegurar que os resultados sejam confiáveis e possam ser generalizados para a população total.

- **Amostragem Não Probabilística em RP**: Em contraste, a amostragem não-probabilística, embora menos complexa, apresenta seus próprios desafios. Por exemplo, na coleta de feedback de clientes durante um evento promocional específico, a amostra é escolhida com base na conveniência. Os participantes são aqueles que estão disponíveis e dispostos a fornecer feedback no momento, o que pode levar a um viés na amostra. Essa abordagem não garante que a amostra seja representativa da população total de clientes. Além disso, a análise dos dados coletados pode ser limitada, pois as conclusões tiradas são aplicáveis apenas aos indivíduos presentes no evento, sem possibilidade de generalização. A seleção intencional e a natureza oportunista dessa amostragem simplificam a coleta de dados, mas requerem cuidado na interpretação e aplicação dos resultados obtidos.

## 4.3 Técnicas de Amostragem

A amostragem é o processo de selecionar uma subparte de uma população para estudo. Isso é feito para economizar tempo, diminuir custos e, muitas vezes, para viabilizar o estudo. Existem muitas técnicas de amostragem diferentes, cada uma com suas próprias vantagens e desvantagens. Selecionar a amostra apropriada é uma das etapas mais importantes no processo de análise de dados, pois cada uma dessas técnicas de amostragem tem suas vantagens e limitações. A escolha entre elas depende dos objetivos específicos da pesquisa em Relações Públicas, dos tipos de dados a serem coletados, da natureza da população-alvo bem como dos recursos disponíveis para realizar a pesquisa.

<div style="text-align: center">
<img src="figures/tecnicas-amostragem.png" width="350" ></img>
</div>

A seguir, apresentamos algumas técnicas de amostragem usadas frequentemente.

1. **Amostragem Aleatória Simples**: Em Relações Públicas, essa técnica pode ser usada para selecionar uma amostra representativa de um público-alvo para uma pesquisa de opinião. Essa técnica é ideal para pesquisas gerais de opinião pública, avaliações de percepção da marca ou medição da eficácia de uma campanha de comunicação. Por exemplo, de uma lista de seguidores de uma marca em uma rede social, selecionar aleatoriamente um número específico para participar de uma pesquisa sobre a percepção da marca.  

2. **Amostragem Sistemática**: Prática para pesquisas em que é necessário selecionar participantes de uma grande lista, como uma base de dados de clientes ou assinantes. A seleção sistemática de cada _n_-ésimo indivíduo -- saltanto de _n_ em _n_ inivídous -- pode ser utilizada para pesquisas de satisfação ou avaliação de serviços. Por exemplo, de uma lista de email de assinantes, selecionar a cada décimo inscrito para receber e avaliar diferentes tipos de _newsletters_.

3. **Amostragem Estratificada**: Utilizada para assegurar que todos os segmentos importantes da população sejam representados como, por exemplo, idade, gênero ou localização geográfica. Em Relações Públicas, isso poderia ser usado para garantir representatividade em uma pesquisa sobre a percepção de uma campanha publicitária. Por exemplo, dividir o público em estratos com base em idade e gênero e, então, retirar amostras aleatórias de cada estrato para entender como diferentes segmentos percebem a campanha.

4. **Amostragem por Conglomerados**: Esta técnica pode ser usada para facilitar a realizar pesquisas de mercado em diferentes localizações geográficas. Por exemplo, se uma agência de RP quiser avaliar a percepção de uma marca em diferentes cidades, ela pode dividir o mercado em conglomerados (regiões) e selecionar algumas dessas regiões para realizar a pesquisa. Isso permite uma compreensão mais abrangente da percepção regional da marca.

5. **Amostragem por Quotas**: Esta técnica envolve selecionar amostras com base em quotas pré-definidas para diferentes subgrupos dentro da população. O objetivo é garantir que características específicas sejam representadas na amostra. Por exemplo, em uma campanha de RP voltada para a conscientização sobre saúde, pode ser importante incluir quotas específicas de homens e mulheres de diferentes faixas etárias. Assim, a amostragem por quotas asseguraria a inclusão proporcional desses grupos para obter uma visão abrangente das diferentes percepções e atitudes em relação à campanha.

6. **Amostragem de Conveniência**: Esta abordagem utiliza indivíduos facilmente acessíveis ou disponíveis para a pesquisa. É uma forma rápida e econômica de coletar dados, mas com poucas chances de ser representativa da população total. Por exemplo, durante o lançamento de um novo produto, um profissional de RP pode realizar uma pesquisa rápida com os participantes do evento de lançamento. Essa amostra de conveniência permite coletar feedback instantâneo sobre o produto, embora os participantes possam não representar todos os potenciais consumidores do produto.

7. **Amostragem de Bola de Neve**: Utilizada para acessar e estudar populações que são difíceis de identificar ou alcançar. Começa com um pequeno grupo de participantes conhecidos que, por sua vez, indicam outros membros de sua rede. Por exemplo, ao estudar a influência de opiniões de líderes de pensamento em um nicho de mercado específico, um profissional de RP pode começar entrevistando alguns influenciadores conhecidos nesse nicho. Esses influenciadores, por sua vez, podem indicar outros colegas ou contatos relevantes, ampliando a pesquisa através da rede de contatos. Por isso o é conhecido como "bola de neve".

No entanto, se a amostra for suficientemente grande e for selecionada aleatoriamente, a média da amostra deve ser uma boa estimativa da média da população.

````python
import pandas as pd

def descreve_datasets(dataset1, dataset2=None, nome1='Dataset 1', nome2='Dataset 2', idioma='pt'):
    descricao1 = dataset1.describe()

    if dataset2 is not None:
        descricao2 = dataset2.describe()
        descricao = pd.concat([descricao1, descricao2], axis=1, keys=[nome1, nome2])
    else:
        descricao = descricao1.to_frame()
        descricao.columns = [nome1]

    if idioma == 'pt':
        descricao = descricao.rename(index={
            'count': 'contagem',
            'mean': 'média',
            'std': 'desvio padrão',
            'min': 'mínimo',
            '25%': 'Q1 (25%)',
            '50%': 'mediana',
            '75%': 'Q3 (75%)',
            'max': 'máximo'
        })
    elif idioma == 'en':
        descricao = descricao.rename(index={
            'count': 'count',
            'mean': 'mean',
            'std': 'standard deviation',
            'min': 'minimum',
            '25%': 'Q1 (25%)',
            '50%': 'median',
            '75%': 'Q3 (75%)',
            'max': 'maximum'
        })
    else:
        raise ValueError('Idioma não suportado')

    print(descricao.to_markdown() + "\n")


import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans

# Definindo a semente do gerador de números aleatórios para reprodutibilidade
np.random.seed(19)

# Gerando dados da população
populacao_tam = 100000
populacao = np.random.normal(170, 10, populacao_tam) # Suponha que temos 100000 estudantes universitários no Brasil
populacao_df = pd.DataFrame(populacao, columns=['altura'])

# Selecionando uma amostra para cada tipo de amostragem
amostra_tam = 5000
# Amostragem Aleatória
amostra_aleatoria = np.random.choice(populacao, amostra_tam)

# Amostragem Sistematica
n = len(populacao_df) // amostra_tam # pega cada nth membro da população
# FIXME o plot estratificado n tá com intervalos iguais
# print(" N é " + str(n))
amostra_sistematica = populacao_df.iloc[::n, :].values.flatten()

# Amostragem Estratificada
populacao_df['estrato'] = pd.cut(populacao_df['altura'], bins=[0, 170, np.inf], labels=['abaixo da média', 'acima da média'])
strat_train_set, strat_test_set = train_test_split(populacao_df, test_size=0.25, stratify=populacao_df['estrato'])
amostra_estratificada = strat_test_set['altura'].values

# Amostragem por Conglomerados
# kmeans = KMeans(n_clusters=10, random_state=0).fit(populacao_df[['altura']])
kmeans = KMeans(n_clusters=3, n_init=10, random_state=0).fit(populacao_df[['altura']])

populacao_df['cluster'] = kmeans.labels_
amostra_por_conglomerados = populacao_df[populacao_df['cluster'] == 0]['altura'].values

# Criando uma lista com todos os tipos de amostragem
tipos_de_amostragem = [("Amostragem Aleatória", amostra_aleatoria), 
                       ("Amostragem Sistemática", amostra_sistematica), 
                       ("Amostragem Estratificada", amostra_estratificada), 
                       ("Amostragem por Conglomerados", amostra_por_conglomerados)]

# Gerando um gráfico de histograma para cada tipo de amostragem
for tipo, amostra in tipos_de_amostragem:
    plt.figure(figsize=(12,6))
    plt.hist(populacao, bins=100, alpha=0.5, label='População')#, density=True)
    plt.hist(amostra, bins=100, alpha=0.5, label=tipo)#, density=True)

    # Adicionando a legenda
    plt.legend(loc='upper right')

    # Adicionando títulos e labels
    plt.title(f'Distribuição da Altura da População - {tipo}')
    plt.xlabel('Altura (cm)')
    plt.ylabel('Frequência')

    # Imprimindo statísticas de resumo da população e da amostra
    print("\n\n===========   " + tipo.upper() + "   ===========")
    print("\n------------ Medidas-Resumo ------------\n")
    descreve_datasets(pd.Series(populacao), pd.Series(amostra), nome1="População", nome2="Amostra", idioma='pt')

    # Mostrando o gráfico
    plt.show()
    print()
````

## 4.4 Erros de Amostragem

Ao trabalhar com amostras, erros podem surgir no processo de amostragem e podem afetar a representatividade e precisão da amostra em relação à população. Esses erros ocorrem quando uma amostra selecionada de uma população não reflete exatamente as características da população total. Em outras palavras, é a diferença entre os resultados obtidos na amostra e os verdadeiros valores que seriam obtidos se toda a população fosse examinada. 

<div style="text-align: center">
<img src="figures/sampling_errors.png" width="350" align="center"></img>
</div>

Os erros amostrais são inerentes ao processo de amostragem, pois é praticamente impossível que uma pequena parte represente perfeitamente o todo. Eles podem ser influenciados por vários fatores, como o tamanho da amostra, o método de seleção e a variabilidade inerente à população. No campo dinâmico das Relações Públicas, onde a compreensão precisa do público é essencial, compreender e minimizar os erros de amostragem torna-se, pois, fundamental para garantir a confiabilidade e a validade dos resultados de uma pesquisa.

Nas Relações Públicas, alguns erros de amostragem são mais relevantes devido ao impacto direto que eles têm na interpretação de dados e na tomada de decisões estratégicas. Destacam-se:

1. **Erro de Viés de Seleção**. O viés de seleção ocorre quando há um viés na seleção da amostra, ou seja, a maneira como você seleciona sua amostra faz com que ela seja não representativa da população. Uma seleção enviesada de amostra pode levar a estalos (_insights_) e estratégias baseadas em uma compreensão distorcida do público. Por exemplo, se apenas usuários ativos de mídias sociais são incluídos em uma pesquisa sobre percepções de marca, isso pode excluir opiniões de segmentos importantes do público que não são tão ativos ou que não usam as mídias sociais utilizadas na pesquisa.

1. **Erro de Amostragem Aleatório**: Mesmo com um processo de seleção aleatória perfeito, ainda haverá um erro de amostragem, que é a diferença entre a verdadeira média da população e a média da amostra. De uma forma geral, quanto maior a amostra, menor será o erro de amostragem. Há um ponto de diminuição, onde amostras maiores não reduzirão muito o erro de amostragem.

1. **Erro de Não Resposta**: Comum em pesquisas e sondagens, esse erro ocorre quando indivíduos selecionados para a amostra não respondem. Em RP, isso pode resultar em uma compreensão incompleta ou distorcida do público-alvo, especialmente se os não respondentes compartilharem características ou opiniões comuns.

1. **Erro de Cobertura**: Este erro acontece quando parte da população foi omitida no momento de escolher as amostras. não é utilizada toda a população para fazer a lista ou o quadro da amostra (a partir do qual a amostra é selecionada) não inclui todos os elementos da população-alvo. Assim, mesmo com uma seleção aleatória, a amostra pode não representar toda a população. Esse erro ocorre quando uma amostra que não cobre toda a população de interesse. Em RP, ao avaliar a eficácia de campanhas ou o alcance da marca, uma amostra pode não cobrir toda a população de interesse pode levar a conclusões imprecisas sobre a percepção geral da marca.

1. **Erro de Viés de Medição**: O erro de viés de medição, também conhecido como erro de medição, é um tipo de erro que ocorre quando há uma distorção sistemática nos dados coletados durante o processo de pesquisa. Este erro não está relacionado à escolha da amostra, mas sim à forma como as informações são coletadas, processadas ou interpretadas. Pode ocorrer devido a perguntas mal formuladas em pesquisas ou interpretações equivocadas dos dados. Em Relações Públicas, o erro de viés de medição pode ter um impacto significativo, pois pode levar a uma compreensão distorcida do público, eficácia das campanhas ou percepção da marca. Por exemplo, se uma pesquisa para avaliar a percepção da marca usa perguntas sugestivas ou ambíguas, os resultados podem não refletir verdadeiramente as atitudes dos consumidores.

1. **Erro de Viés de Informante**: Ocorre quando os dados coletados de participantes ou respondentes não são precisos ou são distorcidos devido a várias razões. Esse tipo de erro afeta a qualidade e a confiabilidade das informações obtidas em uma pesquisa. As principais causas do erro de viés de informante incluem a vontade de agradar o entrevistador ou à percepção social, esquecimento ou memória seletiva, entendimento incorreto (interpretar mal as perguntas ou o propósito da pesquisa), sensibilidade ao assunto como tabus, entre outros.

1. **Erro de Viés de Contexto**: Ocorre quando as circunstâncias ou o ambiente em que a pesquisa é conduzida influenciam as respostas dos participantes de maneira a distorcer os resultados. Esse tipo de viés pode ser introduzido por diversos fatores relacionados ao contexto da coleta de dados, afetando a maneira como os respondentes interpretam as perguntas e como escolhem suas respostas. Por exemplo, fatores comom ambiente físico (formal ou informal, temperatura desagradável), contexto social ou cultural, eventos recentes (falecimento de ente próximo), contexto temporal da pesquisa (após um time de futebol ser campeão), dentre outros.

1. **Erro de Viés de Voluntário**: Em RP, especialmente em pesquisas abertas ou convites para feedback, os indivíduos que escolhem participar podem não ser representativos do público maior, levando a compreensões potencialmente enviesadas.


1. **Dados Atípicos e Outliers**: Dados típicos são valores que são significativamente diferentes da maioria dos outros valores na amostra. O erro associado a dados atípicos e outliers refere-se ao impacto que observações extremas ou anômalas podem ter sobre os resultados de uma pesquisa ou análise de dados. As causas comuns de dados atípicos são erros de coleta ou no processamento dos dados ou erros acidentais na coleta de dados -- ambos _outliers_ --, variação natural de uma opulação heterogênea


Cada um desses erros pode afetar a eficácia e a precisão das estratégias e campanhas em Relações Públicas, tornando fundamental para os profissionais da área entender e trabalhar para mitigar esses erros em suas pesquisas e análises. Para minimizar os erro de amostragem, os profissionais de RP podem procurar criar um ambiente neutro para a pesquisa, considerar cuidadosamente o momento da pesquisa e utilizar diferentes métodos e contextos para coletar dados, aumentando assim a confiabilidade dos resultados obtidos.

## 4.6 Outros métodos para calcular o tamanho da amostra

````python
import math

def tamanho_amostra_cochran_finita(populacao, margem_erro, confianca, proporcao_estimada=0.5):
    """
    Calcula o tamanho da amostra para uma população finita usando a fórmula de Cochran.

    :param populacao: Tamanho total da população.
    :param margem_erro: Margem de erro desejada (em decimal).
    :param confianca: Nível de confiança desejado (em decimal).
    :param proporcao_estimada: Proporção estimada da característica na população (padrão é 0.5).
    :return: Tamanho da amostra calculado.
    """
    z_score = abs(math.erf(confianca / math.sqrt(2)))  # Z-score para o nível de confiança
    tamanho_amostra_infinita = (z_score ** 2) * proporcao_estimada * (1 - proporcao_estimada) / (margem_erro ** 2)
    tamanho_amostra = tamanho_amostra_infinita / (1 + (tamanho_amostra_infinita - 1) / populacao)
    return round(tamanho_amostra)

def tamanho_amostra_cochran_infinita(margem_erro, confianca, proporcao_estimada=0.5):
    """
    Calcula o tamanho da amostra para uma população infinita usando a fórmula de Cochran.

    :param margem_erro: Margem de erro desejada (em decimal).
    :param confianca: Nível de confiança desejado (em decimal).
    :param proporcao_estimada: Proporção estimada da característica na população (padrão é 0.5).
    :return: Tamanho da amostra calculado.
    """
    z_score = abs(math.erf(confianca / math.sqrt(2)))  # Z-score para o nível de confiança
    tamanho_amostra_infinita = (z_score ** 2) * proporcao_estimada * (1 - proporcao_estimada) / (margem_erro ** 2)
    return round(tamanho_amostra_infinita)

def tamanho_amostra_yamane(populacao, margem_erro):
    """
    Calcula o tamanho da amostra usando a fórmula de Yamane para populações finitas.

    :param populacao: Tamanho total da população.
    :param margem_erro: Margem de erro desejada (em decimal).
    :return: Tamanho da amostra calculado.
    """
    tamanho_amostra = populacao / (1 + populacao * (margem_erro ** 2))
    return round(tamanho_amostra)


# Exemplo de uso das funções
populacao_exemplo = 10000
margem_erro_exemplo = 0.05
confianca_exemplo = 0.95

tamanho_amostra_cochran = tamanho_amostra_cochran_finita(populacao_exemplo, margem_erro_exemplo, confianca_exemplo)
tamanho_amostra_cochran_infinita = tamanho_amostra_cochran_infinita(margem_erro_exemplo, confianca_exemplo)
tamanho_amostra_yamane = tamanho_amostra_yamane(populacao_exemplo, margem_erro_exemplo)

print(f'tamanho_amostra_cochran = {tamanho_amostra_cochran}')
print(f'tamanho_amostra_cochran_infinita = {tamanho_amostra_cochran_infinita}')
print(f'tamanho_amostra_yamane = {tamanho_amostra_yamane}')



````

## 5.1 Média

### 5.1.1 Definição e Exemplos

A média aritmética é provavelmente a medida de posição mais conhecida. A média de um conjunto de dados é calculada **somando-se todos os números e depois dividindo-se pelo número total de números**. A fórmula da média é:

$ \bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i $

Onde:
- $\bar{x}$ representa a média da amostra,
- $n$ é o número total de observações na amostra,
- $x_i$ representa cada valor individual na amostra,
- $\sum$ é o símbolo de somatório, indicando a soma de todos os valores $x_i$.

Por exemplo, para ter uma ideia geral do período do entrevistado e da sua pretensão salarial, calculamos as **médias** do período e da pretensão salarial dos entrevistados do conjunto de dados CD-2; conforme calculado pelo programa a seguir. Esses valores médios mostrados abaixo nos dão uma visão geral em qual período os entreistados estão e qual é a sua pretensão salarial. No caso do período, como se trata de uma variável discreta, pode-se arredondar par cima ou para baixo o valor da média para representar de forma discreta o período dos entresvistados.

````python
media_periodo = df_pesquisa_opiniao["Q11 Período"].mean()
print(f'A média do período dos entrevistados é {media_periodo:.1f}')

media_pretensao_salarial = df_pesquisa_opiniao["Q09b Pretensão salarial"].mean()
print(f'A média da pretensão salarial é de R$ {media_pretensao_salarial:.1f}')
````

## 5.2 Mediana

A mediana é a medida que indica o **valor central de um conjunto de dados ordenados**, assim, dividindo o conjunto de dados em duas partes iguais. Para calcular a mediana, siga os passos: 

1. Organize os dados em ordem crescente.
1. Identifica o valor do meio, este será a mediana. Caso haja um número par de dados, a mediana é a média aritmética dos dois dados do meio.

Por exemplo, no Conjunto de Dados CD-3, a mediana é mostrada a seguir.

````python
mediana_pesquisa_opiniao_produto = df_pesquisa_opiniao_produto["Satisfação sobre o produto"].median()
print(f'A mediana das satisfações sobre o produto é {mediana_pesquisa_opiniao_produto}')
````

A mediana é uma medida de posição central robusta devido à sua resistência à influência de valores extremos ou atípicos. Ao contrário da média, que pode ser significativamente distorcida pela presença de tais valores, a mediana mantém sua confiabilidade, oferecendo uma representação mais estável da tendência central dos dados, especialmente em situações onde os valores atípicos são frequentes ou quando a distribuição dos dados é assimétrica.

Nas Relações Públicas, a aplicação da mediana permite analisar dados de forma mais confiável em casos de distribuições de dados assimétricas ou quando não conhecemos a distribuição dos dados. Por exemplo, considere uma empresa que coleta feedback de clientes após um evento de lançamento de produto. Se a maioria das avaliações for positiva, mas houver algumas críticas extremamente negativas, a média dessas avaliações pode ser distorcida por esses poucos pontos de dados negativos. No entanto, ao usar a mediana para analisar a satisfação do cliente, a empresa obtém uma visão mais equilibrada, que não é desproporcionalmente afetada por esses _outliers_. Isso permite que os profissionais de RP obtenham uma compreensão mais precisa do sentimento geral dos clientes, dessa forma, subsidiando a tomada de decisões estratégicas com maior fidedignidade.

## 5.3 Moda

A moda é uma medida de tendência central que **identifica o valor ou valores mais frequentes** em um conjunto de dados. Para encontrar a moda, basta identificar qual(is) valor(es) ocorre(m) com maior frequência. Se um conjunto de dados possui um único valor que aparece com mais frequência, é chamado de **unimodal**. Conjuntos de dados com duas modas são chamados de **bimodais**, e aqueles com mais de duas modas são chamados de **multimodais**. Por fim, podem haver situações em que o conjunto de dados não possui moda, ou seja, não há dados repetidos.

Por exemplo, no Conjunto de Dados CD-2, quem será o(a) entrevitador(a) que mais realizou entrevistas? Já no Conjunto de Dados CD-3, qual é a avaliação mais frequente? Veja os resultados a seguir.

Perceba que a moda é útil quando se deseja entender **qual é o comportamento ou opinião mais comum em um conjunto de dados**. Só que, ao contrário da média e da mediana, **a moda pode ser usada com dados qualitativos (categóricos)**.

Nas Relações Públicas, a moda é especialmente útil para identificar tendências e preferências dominantes em um público-alvo. ## 6.1 Amplitude Total (Faixa Total)

A amplitude é a diferença entre o valor máximo e o valor mínimo em um conjunto de dados: 

$Amplitude\ Total = Valor\ máximo - Valor\ mínimo$


Por exemplo, no Conjunto de Dados CD-2, suponhamos que selecionamos uma amostra da pretensão salarial de 10 estudantes e calculemos a Amplitude Total desses dados. Apresentamos essa amostra em tabela e em um gráfico a seguir, que ilustra o cálculo da Amplitutde Total.

````python
import matplotlib.pyplot as plt
import seaborn as sns


# Plotando um gráfico de barras para ilustrar a amplitude
plt.figure(figsize=(10, 4))
sns.barplot(x=sorted_names, y=sorted_salaries)
plt.ylim(0, 12000)  # limites do eixo y para visualização clara
plt.title(f"Amplitude das Pretensões Salariais = R$ {range_salary}")
plt.xlabel('Estudante')
plt.ylabel('Pretensão Salarial (R$)')

# Adicionando linhas para mostrar a amplitude
min_salary = min(sorted_salaries)
max_salary = max(sorted_salaries)
plt.axhline(y=min_salary, color='r', linestyle='--', label=f'Pretensão mínima = R$ {min_salary}')
plt.axhline(y=max_salary, color='b', linestyle='--', label=f'Pretensão máxima = R$ {max_salary}')

# Adicionando legenda
plt.legend()

# Adicionando anotação para a amplitude
mid_index = len(sorted_salaries) // 2
plt.annotate('', xy=(mid_index, min_salary), xytext=(mid_index, max_salary),
             arrowprops=dict(arrowstyle='<->', color='black'))
plt.text(mid_index + 0.5, (min_salary + max_salary) / 2, f'AMPLITUDE = R$ {range_salary}',
         rotation=90, verticalalignment='center', color='black')

plt.show()
````

Ademais, ela é sensível a valores atípicos (outliers), o que pode mascarar a verdadeira variabilidade dos dados. 

A Amplitutde Total também ignora a distribuição dos dados, por isso ela é útil para termos uma noção de quanto está variando os dados em relação aos valores máximo e mínimo. ## 6.2 Variância e Desvio Padrão

A variância e o desvio padrão são medidas estatísticas fundamentais na compreensão da dispersão de um conjunto de dados, permitindo uma compreensão mais completa sobre a variabilidade de um conjunto de dados. A variância e o desvio padrão são úteis para **identificar o quanto os valores em um conjunto de dados divergem da média**. 

É possível notar essa comparação com a média no cálculo da variância, pois a variância de um conjunto de dados é definida como a média das diferenças quadradas entre cada valor e a média do conjunto de dados. Para um conjunto de dados $X$ com $N$ observações (valores) $x_i$, onde $\bar{x}$ é a média dos valores, a variância, denotada por $\sigma^2$, é dada por:

$\sigma^2 = \frac{1}{N}\sum_{i=1}^{N}(x_i - \bar{x})^2$

Logo:

$\sigma = \sqrt{\frac{1}{N}\sum_{i=1}^{N}(x_i - \bar{x})^2}$

Por convenção, chamamos o **desvio padrão de $\sigma$** e como ele é definido como a **raiz quadrada da variância**representamos a variância por $\sigma^2$.

$\sigma = Desvio\ Padrão = \sqrt{\sigma^2}$

$\sigma^2 = Variância$

Extraimos a raiz quadrada da variância e a chamamos de **desvio padrão** para termos uma  medida de dispersão dos dados em relação à média em **na mesma medida dos dados originais**; uma vez que utilizamos o quadrado no cálculo da variância. Dessa maneira, o desvio padrão pode ser comparado com os valores do conjunto de dados diretamente. 

O desvio padrão é uma das medidas de dispersão mais utilizadas, pois indica quanto, em média, os valores no conjunto de dados desviam da média $\bar{x}$. Um desvio padrão baixo indica que os dados estão agrupados mais perto da média, enquanto um desvio padrão alto indica que os dados estão mais espalhados.

Por exemplo, utilizando o conjunto de dados anterior, criamos o gráfico a seguir, chamado de gráfico de dispersão. Ele oferece uma representação visual dos conceitos de média e desvio padrão, além de mostrar a distribuição dos dados das pretensões salarias dos estudantes.

## 6.3 Medidas Resumo de Quantis

Os quantis são valores derivados do conjunto de dados quando o dividimos em partes iguais. As medidas de quantis mais comuns são os quartis, os decils e os percentis.

### Percentis

Percentis são medidas que dividem um conjunto de dados em 100 partes iguais. São muito úteis na estatística para entender a distribuição dos dados. Por exemplo, **o percentil 50, também conhecido como mediana**, é o valor no qual 50% dos dados são menores e 50% dos dados são maiores.

Os percentis são muito úteis para entender onde um ponto de dados específico se encaixa no contexto de todo o conjunto de dados. Por exemplo, se a pretensão salarial de um estudante estiver no percentil 90, isso significa que esse estudante possui pretensão salarial maior que 90% dos demais.

````python
import numpy as np

# Calculando alguns percentis
percentil_30 = np.percentile(sorted_salaries, 30)
percentil_75 = np.percentile(sorted_salaries, 75)
percentil_90 = np.percentile(sorted_salaries, 90)

# Imprimindo os resultados
print("Percentis para as pretensões salariais:")
print(f"Percentil 30: R$ {percentil_30:.0f}")
print(f"Percentil 75: R$ {percentil_75:.0f}")
print(f"Percentil 90: R$ {percentil_90:.0f}")
````

### Quartis

Os quartis são um tipo específico de quantil que **dividem um conjunto de dados ordenados em quatro partes** iguais. Os quartis nos fornece são úteis para fornecer uma visão rápida sobre a distribuição dos dados, identificar valores atípicos e entender a dispersão dos dados em torno da **mediana**.

Existem três quartis:

1. **Primeiro quartil (Q1)**: também conhecido como quartil inferior, representa o valor abaixo do qual se encontram **25%** dos dados quando ordenados do menor para o maior. Em outras palavras, Q1 é o valor que divide os primeiros 25% dos dados do restante.

2. **Segundo quartil (Q2)**: é a **mediana**, que divide o conjunto de dados ao meio, ou seja, **50%** dos dados são menores que Q2 e 50% dos dados são maiores.

3. **Terceiro quartil (Q3)**: também conhecido como quartil superior, é o valor abaixo do qual **75%** dos dados se encontram. Q3, portanto, separa os 25% superiores dos dados do restante.



Por, exemplo, vamos calcular os quartis das pretensões salarias dos estudantes a seguir:

````python
import numpy as np

# Supondo que 'sorted_salaries' já tenha sido definido
# sorted_salaries = np.sort(np.append(np.round(np.random.normal(5000, 1000, 9)).astype(int), 10000))

# Calculando os quartis
quartil_1 = np.percentile(sorted_salaries, 25)  # Primeiro quartil (Q1)
quartil_2 = np.percentile(sorted_salaries, 50)  # Segundo quartil (Q2) ou Mediana
quartil_3 = np.percentile(sorted_salaries, 75)  # Terceiro quartil (Q3)

# Imprimindo os resultados
print("Quartis para as pretensões salariais:")
print(f"Primeiro Quartil (Q1): R$ {quartil_1:.0f}")
print(f"Segundo Quartil (Q2 ou Mediana): R$ {quartil_2:.0f}")
print(f"Terceiro Quartil (Q3): R$ {quartil_3:.0f}")
````

## 6.4 Amplitude Interquartil (Faixa Interquartil)

A Amplitude Interquartil (também conhecida como Faixa Interquartil ou _Interquartile Range - IQR_) é uma medida de dispersão estatística que representa a diferença entre o terceiro quartil (Q3) e o primeiro quartil (Q1):

$IQR = Q_3 - Q_1$

Ou seja, a Amplitude Interquartil abrange o intervalo entre os 25% e 75% dos dados quando ordenados do menor para o maior. Isso resulta em um valor que indica a "largura" da parte central dos dados, proporcionando **uma medida da dispersão dos dados ao redor da mediana**:

- Se a Amplitude Interquartil é grande, isso indica que os dados estão mais dispersos.
- Se for pequena, isso indica que os dados estão mais agrupados, menos dispersos.


Como a Amplitude Interquartil mede a **dispersão em torno da mediana**, isso a torna **uma medida de dispersão robusta**, isto é, ela é resistente a valores atípicos. A Amplitude Interquartil é útil para entender a variabilidade dos dados quando a distribuição não é simétrica ou quando há valores atípicos significativos. 

````python
import numpy as np

# Calculando a Faixa Interquartil (IQR)
iqr = quartil_3 - quartil_1

# Imprimindo os resultados
print(f"Primeiro Quartil (Q1): R$ {quartil_1:.0f}")
print(f"Terceiro Quartil (Q3): R$ {quartil_3:.0f}")
print(f"Faixa Interquartil (IQR = Q3 - Q1): R$ {iqr:.0f}")
````

## 6.5 Boxplot (Diagrama de Caixas)

### Componentes de um Boxplot

Um gráfico de caixa, também conhecido como **boxplot, permite visualizar a distribuição, dispersão e simetria dos dados através de seus quartis**. Cada elemento do boxplot está explicado a seguir:

- A **caixa central** do gráfico encapsula a metade intermediária dos dados, estendendo-se do primeiro quartil (Q1) ao terceiro quartil (Q3), o que fornece uma visão da faixa interquartil ($IQR = Q_3 - Q_1$). 
- A linha que atravessa a caixa indica a **mediana** ($Q_2$), oferecendo uma compreensão clara do valor central do conjunto de dados. 
- Os **"bigodes"** se estendem a partir da caixa até os valores máximos e mínimos.
   - Comumente, excluímos os valores atípicos (_outliers_ ou não) ao construirmos um boxplot. Então, os bigodes ficarão dentro de um limite especificado, 1,5 vezes a IQR acima do terceiro quartil ($Q_3$) e abaixo do primeiro quartil ($Q_1$). Assim, permite-se a identificação de valores extremos ou outliers, que são representados por pontos individuais fora dos bigodes.

Dessa forma, o boxplot **resume visualmente várias características** estatísticas de um conjunto de dados.

## 7.2 Definição

A Distribuição de Frequências é uma **representação tabular ou gráfica de dados que descreve a ocorrência de variáveis**. Cada entrada na tabela ou gráfico contém a quantidade ou a percentagem de ocorrências de um determinado intervalo de valores dentro do conjunto de dados total. Este conceito é uma parte fundamental da estatística e ajuda a fornecer um entendimento básico sobre a natureza e a distribuição dos dados coletados.

Com relação ao tipo de variável, a Distribuição de Frequências apresentam as seguintes características:

- **Variáveis Qualitativas**: Para dados categóricos ou qualitativos, a distribuição de frequência contará o número de vezes que cada categoria aparece no conjunto de dados.
- **Variáveis Quantitativas**: Para dados quantitativos, a distribuição de frequência contará o número de dados dentro de certos intervalos.

Por exemplo, suponha que um professor tenha registrado as notas das provas de uma turma. Nesse caso, temos uma variável quantitativa discreta, a nota, e ela varia de 0 a 100. O professor está interessado em entender a distribuição dessas notas para analisar o desempenho geral da turma. Dessa maneira, a distribuição de frequências pode ser tabulada assim:

| Intervalo de Notas | Frequência |
|--------------------|------------|
| 40-59              | 5          |
| 60-79              | 6          |
| 80-100             | 7          |

A interpretação dessa distribuição de frequências é a seguinte:

- 5 estudantes tiraram notas entre 40 e 59.
- 6 estudantes tiraram notas entre 60 e 79.
- 7 estudantes tiraram notas entre 80 e 100.


Portanto, uma distribuição de frequência, seja tabular ou gráfica, serve como uma ferramenta muito útil para resumir e interpretar grandes conjuntos de dados, tornando a informação compreensível e facilmente acessível. Geralmente, a representação da distribuição de frequências de grandes conjuntos de dados é feita através de histogramas ou gráficos de barras para melhor visualização. 


## 7.3 Histogramas

### Definição e Utilidade

Um histograma é uma representação gráfica de uma distribuição de frequência, amplamente utilizada para ilustrar a forma visual da distribuição dos dados. Ao utilizar retângulos de diferentes alturas, alinhados lado a lado, é possível visualizar com clareza as frequências (ou contagens) de dados dentro de intervalos específicos, também conhecidos como classes. A **base de cada retângulo representa uma classe**, enquanto **a altura do retângulo indica a frequência dos dados** nesse intervalo específico.

O propósito de um histograma vai além da simples contagem de observações. Ele serve como uma ferramenta para visualizar e entender a distribuição subjacente dos dados. Ao observar a **forma**, a **amplitude** e as **variações na altura** dos retângulos do histograma é possível identificar padrões, tais como simetria, assimetria, picos múltiplos (modas) e possíveis valores atípicos.

Por exemplo, em um histograma que representa as notas de alunos em uma prova, é possível identificar rapidamente a concentração de notas, os valores mais frequentes e a variação na distribuição das notas. Isso pode oferecer informações sobre o desempenho geral dos alunos, a dificuldade da prova, entre outros aspectos. Por exemplo vejamos o histograma a seguir do exemplo da Seção 7.1 das notas dos alunos.

````python
import matplotlib.pyplot as plt
import numpy as np

# Dados fornecidos
intervalos = ['30-49', '50-89', '90-100']
frequencia = [5, 6, 7]

# Cores para os intervalos
cores = ['antiquewhite', 'slateblue', 'aquamarine']

# Criando o gráfico de barras
plt.figure(figsize=(3,3))  # Ajustando o tamanho do gráfico conforme necessário
plt.bar(intervalos, frequencia, color=cores)
plt.title('Gráfico de Barras da Distribuição de Frequências das Notas dos Alunos')
plt.xlabel('Intervalo de Notas')
plt.ylabel('Frequência')

# Exibindo o gráfico
plt.tight_layout()  # Ajusta automaticamente os subplots para que caibam no espaço da figura
plt.show()
````

### Construindo um Histograma

Para construir um histograma, temos dois passos:

1. Os dados são primeiramente **organizados em classes de igual largura** ou de larguras variáveis. Estas classes são faixas de valores, e os dados são contados de acordo com a classe à qual pertencem, pois a escolha limites de classe apropriados podem ajudar a revelar a verdadeira natureza da distribuição dos dados, enquanto a escolha inadequada pode ocultar padrões importantes ou levar a interpretações equivocadas.
2. Em seguida, construímos **um gráfico onde cada classe é representada por uma barra** e a **altura de cada barra indica a frequência dos dados** que caem dentro daquela classe. As barras também podem representar a densidade de frequência, no caso de larguras variadas

Portanto, o eixo horizontal do histograma (eixo X) representa os intervalos de classes, enquanto o eixo vertical (eixo y) representa as frequências. 

Um aspecto importante sobre a construção de histograms são o **tamanho das classes**. Classes muito largas podem suavizar demais a distribuição, fazendo com que se percam detalhes sobre variações menores nos dados. Por outro lado, classes muito estreitas podem resultar em um histograma com muitas barras com pouca ou nenhuma observação, o que pode sugerir uma variabilidade excessiva que na realidade não existe nos dados.

Por exemplo, vamos agora visualizar a mesma distribuição de frequências do exemplo anteior com dois gráficos: um gráficos de barras e um histograma. Perceba que, **no histograma, a faixa de valor $50-89$ é a mais larga** de todas, pois o histograma respeita a escala de valores dos intervalos de dados (faixas), diferente do gráfico de barras. Em outras palavras, o histograma apresenta mais informações sobre o conjunto de dados do que o gráfico de barras.

````python
import matplotlib.pyplot as plt
import numpy as np

# Dados fornecidos
intervalos = ['30-49', '50-89', '90-100']
frequencia = [5, 6, 7]

# Cores para os intervalos
cores = ['antiquewhite', 'slateblue', 'aquamarine']

# Convertendo os intervalos em uma lista de limites para os bins
limites = [int(intervalo.split('-')[0]) for intervalo in intervalos] + [int(intervalos[-1].split('-')[1])]

# Criando subplots
fig, axs = plt.subplots(2, 1, figsize=(4,6))

# Gráfico de barras
axs[0].bar(intervalos, frequencia, color=cores)
axs[0].set_title('Gráfico de Barras da Distribuição de Frequências das Notas dos Alunos')
axs[0].set_xlabel('Intervalo de Notas')
axs[0].set_ylabel('Frequência')

# Histograma - Criando um conjunto de dados baseado nos intervalos e frequências
for i, (inicio, fim) in enumerate(zip(limites[:-1], limites[1:])):
    dados = np.random.uniform(inicio, fim, frequencia[i])
    axs[1].hist(dados, bins=[inicio, fim], color=cores[i], edgecolor='black', alpha=0.7, rwidth=0.85)

axs[1].set_title('Histograma da Distribuição de Frequências das Notas dos Alunos')
axs[1].set_xlabel('Notas')
axs[1].set_ylabel('Frequência')

# Exibindo os gráficos
plt.tight_layout()
plt.show()
````

## 7.4 Tipos de Distribuição de Frequências

Nesta seção, vamos aprofundar nos tipos de distribuições de frequência, discutindo sobre o comportamento das medidas resumo (como média, mediana e moda) em cada tipo de distribuição. Além disso, vamos relacionar essas distribuições com representações gráficas como o boxplot e as curvas de densidade de probabilidade. Curvas de densidade de probabilidade, ou simplesmente curvas de densidade, são uma representação suavizada do padrão subjacente dos dados em um histograma. Em outras palavras, essas curvas ajudam a visualizar a distribuição dos dados de uma maneira mais contínua e menos fragmentada do que as distribuições de frequência.


### Distribuição Uniforme

Em uma distribuição uniforme, todas as variáveis ou classes têm a mesma frequência ou probabilidade de ocorrência. Ou seja, em um conjunto de dados com distribuição uniforme, cada valor ocorre com a mesma frequência ou frequência similiar. Isso significa que a média, mediana e moda podem cair em qualquer ponto dentro de um memso intervalo, pois todos as observações têm frequências parecidas. 

O boxplot de uma distribuição uniforme mostrará uma caixa de tamanho uniforme se os dados estiverem perfeitamente distribuídos.

Em uma distribuição uniforme, a curva de densidade seria uma linha reta, indicando que cada valor tem a mesma probabilidade de ocorrência.

### Distribuição Normal

Uma distribuição normal é caracterizada pela sua forma de sino simétrica. Neste caso, a média, mediana e a moda são iguais e se localizam no centro do pico. 

O boxplot associado a uma distribuição normal será simétrico, indicando que os dados são distribuídos de maneira igual em torno da média, porém ele apresenta uma concentração maior entre Q1 e Q3 quando comparado com o boxplot da Distribuição Uniforme por exemplo.

Na distribuição normal, a curva de densidade tem a forma clássica de sino, indicando que a maioria das observações está concentrada em torno da média, e se torna menos frequente à medida que se afasta da média em ambas as direções.

### Distribuição Binomial

Em uma Distribuição Binomial, há apenas dois possíveis resultados ou classes, como "sucesso" e "fracasso". A média e mediana podem variar, mas geralmente ficam no centro se ambos os resultados forem igualmente prováveis. 

O boxplot pode variar dependendo das probabilidades de sucesso e fracasso, e a simetria será evidente se as probabilidades forem iguais.

Na distribuição binomial, a curva de densidade pode assumir diferentes formas dependendo da probabilidade de sucesso e do número de tentativas. Não é uma curva suave, mas mostra a probabilidade de cada número específico de sucessos.

### Distribuição Assimétrica (Skewness)

A Distribuição Assimétrica é inclinada, podendo ser positiva ou negativa. Em uma Distribuição Assimétrica positiva, a cauda à direita é mais longa, implicando que a média é maior que a mediana, que é, por sua vez, maior que a moda. Em uma Distribuição Assimétrica negativa, a cauda à esquerda é mais longa, indicando que a média é menor que a mediana, que é menor que a moda. 

O boxplot mostra claramente a assimaetria da Distribuição Assimétrica.

Em distribuições com skewness, a curva de densidade inclina-se para o lado onde os dados são menos frequentes, refletindo a assimetria dos dados.

### Distribuição Bimodal

Uma distribuição bimodal tem dois picos distintos, indicando que há duas modas. A média e a mediana situam-se entre esses dois picos. Na Distribuição Bimodal, a moda não serve como uma única medida representativa da centralidade dos dados, pois existem dois picos distintos ou modas. Nesses casos, é importante considerar ambas as modas para entender completamente a distribuição dos dados. 

O boxplot pode mostrar duas caixas distintas ou uma distância interquartil mais ampla, dependendo da separação das modas e da dispersão dos dados.

Em uma distribuição bimodal, a curva de densidade terá dois picos para representar os dois conjuntos distintos de dados dentro da amostra.


> **Exemplo**. Simulamos vários conjuntos de dados gerados aleatoriamente, cada um tem uma distribuição específica. Cada conjunto de dados é visualmente apresentada com seu histograma correspondente e boxplot . A partir desses gráficos, você pode observar claramente as características únicas de cada tipo de distribuição e como a média, a mediana e a moda, variam de acordo. Perceba que ambos gráficos para cada conjunto de dados estão alinhados em relação à sua escala no eixo X.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configuração para o estilo dos gráficos
sns.set(style="whitegrid")

sample_size = 10000

# Dados de exemplo
data = {
    'Uniforme': np.random.rand(sample_size),
    'Normal': np.random.randn(sample_size),
    'Binomial': np.random.binomial(n=10, p=0.5, size=sample_size),
    'Positivamente Inclinada': np.random.exponential(scale=1, size=sample_size),
    'Bimodal': np.concatenate((np.random.normal(loc=5, scale=2, size=sample_size),
                               np.random.normal(loc=10, scale=2, size=sample_size)))
}

# Criando gráficos
fig, axes = plt.subplots(2, len(data.keys()), figsize=(15,10))

# Iterando através dos dados para criar histogramas e boxplots
for i, dist in enumerate(data.keys()):
    sns.histplot(data[dist], bins=20, kde=True, ax=axes[0, i]).set(title=f'Distribuição {dist}', xlabel='Observações', ylabel='Frequência')
    sns.boxplot(x=data[dist], ax=axes[1, i]).set(xlabel='Observações')


# Ajustando os layout
plt.tight_layout()
plt.show()

````

## Tipos Essenciais de Gráficos e Suas Aplicações

A arte de visualizar dados envolve a escolha criteriosa dos formatos gráficos que melhor representam e comunicam as informações desejadas. Cada tipo de gráfico é intrínseco a determinadas formas de dados e objetivos de análise. Vejamos em detalhes alguns dos gráficos mais frequentemente empregados.

### Gráfico de Barras
Este gráfico é excepcional para representar e comparar quantidades ou frequências associadas a diferentes categorias. É visualmente intuitivo, tornando fácil a identificação de diferenças significativas entre categorias.

**Aplicação Prática:** Se desejarmos comparar o número de vendas de diferentes produtos em um mês, o gráfico de barras oferece uma visualização clara e concisa das vendas relativas.

````python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C']
vendas = [150, 200, 175]

plt.bar(produtos, vendas, color=['blue', 'green', 'red'])
plt.title('Vendas de Produtos em Janeiro')
plt.xlabel('Produtos')
plt.ylabel('Número de Vendas')
plt.show()
````

### Gráfico de Setores (Pizza)
Apesar de comum, o uso deve ser moderado para evitar complexidade visual. É eficiente para mostrar proporções relativas de categorias em um conjunto total.

**Aplicação Prática:** Ao visualizar a distribuição percentual de respostas de uma pesquisa de opinião, o gráfico de pizza destaca a proporção de cada resposta.

````python
import matplotlib.pyplot as plt
import squarify

# Dados existentes
respostas = ['Concordo', 'Discordo', 'Neutro']
percentuais = [50, 30, 20]
cores = ['gold', 'lightskyblue', 'lightcoral']

# Gráfico de pizza
plt.subplot(121)
plt.pie(percentuais, labels=respostas, colors=cores, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Gráfico de Pizza')

# Treemap
plt.subplot(122)
squarify.plot(sizes=percentuais, label=respostas, color=cores, alpha=0.7, text_kwargs={'fontsize':10})
plt.title('Treemap')
plt.axis('off')  # Desativa os eixos para o treemap

# Ajusta o layout para evitar sobreposição
plt.tight_layout()

# Mostra os gráficos
plt.show()
````

### Gráfico de Linhas
Este é indispensável para representar tendências temporais, mostrando a evolução de variáveis numéricas ao longo do tempo.

**Aplicação Prática:** Utilizado para traçar a trajetória de crescimento das vendas anuais, oferecendo informações sobre padrões sazonais ou anuais.

````python
anos = range(2018, 2023)
vendas_anuais = [200, 250, 300, 350, 400]

plt.plot(anos, vendas_anuais, marker='o', linestyle='-', color='purple')
plt.title('Crescimento das Vendas Anuais')
plt.xlabel('Ano')
plt.ylabel('Vendas (em milhares)')
plt.grid(True)
plt.show()
````

### Histograma
O histograma é útil para apresentar a distribuição de dados numéricos, oferecendo informações sobre a forma, amplitude e centralidade dos dados.

**Aplicação Prática:** Utilizado para entender a distribuição de pontuações em um teste, facilitando a identificação de padrões de desempenho dos estudantes.

````python
import numpy as np

pontuacoes = np.array([55, 60, 65, 70, 70, 75, 80, 85, 85, 90, 95, 100])

plt.hist(pontuacoes, bins=9, color='orange', alpha=0.7)
plt.title('Distribuição das Pontuações do Teste')
plt.xlabel('Pontuação')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
````

### Boxplot
Este gráfico oferece uma representação visual compacta das medidas estatísticas centrais e da dispersão, sendo altamente eficaz para identificar outliers.

**Aplicação Prática:** Em estudos clínicos, pode ser utilizado para visualizar a variação dos níveis de colesterol entre diferentes grupos etários.

````python
import seaborn as sns
import pandas as pd

colesterol = [180, 200, 220, 240, 260, 220, 205, 210, 180, 230]
grupo_etario = ['Jovens'] * 5 + ['Idosos'] * 5

data = pd.DataFrame({'Colesterol': colesterol, 'Grupo Etário': grupo_etario})
sns.boxplot(x='Grupo Etário', y='Colesterol', data=data)
plt.title('Variação dos Níveis de Colesterol entre Grupos Etários')
plt.show()
````

## Estratégias para Seleção Adequada de Gráficos

### Orientações Genéricas

A acuidade na seleção do gráfico certo é um passo fundamental na comunicação eficaz dos dados. A seguir, estão estratégias para a escolha de um gráfico:

- **Quantidades e Proporções:** Gráficos de barras e setores são ideais para representar e comparar quantidades ou proporções distintas entre diferentes categorias.
- **Tendências Temporais:** Prefira gráficos de linhas, pois eles são excepcionais para representar a evolução dos dados ao longo do tempo. 
- **Relações entre Variáveis:** Gráficos de dispersão, pois conseguimos dentificar padrões e correlações entre variáveis.
- **Distribuição de Dados:**  Histogramas são fundamentais para desvendar a estrutura dos dados e identificar padrões.
- **Dispersão e Centralidade:**  Boxplots sintetizam estes elementos eficientemente, oferecendo uma visão concisa das medidas centrais e da dispersão dos dados. 

Embora essas diretrizes sejam úteis, é fundamental considerar o público-alvo e o contexto. A complexidade, a interpretabilidade e a estética visual desempenham papéis cruciais na eficácia da comunicação dos dados. Sempre teste e revise suas visualizações para garantir que sejam intuitivas, informativas e engajadoras para o seu público específico.

Por fim, a escolha de um gráfico possuem também um caráter subjetivo e algumas perguntas podem nos ajudar a nortear a escolha dos gráficos:

- Este gráfico ajuda a responder às minhas perguntas?
- Ele representa claramente os dados?
- As informações são facilmente interpretáveis para o público-alvo?

# 9. Interpretando Dados e Contando uma História

Na seção anterior, vimos que a visualização de dados é a ponte entre os dados brutos e a informação útil, diminuindo a complexidade da representação numérica através da representação visual, um passo fundamental para transformar números em narrativas. Agora, vamos finalmente aprender como interpretá-los e como comunicar nossas descobertas através de narrativas, ou seja, contando histórias. 

## Abordagem da Pixar para Criar Apresentações (_Pitch_)

Uma forma bastante simples e eficaz de contar histórias é utilizando a téncnica de narrativa da Pixar [Leon, 2019]. Resumidamente, essa abordagem propõe as seguintes etapas: contextualização, introdução do problema, apresentação da solução, consequências diretas da solução e beneficios 

```
Era uma vez... um problema.
Todos os dias... o problema impactava negativamente as vidas das pessoas.
Até que um dia... surgiu uma solução para o problema.
Por causa disso... uma organização foi fundada para resolver o problema.
Por causa disso... a organização começou a aplicar a solução ao problema.
Até que finalmente... o peso do problema foi aliviado.
```

## Exemplo: Desempenho de uma Empresa de Comunicação Social em Redes Sociais

### Contexto 

Uma empresa de comunicação social enfrentava um desafio significativo: o engajamento com o público estava estagnado, e as métricas de interação eram baixas. A estratégia de conteúdo existente não estava dando certo e os usuários encontravam as mensagens da marca distantes e pouco autênticas. A falta de uma conexão humana e significativa era evidente, resultando em baixos níveis de interação, de compartilhamentos e de retornos positivos nas plataformas de mídia social.

Era preciso reavaliar e transformar a abordagem de comunicação para reverter esse cenário através de uma nova estratégia de comunicação que se conectasse de forma mais íntima e genuína com o público. Então. a velha prática de transmitir mensagens unidimensionais foi substituída por uma narrativa envolvente e dialogada, promovendo a interação e a participação ativa do público. 

A fim de mensurar o desempenho dessa nova estratágia, foram utilizadas as seguntes variáveis

- **Engajamento do público***. O engajamento do público é um termômetro direto da ressonância e relevância do conteúdo: quanto mais os usuários interagem, mais o conteúdo é percebido como valioso.
- **Sentimentos positivos** Sentimentos positivos são indicativos da qualidade da conexão emocional estabelecida, um aspecto muito importante na construção de relações autênticas e duradouras com o público.
- **Compartilhamentos**. Já os compartilhamentos refletem a propensão dos usuários em associar sua identidade ao conteúdo da marca, amplificando o alcance e a visibilidade das mensagens.

Após fazer a anális de dados, foi escolhido o gráfico de curvas para representar os dados baseados nos critérios explicados anteriormente. Dessa maneira, o gráfico a seguir apresenta a evolução do engajamento do público, sentimentos positivos e compartilhamentos ao longo de quatro meses após a implementação de uma nova estratégia de comunicação social. 


````python
import matplotlib.pyplot as plt
import numpy as np

# Dados hipotéticos ao longo de 4 meses
meses = np.array(['Mês 1', 'Mês 2', 'Mês 3', 'Mês 4'])
engajamento = np.array([23, 45, 47, 72])  # Dados mais aleatórios para o engajamento do público
sentimentos_positivos = np.array([18, 38, 56, 63])  # Dados mais aleatórios para sentimentos positivos
compartilhamentos = np.array([6, 12, 35, 80])  # Dados mais aleatórios para compartilhamentos

# Criando o gráfico de linhas
plt.figure(figsize=(10, 7))

plt.plot(meses, engajamento, marker='o', label='Engajamento do Público', color='b')
plt.plot(meses, sentimentos_positivos, marker='o', label='Sentimentos Positivos', color='g', linestyle=':')
plt.plot(meses, compartilhamentos, marker='o', label='Compartilhamentos', color='r', linestyle='--')

# Personalizando os rótulos e título
plt.xlabel('Meses após Nova Estratégia')
plt.ylabel('Percentual (%)')
plt.title('Avaliação da Nova Estratégia de Comunicação Social')

# Adicionando a legenda
plt.legend()

# Exibindo o gráfico
plt.grid(True)
plt.show()
````

### Ressurgimento Conectado: A Jornada da Transformação Comunicativa

No gráfico anterior, é visível um crescimento consistente em todas as três métricas, indicando uma resposta positiva à nova estratégia. Agora, vamos contar essa história utilizando a abordagem para criar apresentações da Pixar.


> Era uma vez uma empresa de comunicação social que enfrentava um desafio crítico. > 
> Até que um dia, aconteceu uma transformação. A empresa, reconhecendo a necessidade de mudança, reformulou sua estratégia de comunicação. Introduziu narrativas envolventes, promovendo uma comunicação bidirecional que ecoava a humanidade e a autenticidade.
> 
> Por causa disso, nasceu uma nova era de engajamento. Os gráficos mostraram uma narrativa de triunfo, o engajamento do público elevou-se, sentimentos positivos floresceram e os compartilhamentos cresceram exponencialmente. > 
> Por causa disso, a empresa não era apenas uma transmissora de mensagens, mas uma entidade fortemente conectada com o seu público. As mensagens não eram mais monólogos, mas diálogos ricos que alimentavam uma comunidade em expansão.

# 10. Passo a Passo Resumido

Após este estudo introdutório, proponho uma sugestão de lista de passos a serem seguidos para realizar uma análise exploratória de dados. Lembre-se de que a AED é um processo iterativo e flexível, onde frequentemente você precisará refazer algumas etapas à medida que novas informações e descobertas surgirem.

É muito importante ressaltar também que **esse compilado de ações não substitui o estudo sobre análise exploratória da dados**, mas sim uma forma sistemática de passos que podem ser seguidos para facilitar a análise. De fato, esse passo a passo muito provavelmente não será útil para quem não tiver alguma noção mínima de análise exploratória da dados; seja adquirida nesse curso ou em qualquer outra fonte de estudo.


1. **Definição de Análise Exploratória de Dados**: Inicie compreendendo o que é a Análise Exploratória de Dados (AED), sua importância e como ela se encaixa no contexto da sua pesquisa ou projeto. Entenda que a AED é uma etapa preliminar fundamental que permite familiarizar-se com os dados, identificar padrões, anomalias, testar hipóteses e verificar suposições.

2. **Modelagem**: Antes de analisar os dados, defina claramente os objetivos da sua análise. Quais perguntas você está tentando responder? Isso ajudará a orientar suas técnicas de modelagem e escolha de ferramentas.

3. **Variáveis: Tipos, Dependência, Correlação**: Identifique as variáveis do conjunto de dados e classifique-as em quantitativas (numéricas) e qualitativas (categóricas). Avalie a relação entre as variáveis para entender dependências e possíveis correlações. Ferramentas como gráficos de dispersão e coeficientes de correlação podem ajudar nessa avaliação.

4. **Amostragem: Tipos, Técnicas, Erros**: Se estiver trabalhando com uma amostra dos dados, escolha adequadamentes o método de amostragem a ser empregado. Para isso, avalie qual técnica de amostragem vai se enquadrar melhor às fontes dos dados, às formas possíveis de coletar os dados, sempre tendo como Norte a(s) pergunta(s) que pretende responder para tentar coletar uma amostra mais representativa possível dos dados. 

5. **Medidas Resumo de Posição: Média, Mediana e Moda**: Utilize essas medidas para obter uma compreensão geral da distribuição dos seus dados. Cada uma dessas medidas fornece informações diferentes sobre a tendência central dos dados.

6. **Medidas Resumo de Dispersão**: Complemente as medidas de posição com medidas de dispersão como variância, desvio padrão, amplitude e quartis por exemplo. Isso ajudará a entender a variabilidade nos seus dados e se a média representa bem os dados por exemplo.

7. **Distribuição de Frequências de Dados**: Construa tabelas de frequência e histogramas para visualizar como os dados estão distribuídos. Isso pode ajudar a identificar padrões, assimetrias e possíveis outliers.

8. **Visualização de Dados**: Utilize gráficos e ferramentas de visualização como gráficos de barras, box plots, gráficos de linha, mapas de calor, etc. para explorar visualmente os dados, detectar tendências, padrões e anomalias.

9. **Interpretando Dados e Contando uma História**: Combine os aprendizado obtido nas etapas anteriores e construa uma narrativa coerente sobre seus dados. A interpretação dos resultados deve ser feita com cuidado, considerando o contexto dos dados, as limitações da análise e as possíveis implicações das descobertas.
