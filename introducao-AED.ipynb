## 1. Definição de Análise Exploratória de Dados (AED)

A Análise Exploratória de Dados (AED) é uma etapa elemntar na análise de dados que envolve uma série de técnicas para visualizar, resumir e entender a informação que nossos dados estão nos fornecendo. Esta fase da análise de dados é fundamental para gerar ideias (insights), testar hipóteses e verificar suposições.

A AED é uma abordagem que enfatiza a exploração visual dos dados, em vez de testes estatísticos formais ou modelagem preditiva. Seu objetivo principal é revelar a estrutura dos dados, detectar outliers ou anomalias, testar suposições, desenvolver modelos intuitivos e determinar a seleção adequada de ferramentas e técnicas para análises futuras.

Agora, vamos primeiro entender o que são variáveis e seus respectivos tipos.

## 2. Tipos de Variáveis

No contexto da Análise Exploratória de Dados, uma variável é uma característica, número ou quantidade que pode ser medida ou contada. As variáveis podem assumir vários valores e cada valor da variável pode corresponder a uma observação ou medida diferente.

As variáveis podem ser categorizadas principalmente em dois tipos: qualitativas e quantitativas.

1. **Variáveis Qualitativas**: Também conhecidas como variáveis categóricas, elas são aquelas que expressam uma característica ou qualidade que não pode ser medida numericamente. Elas podem ser divididas em:

   - **Variáveis Nominais**: não existe uma ordem ou ranking entre as categorias. Exemplos incluem gênero, cor dos olhos, tipo de material, etc.
   
   - **Variáveis Ordinais**: as categorias têm uma ordem ou ranking específico. Exemplos incluem classificações de filmes, níveis de educação, etc.

2. **Variáveis Quantitativas**: Estas são variáveis que podem ser medidas numericamente. Elas podem ser divididas em:

   - **Variáveis Discretas**: são contáveis e normalmente são números inteiros. Exemplos incluem o número de animais de estimação em uma casa, o número de alunos em uma sala de aula, etc.
   
   - **Variáveis Contínuas**: podem assumir qualquer valor dentro de um intervalo específico. Exemplos incluem altura, peso, tempo, temperatura, etc.

## 3. Conceitos Básicos em Estatística

A Estatística é uma disciplina que se ocupa da coleta, análise, interpretação, apresentação e organização de dados. A seguir, estão alguns conceitos básicos que são fundamentais para a compreensão da estatística e, consequentemente, para a análise exploratória de dados.

### População e Amostra

População é o conjunto completo de observações, indivíduos, itens ou eventos que queremos estudar, enquanto a amostra é um subconjunto dessa população. Por exemplo, se quisermos estudar a altura dos estudantes universitários no Brasil, a população seria todos os estudantes universitários no Brasil. Já uma amostra poderia ser os discentes matriculados na turma de Estatística do curso de Relações Públicas da Universidade Federal de Alagoas por exemplo.

> **Exemplo**. Suponha que temos a altura de todos os estudantes universitários no Brasil e queremos tirar uma amostra desses dados para estudos adicionais. Usaremos a biblioteca `numpy` para simular os dados da população e a função `random.choice` para selecionar uma amostra aleatória. Para fins de ilustração, suponhamos que a altura dos estudantes universitários no Brasil siga uma distribuição normal com uma média de 170 cm e um desvio padrão de 10 cm.
> 
> Estamos gerando uma "população" de alturas de estudantes universitários, e então tirando uma "amostra" dessa população. Depois, calculamos a média da altura para a população e para a amostra e as imprimimos. 
> 
> Note que a média da amostra pode não ser exatamente igual à média da população, devido à variação aleatória inerente ao processo de amostragem. No entanto, se a amostra for suficientemente grande e for selecionada aleatoriamente, a média da amostra deve ser uma boa estimativa da média da população.

```python
# Importando as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt

# Definindo a semente do gerador de números aleatórios para reprodutibilidade
np.random.seed(0)

# Gerando dados da população
altura_populacao = np.random.normal(170, 10, 1000000) # Suponha que temos 1 milhão de estudantes universitários no Brasil

# Selecionando uma amostra
amostra = np.random.choice(altura_populacao, 100) # Selecionamos 100 estudantes para a nossa amostra

# Calculando a média da população e da amostra
media_populacao = altura_populacao.mean()
media_amostra = amostra.mean()

print(f"Média da altura da população: {media_populacao:.2f} cm")
print(f"Média da altura da amostra: {media_amostra:.2f} cm")

# Criando o histograma para a população
plt.figure(figsize=(12,6))
plt.hist(altura_populacao, bins=100, alpha=0.5, label='População', density=True)

# Adicionando a legenda
plt.legend(loc='upper right')

# Adicionando títulos e labels
plt.title('Distribuição da Altura da População')
plt.xlabel('Altura (cm)')
plt.ylabel('Frequência Normalizada')

# Mostrando o gráfico
plt.show()

# Criando o histograma para a amostra
plt.figure(figsize=(12,6))
plt.hist(amostra, bins=100, alpha=0.5, label='Amostra', density=True)

# Adicionando a legenda
plt.legend(loc='upper right')

# Adicionando títulos e labels
plt.title('Distribuição da Altura da Amostra')
plt.xlabel('Altura (cm)')
plt.ylabel('Frequência Normalizada')

# Mostrando o gráfico
plt.show()
```

### Variáveis Dependentes e Independentes

Na análise de dados, muitas vezes estamos interessados em entender a relação entre duas variáveis. A **variável dependente** é o que estamos tentando prever ou explicar, enquanto a **variável independente** é a que acreditamos ter algum efeito sobre a variável dependente. 

Por exemplo, se quisermos prever o desempenho de um aluno com base na quantidade de horas que ele estuda por semana, o desempenho do aluno seria a variável dependente e a quantidade de horas de estudo seria a variável independente. Acreditamos que o desempenho do aluno (variável dependente) depende de, ou é influenciado por, a quantidade de horas que ele estuda (variável independente). Da mesma forma, em um experimento científico para testar os efeitos de um medicamento, a variável independente seria a administração do medicamento, enquanto a variável dependente poderia ser a melhora ou piora dos sintomas.

> **Exemplo**. Vamos prever uma variável contínua (o desempenho do aluno) com base em outra variável contínua (as horas de estudo). Podemos usar uma regressão linear simples para este exemplo. Aqui está um exemplo de código usando a biblioteca `sklearn` do Python.
> 
> No exemplo acima, estamos simulando um conjunto de dados onde o desempenho do aluno é proporcional às horas de estudo, com algum ruído adicionado para tornar os dados mais realistas. Em seguida, treinamos um modelo de regressão linear no conjunto de treinamento e fazemos previsões no conjunto de teste. Por fim, comparamos as previsões do modelo com os valores reais e plotamos a linha de regressão.

```python
# Importando as bibliotecas necessárias
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt
import pandas as pd

# Definindo a semente do gerador de números aleatórios para reprodutibilidade
np.random.seed(0)

# Gerando dados fictícios
horas_estudo = np.random.normal(10, 2, 100) # Gerando horas de estudo aleatórias
desempenho_aluno = horas_estudo * 2 + np.random.normal(0, 2, 100) # Desempenho é proporcional às horas de estudo, adicionamos algum ruído

# Reformatando os dados para usar com sklearn
horas_estudo = horas_estudo.reshape(-1,1)
desempenho_aluno = desempenho_aluno.reshape(-1,1)

# Dividindo os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(horas_estudo, desempenho_aluno, test_size=0.2, random_state=0)

# Treinando o modelo
regressor = LinearRegression()  
regressor.fit(X_train, y_train) # Treinando o algoritmo

# Fazendo previsões no conjunto de teste
y_pred = regressor.predict(X_test)

# Comparando a previsão com os dados reais
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
print(df)

# Plotando a linha de regressão
plt.scatter(X_test, y_test,  color='gray')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.show()
```

### Distribuição de Dados

Este é um conceito fundamental que descreve a frequência de diferentes valores ou intervalos de valores em um conjunto de dados. A forma da distribuição pode nos dizer muito sobre nossos dados, incluindo a média, mediana e modo, a presença de outliers, etc. Por exemplo, imagine que você esteja analisando os resultados de uma prova aplicada em uma sala de aula. A pontuação de cada estudante seria um valor dentro do seu conjunto de dados. A distribuição desses valores poderia revelar se a maioria dos alunos teve um desempenho similar (resultando em uma distribuição normal ou simétrica), ou se houve muitas variações nas notas (resultando em uma distribuição assimétrica).

Se a distribuição das notas for normal, a média, a mediana e a moda terão aproximadamente o mesmo valor, situado no centro da distribuição. Isso significaria que a maioria dos alunos obteve notas próximas à média.

Por outro lado, se a distribuição das notas for assimétrica, por exemplo, inclinada para a direita (também conhecida como distribuição positiva ou de cauda longa à direita), isso significaria que a maioria dos alunos obteve notas mais baixas, enquanto poucos alunos obtiveram notas muito altas. Nesse caso, a média seria maior que a mediana e a moda, pois é influenciada pelos valores mais altos.

A análise da distribuição de dados também pode revelar a presença de outliers, que são valores atipicamente altos ou baixos. No exemplo das notas, um outlier poderia ser um aluno que teve uma nota muito baixa ou muito alta em comparação com o restante da turma.

Portanto, entender a distribuição dos dados é crucial para a interpretação adequada dos resultados da análise e para a tomada de decisões baseadas em dados.

Estes são apenas alguns dos conceitos básicos em estatística. Há muitos outros, mas com esses conceitos fundamentais, você já terá uma base sólida para começar a entender e a explorar seus dados.

> **Exemplo**. Vamos simular esse cenário criando um conjunto de dados de notas de alunos e visualizando a distribuição dessas notas. Para isso, usaremos o pacote `numpy` para geração de dados e o pacote `matplotlib` para a visualização dos dados. 
> 
> Neste exemplo, criamos uma distribuição normal de notas de alunos com uma média de 70 e um desvio padrão de 10. Então, visualizamos a distribuição das notas usando um histograma. A saída deste código será o gráfico do histograma, bem como a média e a mediana das notas.
> 
> Lembre-se, este é apenas um exemplo de simulação. No mundo real, você provavelmente estaria trabalhando com conjuntos de dados reais coletados através de pesquisas ou observações.

```python
# Importando as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt

# Definindo a semente do gerador de números aleatórios para reprodutibilidade
np.random.seed(0)

# Gerando um conjunto de dados representando as notas dos alunos
# Aqui estamos simulando uma distribuição normal de notas, com média 70 e desvio padrão 10
notas = np.random.normal(70, 10, 100)

# Criando o histograma
plt.hist(notas, bins=10, edgecolor='black')

# Adicionando títulos e labels
plt.title('Distribuição de Notas dos Alunos')
plt.xlabel('Nota')
plt.ylabel('Número de Alunos')

# Mostrando o gráfico
plt.show()

# Imprimindo as estatísticas
print(f'Média das notas: {notas.mean():.2f}')
print(f'Mediana das notas: {np.median(notas):.2f}')
```
